## 📝 Introdução

A **Programação Orientada a Objetos (POO)** é um paradigma de programação que organiza o código em torno de **objetos** — entidades que combinam **dados (atributos)** e **comportamentos (métodos)**. Python é uma linguagem multiparadigma que oferece suporte completo a POO.

Enquanto a programação procedural se baseia em funções que manipulam dados, a POO agrupa dados e funções relacionados em **classes**, facilitando a reutilização, organização e manutenção do código.

## 📚 Explicação Detalhada

### Classe e Objeto

Uma **classe** é um modelo (molde) que define a estrutura e o comportamento de um objeto. Um **objeto** é uma instância específica criada a partir desse molde.

```python
class Carro:
    pass

meu_carro = Carro()  # objeto (instância)
```

### Método `__init__` (Construtor)

O método `__init__` é executado automaticamente quando um objeto é criado. Ele inicializa os atributos da instância.

### Parâmetro `self`

`self` representa a própria instância do objeto. É o primeiro parâmetro de qualquer método de instância e permite acessar os atributos e outros métodos do objeto.

### Atributos de Classe vs Atributos de Instância

- **Atributos de instância**: definidos dentro de `__init__`, pertencem a cada objeto individualmente.
- **Atributos de classe**: definidos diretamente no corpo da classe, são compartilhados por todas as instâncias.

### Métodos de Instância

São funções definidas dentro da classe que operam sobre os dados da instância. Sempre recebem `self` como primeiro argumento.

### Método `__str__`

Define a representação em string do objeto (usada por `print()` e `str()`).

## 💻 Exemplos

````python
# Exemplo 1: Criando uma classe simples
class Cachorro:
    pass

rex = Cachorro()
print(type(rex))
print(rex)
````

````python
# Exemplo 2: Classe com __init__ e self
class Cachorro:
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade

rex = Cachorro("Rex", 3)
print(f"Nome: {rex.nome}, Idade: {rex.idade}")
````

````python
# Exemplo 3: Atributos de classe vs instância
class Cachorro:
    especie = "Canis familiaris"  # atributo de classe

    def __init__(self, nome, idade):
        self.nome = nome          # atributo de instância
        self.idade = idade        # atributo de instância

rex = Cachorro("Rex", 3)
bob = Cachorro("Bob", 5)
print(rex.nome, rex.especie)
print(bob.nome, bob.especie)
Cachorro.especie = "Canis lupus"
print(rex.especie)  # mudou para todas as instâncias
````

````python
# Exemplo 4: Métodos de instância
class Cachorro:
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade

    def latir(self):
        return f"{self.nome} diz: Au au!"

    def aniversario(self):
        self.idade += 1
        return f"{self.nome} fez {self.idade} anos!"

rex = Cachorro("Rex", 3)
print(rex.latir())
print(rex.aniversario())
print(rex.aniversario())
````

````python
# Exemplo 5: Método __str__
class Cachorro:
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade

    def __str__(self):
        return f"Cachorro(nome={self.nome}, idade={self.idade})"

rex = Cachorro("Rex", 3)
print(rex)          # usa __str__
print(str(rex))     # mesma coisa
print(repr(rex))    # representação padrão (fallback)
````

````python
# Exemplo 6: Classe completa - Conta Bancária
class ContaBancaria:
    banco = "Python Bank"  # atributo de classe

    def __init__(self, titular, saldo=0):
        self.titular = titular
        self.__saldo = saldo  # encapsulamento (será visto na Parte 2)

    def depositar(self, valor):
        if valor > 0:
            self.__saldo += valor
            return f"Depósito de R$ {valor:.2f} realizado"
        return "Valor inválido"

    def sacar(self, valor):
        if 0 < valor <= self.__saldo:
            self.__saldo -= valor
            return f"Saque de R$ {valor:.2f} realizado"
        return "Saldo insuficiente ou valor inválido"

    def __str__(self):
        return f"Conta de {self.titular} - Saldo: R$ {self.__saldo:.2f}"

conta = ContaBancaria("João", 1000)
print(conta)
print(conta.depositar(500))
print(conta)
print(conta.sacar(200))
print(conta)
print(f"Banco: {ContaBancaria.banco}")
````

````python
# Exemplo 7: Vários objetos independentes
class Aluno:
    def __init__(self, nome, matricula):
        self.nome = nome
        self.matricula = matricula
        self.notas = []

    def adicionar_nota(self, nota):
        self.notas.append(nota)

    def media(self):
        if not self.notas:
            return 0
        return sum(self.notas) / len(self.notas)

    def __str__(self):
        return f"Aluno: {self.nome} ({self.matricula}) - Média: {self.media():.1f}"

aluno1 = Aluno("Ana", "2024001")
aluno2 = Aluno("Carlos", "2024002")
aluno1.adicionar_nota(8.5)
aluno1.adicionar_nota(9.0)
aluno2.adicionar_nota(7.0)
aluno2.adicionar_nota(6.5)
print(aluno1)
print(aluno2)
````

## ✅ Exercícios Resolvidos

````python
# Exercício 1: Criar classe Livro com titulo, autor, ano e método info()
class Livro:
    def __init__(self, titulo, autor, ano):
        self.titulo = titulo
        self.autor = autor
        self.ano = ano

    def info(self):
        return f"'{self.titulo}' por {self.autor} ({self.ano})"

livro = Livro("1984", "George Orwell", 1949)
print(livro.info())
````

````python
# Exercício 2: Classe Retângulo com área e perímetro
class Retangulo:
    def __init__(self, largura, altura):
        self.largura = largura
        self.altura = altura

    def area(self):
        return self.largura * self.altura

    def perimetro(self):
        return 2 * (self.largura + self.altura)

    def __str__(self):
        return f"Retângulo {self.largura}x{self.altura}"

r = Retangulo(5, 3)
print(r)
print(f"Área: {r.area()}")
print(f"Perímetro: {r.perimetro()}")
````

````python
# Exercício 3: Classe Pessoa com contador de instâncias (atributo de classe)
class Pessoa:
    total_pessoas = 0

    def __init__(self, nome):
        self.nome = nome
        Pessoa.total_pessoas += 1

    def __str__(self):
        return f"Pessoa: {self.nome}"

p1 = Pessoa("Maria")
p2 = Pessoa("João")
p3 = Pessoa("Pedro")
print(p1)
print(f"Total de pessoas criadas: {Pessoa.total_pessoas}")
````

## 📌 Resumo

- **Classe**: molde que define atributos e métodos (`class Nome:`)
- **Objeto**: instância concreta de uma classe
- **`__init__`**: construtor que inicializa atributos da instância
- **`self`**: referência à própria instância, primeiro parâmetro de métodos
- **Atributos de classe**: compartilhados por todas as instâncias
- **Atributos de instância**: específicos de cada objeto
- **Métodos de instância**: funções que operam sobre a instância
- **`__str__`**: representação textual do objeto para `print()`

## 💡 Curiosidades

- Python permite **modificar atributos diretamente** (não há `private` de verdade), mas convenções como `_` (protegido) e `__` (privado) existem.
- O nome `self` é uma **convenção**, não uma palavra reservada. Você poderia usar `this` ou qualquer outro nome, mas usar `self` é padrão da comunidade.
- Python suporta herança múltipla, diferentemente de Java e C#.
- O paradigma POO foi popularizado por linguagens como Smalltalk (1970s) e C++ (1980s).

## ✅ Boas Práticas

- Nomeie classes usando `PascalCase` (ex: `MinhaClasse`)
- Nomeie métodos e atributos usando `snake_case` (ex: `meu_metodo()`)
- Sempre use `self` como primeiro parâmetro dos métodos de instância
- Inicialize todos os atributos no `__init__`, mesmo que com `None`
- Implemente `__str__` para facilitar debugging
- Mantenha cada classe focada em uma única responsabilidade
- Use atributos de classe para dados compartilhados, não para valores específicos de instância

## ⚠️ Erros Comuns

| Erro | Exemplo | Correção |
|------|---------|----------|
| Esquecer `self` no método | `def metodo():` | `def metodo(self):` |
| Esquecer `self` ao acessar atributo | `nome` em vez de `self.nome` | Usar `self.nome` |
| Confundir classe e instância | `minha_classe.atributo` (instância) vs `Classe.atributo` (classe) | Entender a diferença |
| Não chamar `__init__` da superclasse | Em herança, esquecer `super().__init__()` | Chamar `super().__init__()` |
| Modificar atributo de classe via instância | `obj.atributo_classe = valor` (cria atributo de instância) | Usar `Classe.atributo_classe = valor` |
| Definir atributo fora de `__init__` | `def metodo(self): self.x = 1` (funciona, mas é confuso) | Definir todos em `__init__` |
