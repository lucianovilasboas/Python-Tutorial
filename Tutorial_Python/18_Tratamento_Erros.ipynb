## 🎯 Objetivos

- Compreender o que são exceções e por que tratá-las
- Usar `try`, `except`, `else` e `finally` para controle de erros
- Tratar múltiplas exceções em um mesmo bloco
- Criar e lançar exceções customizadas com `raise`
- Conhecer exceções comuns: `ValueError`, `TypeError`, `FileNotFoundError`
- Usar `assert` para validações em tempo de desenvolvimento

## 📝 Introdução

Erros acontecem! Seja por entrada inválida do usuário, arquivo inexistente ou problema de conexão, seu programa precisa lidar com imprevistos de forma elegante. O Python oferece o bloc `try/except` para capturar e tratar exceções, além do `assert` para verificar condições que devem ser verdadeiras. Aprender a tratar erros corretamente é essencial para criar programas robustos e profissionais.

## 📚 Explicação Detalhada

### Estrutura try/except

```python
try:
    # Código que pode gerar erro
except TipoDoErro:
    # Código executado se o erro ocorrer
else:
    # Código executado se NENHUM erro ocorrer
finally:
    # Código executado SEMPRE (com ou sem erro)
```

### Principais Exceções Internas

| Exceção | Descrição | Exemplo |
|---------|-----------|---------|
| `ValueError` | Argumento com valor inválido | `int("abc")` |
| `TypeError` | Operação com tipo incompatível | `"texto" + 5` |
| `FileNotFoundError` | Arquivo não encontrado | `open("inexistente.txt")` |
| `ZeroDivisionError` | Divisão por zero | `10 / 0` |
| `IndexError` | Índice fora da lista | `lista[100]` |
| `KeyError` | Chave inexistente em dict | `dict["chave"]` |
| `ImportError` | Módulo não encontrado | `import modulo_inexistente` |

### Assert

```python
assert condição, "Mensagem de erro"
```
Se a condição for `False`, levanta `AssertionError`.

## 💻 Exemplos

### Exemplo 1: Try/Except básico

````python
try:
    numero = int(input("Digite um número: "))
    print(f"Você digitou {numero}")
except ValueError:
    print("Erro: você não digitou um número válido!")
````

### Exemplo 2: Múltiplos excepts

````python
try:
    a = int(input("Numerador: "))
    b = int(input("Denominador: "))
    resultado = a / b
    print(f"Resultado: {resultado}")
except ValueError:
    print("Erro: digite apenas números inteiros!")
except ZeroDivisionError:
    print("Erro: não é possível dividir por zero!")
````

### Exemplo 3: Tratando múltiplas exceções em uma linha

````python
try:
    lista = [1, 2, 3]
    indice = int(input("Índice (0-2): "))
    print(f"Valor: {lista[indice]}")
except (ValueError, IndexError) as erro:
    print(f"Erro do tipo {type(erro).__name__}: {erro}")
````

### Exemplo 4: Try/Except/Else/Finally

````python
try:
    arquivo = open("dados.txt", "r", encoding="utf-8")
    conteudo = arquivo.read()
except FileNotFoundError:
    print("Arquivo não encontrado!")
else:
    print("Arquivo lido com sucesso!")
    print(conteudo[:100])
finally:
    print("\nBloco finally: sempre executado.")
    try:
        arquivo.close()
    except NameError:
        pass
````

### Exemplo 5: raise - Lançando exceções

````python
def dividir(a, b):
    if b == 0:
        raise ValueError("Denominador não pode ser zero!")
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Ambos os valores devem ser números!")
    return a / b

try:
    print(dividir(10, 2))
    print(dividir(10, 0))
except ValueError as e:
    print(f"Erro: {e}")
except TypeError as e:
    print(f"Erro: {e}")
````

### Exemplo 6: Capturando qualquer exceção (genérico)

````python
try:
    import modulo_inexistente
except Exception as erro:
    print(f"Ocorreu um erro: {type(erro).__name__} - {erro}")
finally:
    print("Bloco finally executado.")
````

### Exemplo 7: Criando exceção customizada

````python
class SaldoInsuficienteError(Exception):
    "Exceção para saldo insuficiente em conta bancária"
    pass

def sacar(saldo, valor):
    if valor > saldo:
        raise SaldoInsuficienteError(f"Saldo: R$ {saldo:.2f}, Saque: R$ {valor:.2f}")
    return saldo - valor

try:
    novo_saldo = sacar(1000, 1500)
    print(f"Novo saldo: R$ {novo_saldo:.2f}")
except SaldoInsuficienteError as e:
    print(f"Saque negado: {e}")
````

### Exemplo 8: Usando assert

````python
def calcular_media(notas):
    assert len(notas) > 0, "Lista de notas não pode estar vazia!"
    assert all(0 <= n <= 10 for n in notas), "Todas as notas devem estar entre 0 e 10!"
    return sum(notas) / len(notas)

try:
    print(calcular_media([7.5, 8.0, 9.5]))
    print(calcular_media([]))
except AssertionError as e:
    print(f"AssertionError: {e}")
````

## ✅ Exercícios Resolvidos

### Exercício 1: Calculadora segura com try/except

````python
def calculadora_segura():
    try:
        a = float(input("Primeiro número: "))
        op = input("Operação (+, -, *, /): ")
        b = float(input("Segundo número: "))

        if op == "+":
            print(f"Resultado: {a + b}")
        elif op == "-":
            print(f"Resultado: {a - b}")
        elif op == "*":
            print(f"Resultado: {a * b}")
        elif op == "/":
            print(f"Resultado: {a / b}")
        else:
            print("Operação inválida!")
    except ValueError:
        print("Erro: digite números válidos!")
    except ZeroDivisionError:
        print("Erro: divisão por zero não permitida!")

calculadora_segura()
````

### Exercício 2: Validação de senha com raise

````python
class SenhaFracaError(Exception):
    pass

def validar_senha(senha):
    if len(senha) < 8:
        raise SenhaFracaError("A senha deve ter no mínimo 8 caracteres!")
    if not any(c.isdigit() for c in senha):
        raise SenhaFracaError("A senha deve conter pelo menos um número!")
    if not any(c.isupper() for c in senha):
        raise SenhaFracaError("A senha deve conter pelo menos uma letra maiúscula!")
    return True

senhas_teste = ["abc123", "Senha1234", "12345678"]
for s in senhas_teste:
    try:
        validar_senha(s)
        print(f"'{s}' → Válida!")
    except SenhaFracaError as e:
        print(f"'{s}' → {e}")
````

### Exercício 3: Lendo arquivo com finally

````python
def ler_arquivo_seguro(nome_arquivo):
    arquivo = None
    try:
        arquivo = open(nome_arquivo, "r", encoding="utf-8")
        return arquivo.read()
    except FileNotFoundError:
        return "Arquivo não encontrado."
    except PermissionError:
        return "Sem permissão para ler o arquivo."
    finally:
        if arquivo:
            arquivo.close()
            print("Arquivo fechado com finally.")

print(ler_arquivo_seguro("arquivo_inexistente.txt"))
````

## 📌 Resumo

- `try/except` captura e trata exceções evitando que o programa quebre
- Use `else` para código que executa apenas se não houver erro
- Use `finally` para código que executa sempre (ex: fechar arquivos)
- `raise` permite lançar exceções manualmente (inclusive customizadas)
- Múltiplos `except` tratam diferentes tipos de erro
- `assert` verifica condições em tempo de desenvolvimento (desabilitado com `-O`)

## 💡 Curiosidades

- Exceções são objetos! Você pode criar suas próprias classes de exceção herdando de `Exception`
- O bloco `finally` é executado mesmo se houver um `return` dentro do `try` ou `except`
- `assert` não deve ser usado para validação de entrada do usuário — apenas para testes internos
- Em Python, é considerado mais fácil pedir perdão do que permissão (EAFP — Easier to Ask for Forgiveness than Permission)

## ✅ Boas Práticas

- Trate exceções específicas em vez de usar `except:` genérico (captura até `KeyboardInterrupt`)
- Use `as` para capturar a mensagem do erro: `except ValueError as e`
- Não use `assert` para validação de dados do usuário — use `if` com `raise`
- Feche recursos (arquivos, conexões) no `finally` ou use `with` (context manager)
- Crie exceções customizadas apenas quando fizer sentido semântico no seu domínio

## ⚠️ Erros Comuns

| Erro | Causa | Solução |
|------|-------|---------|
| `except:` captura tudo (inclusive `Ctrl+C`) | Usar except genérico sem especificar | Use `except Exception as e:` para capturar exceções comuns |
| `AssertionError` inesperado | `assert` usado para validar entrada do usuário | Use `if` com `raise ValueError()` em vez de `assert` |
| Esquecer de tratar `ZeroDivisionError` | Não considerar divisão por zero | Sempre trate ou verifique antes de dividir |
| Variável não definida no `finally` | Tentar acessar variável criada dentro do `try` | Inicialize a variável antes do `try` |
| Exceção não tratada no `else` | Código no `else` também pode gerar erro | Coloque outro try/except dentro do else se necessário |
