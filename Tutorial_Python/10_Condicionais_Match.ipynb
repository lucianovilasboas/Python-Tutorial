## 📝 Introdução

Condicionais permitem que o programa tome decisões baseadas em condições. O comando `if` é a estrutura fundamental de controle de fluxo. A partir do Python 3.10, o `match` statement oferece uma forma poderosa e elegante de correspondência de padrões,similar ao switch/case de outras linguagens.

## 📚 Explicação Detalhada

### if / elif / else
```python
if condicao:
    # bloco executado se condicao for True
elif outra_condicao:
    # bloco executado se a primeira for False e esta for True
else:
    # bloco executado se todas forem False
```

### Operador Ternário
```python
valor if condicao else outro_valor
```

### Match Statement (Python 3.10+)
```python
match valor:
    case padrão1:
        # ação
    case padrão2:
        # ação
    case _:
        # default (wildcard)
```

O `match` suporta:
- Padrões literais
- Padrões com variáveis
- Padrões OR (`|`)
- Padrões de sequência (listas, tuplas)
- Padrões de mapeamento (dicionários)
- Guards (`case x if x > 0:`)

## 💻 Exemplos

````python
# Exemplo 1: if básico
idade = int(input("Digite sua idade: "))
if idade >= 18:
    print("Você é maior de idade.")
else:
    print("Você é menor de idade.")
````

````python
# Exemplo 2: elif (múltiplas condições)
nota = float(input("Digite a nota (0 a 10): "))
if nota >= 9:
    conceito = "A"
elif nota >= 7:
    conceito = "B"
elif nota >= 5:
    conceito = "C"
else:
    conceito = "D"
print(f"Conceito: {conceito}")
````

````python
# Exemplo 3: Operadores lógicos em condicionais
ano = int(input("Digite o ano: "))
if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"{ano} é bissexto.")
else:
    print(f"{ano} não é bissexto.")
````

````python
# Exemplo 4: Operador ternário
numero = int(input("Digite um número: "))
resultado = "par" if numero % 2 == 0 else "ímpar"
print(f"O número {numero} é {resultado}.")

# Ternário aninhado (use com moderação)
x = 15
classif = "alto" if x > 10 else "médio" if x > 5 else "baixo"
print(f"Classificação: {classif}")
````

````python
# Exemplo 5: Condicionais aninhadas
idade = 25
tem_carteira = True
if idade >= 18:
    if tem_carteira:
        print("Pode dirigir.")
    else:
        print("Precisa tirar carteira.")
else:
    print("Menor de idade não pode dirigir.")
````

````python
# Exemplo 6: pass em condicionais
x = 10
if x > 5:
    pass  # TODO: implementar lógica depois
print("O pass apenas ignora o bloco.")
````

````python
# Exemplo 7: match - padrões literais
codigo = int(input("Digite o código (1-3): "))
match codigo:
    case 1:
        print("Código 1: Produto A")
    case 2:
        print("Código 2: Produto B")
    case 3:
        print("Código 3: Produto C")
    case _:
        print("Código inválido!")
````

````python
# Exemplo 8: match - OR patterns e guards
comando = input("Digite um comando: ").strip().lower()
match comando:
    case "sair" | "exit" | "quit":
        print("Encerrando o programa...")
    case "ajuda" | "help":
        print("Comandos disponíveis: sair, ajuda, versao")
    case _ as outro:
        print(f"Comando desconhecido: {outro}")
````

````python
# Exemplo 9: match com sequências
ponto = (5, 0)
match ponto:
    case (0, 0):
        print("Origem")
    case (x, 0):
        print(f"Sobre o eixo X em x={x}")
    case (0, y):
        print(f"Sobre o eixo Y em y={y}")
    case (x, y):
        print(f"Ponto em ({x}, {y})")
````

````python
# Exemplo 10: match com guard
numero = int(input("Digite um número: "))
match numero:
    case n if n > 0:
        print(f"{n} é positivo")
    case n if n < 0:
        print(f"{n} é negativo")
    case _:
        print("É zero")
````

## ✅ Exercícios Resolvidos

````python
# Exercício Resolvido 1: Calculadora simples com match
a = float(input("Digite o primeiro número: "))
b = float(input("Digite o segundo número: "))
op = input("Digite a operação (+, -, *, /): ").strip()

match op:
    case "+":
        print(f"Resultado: {a + b}")
    case "-":
        print(f"Resultado: {a - b}")
    case "*":
        print(f"Resultado: {a * b}")
    case "/":
        if b != 0:
            print(f"Resultado: {a / b}")
        else:
            print("Erro: divisão por zero!")
    case _:
        print("Operação inválida!")
````

````python
# Exercício Resolvido 2: IMC com if/elif/else
peso = float(input("Digite seu peso (kg): "))
altura = float(input("Digite sua altura (m): "))
imc = peso / (altura ** 2)

if imc < 18.5:
    classificacao = "Abaixo do peso"
elif imc < 25:
    classificacao = "Peso normal"
elif imc < 30:
    classificacao = "Sobrepeso"
else:
    classificacao = "Obesidade"
print(f"IMC: {imc:.2f} — {classificacao}")
````

````python
# Exercício Resolvido 3: Classificador de triângulos
a = float(input("Lado 1: "))
b = float(input("Lado 2: "))
c = float(input("Lado 3: "))

# Primeiro verifica se é triângulo
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        tipo = "Equilátero"
    elif a == b or a == c or b == c:
        tipo = "Isósceles"
    else:
        tipo = "Escaleno"
    print(f"É um triângulo {tipo}.")
else:
    print("Não forma um triângulo.")
````

## 📌 Resumo

- `if` / `elif` / `else` para decisões baseadas em condições booleanas
- Operador ternário: `valor if cond else outro`
- `match` / `case` para correspondência de padrões (Python 3.10+)
- Use `_` como wildcard (default) no match
- Guards (`if`) dentro de `case` para condições adicionais
- `pass` como placeholder para blocos vazios

## 💡 Curiosidades

- O `match` statement foi inspirado em linguagens funcionais como Haskell e Rust
- Diferente do `switch` de C, o `match` não precisa de `break` — apenas um caso é executado
- O `match` pode decompor estruturas complexas como listas, tuplas e dicionários

## ✅ Boas Práticas

- Evite aninhamento profundo de if/else; prefira elif ou early return
- Use `match` quando houver muitos casos (mais legível que vários elif)
- Sempre trate o caso padrão (`else` ou `case _:`)
- Mantenha condições simples e nomeie variáveis booleanas com clareza

## ⚠️ Erros Comuns

| Erro | Exemplo | Solução |
|------|---------|--------|
| Dois pontos esquecidos | `if x > 5` | Adicione `:` ao final |
| Indentação incorreta | Bloco não alinhado | Use 4 espaços consistentemente |
| Usar `=` em vez de `==` | `if x = 5:` | Use `==` para comparação |
| Esquecer o wildcard `_` | `match x: case 1: ...` | Adicione `case _:` para default |
| Match sem cobertura total | Apenas alguns casos | Sempre inclua `case _:` |
