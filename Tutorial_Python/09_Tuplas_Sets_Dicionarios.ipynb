## 📝 Introdução

Além das listas, Python oferece outras estruturas de dados poderosas. **Tuplas** são sequências imutáveis; **sets** são coleções não ordenadas sem elementos repetidos; **dicionários** mapeiam chaves únicas a valores. Cada uma tem seu caso de uso ideal.

## 📚 Explicação Detalhada

### Tuplas
- Ordenadas e **imutáveis** (não podem ser alteradas após criação)
- Criadas com `()` ou `tuple()`
- Úteis para dados que não devem ser modificados (coordenadas, configurações)
- Suportam desempacotamento

### Sets
- **Não ordenados**, sem índices
- **Sem duplicatas** (elementos únicos)
- Criados com `{}` ou `set()`
- Suportam operações matemáticas de conjuntos (união, interseção, diferença)
- `frozenset` é a versão imutável

### Dicionários
- Mapeiam **chaves** únicas a **valores**
- Criados com `{}` (pares `chave: valor`) ou `dict()`
- Mutáveis e ordenados (a partir do Python 3.7)
- Chaves devem ser imutáveis (strings, números, tuplas)

## 💻 Exemplos

```python
# Exemplo 1: Criando tuplas
t1 = (1, 2, 3)
t2 = "a", "b", "c"  # sem parênteses também funciona
t3 = tuple([4, 5, 6])
print(f"t1: {t1}, tipo: {type(t1)}")
print(f"t2: {t2}")
print(f"t3: {t3}")
print(f"t1[1]: {t1[1]}")
```

```python
# Exemplo 2: Imutabilidade e desempacotamento
coordenadas = (10, 20)
# coordenadas[0] = 5  # Isso geraria TypeError!

x, y = coordenadas
print(f"x = {x}, y = {y}")

# Desempacotamento com *
primeiro, *resto = (1, 2, 3, 4, 5)
print(f"primeiro: {primeiro}, resto: {resto}")
```

```python
# Exemplo 3: Métodos de tuplas
t = (1, 3, 2, 3, 4, 3)
print(f"count(3): {t.count(3)}")
print(f"index(2): {t.index(2)}")
print(f"len: {len(t)}")

# Concatenar tuplas
t4 = t + (5, 6)
print(f"Concatenada: {t4}")
```

```python
# Exemplo 4: Sets - criação e adição
s1 = {1, 2, 3}
s2 = set([3, 4, 5])
print(f"s1: {s1}")
print(f"s2: {s2}")

s1.add(4)
print(f"s1 após add(4): {s1}")
s1.add(1)  # não adiciona duplicata
print(f"s1 após add(1) duplicado: {s1}")

s1.remove(2)
print(f"s1 após remove(2): {s1}")
```

```python
# Exemplo 5: Operações de conjuntos (sets)
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print(f"União: {a | b}")
print(f"Interseção: {a & b}")
print(f"Diferença (a - b): {a - b}")
print(f"Diferença (b - a): {b - a}")
print(f"Diferença simétrica: {a ^ b}")
```

```python
# Exemplo 6: frozenset
fs = frozenset([1, 2, 3])  # imutável
print(f"frozenset: {fs}")
print(f"Tipo: {type(fs)}")
# fs.add(4)  # Isso geraria AttributeError!
```

```python
# Exemplo 7: Dicionários - criação e acesso
aluno = {
    "nome": "João",
    "idade": 20,
    "curso": "Ciência da Computação"
}
print(f"Nome: {aluno['nome']}")
print(f"Idade: {aluno.get('idade')}")
print(f"Nota: {aluno.get('nota', 'Não informada')}")  # valor padrão

aluno["idade"] = 21  # alterando
aluno["nota"] = 9.5  # adicionando
print(f"Aluno atualizado: {aluno}")
```

```python
# Exemplo 8: Métodos de dicionários
dic = {"a": 1, "b": 2, "c": 3}
print(f"Chaves: {list(dic.keys())}")
print(f"Valores: {list(dic.values())}")
print(f"Itens: {list(dic.items())}")

for chave, valor in dic.items():
    print(f"{chave} → {valor}")

# Remover
valor = dic.pop("b")
print(f"Valor removido 'b': {valor}")
print(f"Dicionário após pop: {dic}")
```

```python
# Exemplo 9: Dicionários aninhados
turma = {
    101: {"nome": "Ana", "notas": [8.0, 9.0, 7.5]},
    102: {"nome": "Bruno", "notas": [6.0, 7.0, 5.5]},
    103: {"nome": "Carla", "notas": [10.0, 9.5, 9.8]},
}
for matricula, dados in turma.items():
    media = sum(dados["notas"]) / len(dados["notas"])
    print(f"{matricula} - {dados['nome']}: média {media:.2f}")
```

## ✅ Exercícios Resolvidos

```python
# Exercício Resolvido 1: Contagem de caracteres com dicionário
frase = input("Digite uma frase: ").lower()
contagem = {}
for char in frase:
    if char != " ":
        contagem[char] = contagem.get(char, 0) + 1
print("Frequência de caracteres:")
for char, freq in sorted(contagem.items()):
    print(f"'{char}': {freq}")
```

```python
# Exercício Resolvido 2: União de sets
conj1 = set(map(int, input("Digite números para o conjunto 1: ").split()))
conj2 = set(map(int, input("Digite números para o conjunto 2: ").split()))
uniao = conj1 | conj2
intersec = conj1 & conj2
print(f"União: {sorted(uniao)}")
print(f"Interseção: {sorted(intersec)}")
```

```python
# Exercício Resolvido 3: Agenda telefônica
agenda = {}
agenda["João"] = "(11) 99999-0001"
agenda["Maria"] = "(21) 98888-0002"
agenda["Pedro"] = "(31) 97777-0003"

nome = input("Digite o nome para consultar: ")
telefone = agenda.get(nome, "Contato não encontrado")
print(f"Telefone de {nome}: {telefone}")
```

## 📌 Resumo

- **Tuplas**: imutáveis, ordenadas, permitem duplicatas — use para dados fixos
- **Sets**: não ordenados, sem duplicatas, suportam operações de conjunto
- **Dicionários**: chave → valor, mutáveis, chaves únicas e imutáveis
- `frozenset` é a versão imutável do set
- Use `.get()` para acessar dicionários com valor padrão seguro

## 💡 Curiosidades

- Dicionários são implementados como tabelas hash — acesso extremamente rápido (O(1))
- Sets também usam hash, por isso exigem elementos imutáveis
- A partir do Python 3.7, dicionários mantêm a ordem de inserção (recurso de linguagem, não só de implementação)

## ✅ Boas Práticas

- Use tuplas para dados que não devem mudar (constantes, coordenadas)
- Use sets para remover duplicatas ou testar pertinência
- Prefira `.get()` em vez de acesso direto `dic["chave"]` quando a chave pode não existir
- Use `defaultdict` do módulo `collections` para contagens e agrupamentos

## ⚠️ Erros Comuns

| Erro | Exemplo | Solução |
|------|---------|--------|
| Tentar modificar tupla | `tupla[0] = 99` | Use lista se precisar modificar |
| Chave não encontrada | `dic["chave_inexistente"]` | Use `dic.get("chave", default)` |
| Adicionar lista em set | `set([1, [2,3]])` | Elementos de set devem ser imutáveis |
| Confundir `set` com `{}` vazio | `x = {}` cria dicionário, não set | Use `x = set()` para set vazio |
