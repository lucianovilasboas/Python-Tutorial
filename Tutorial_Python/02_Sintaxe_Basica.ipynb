## 📝 Introdução

A sintaxe de uma linguagem de programação é o conjunto de regras que define como os programas são escritos e interpretados. Assim como na língua portuguesa temos regras gramaticais, no Python existem regras específicas que devemos seguir para que o computador entenda nossas instruções.

Python foi projetado para ser uma linguagem de fácil leitura e escrita. Sua sintaxe limpa e intuitiva é uma das razões pelas quais é tão recomendada para iniciantes. Neste notebook, vamos explorar os elementos básicos da sintaxe Python.

## 📚 Explicação Detalhada

### 1. Indentação

Diferente de outras linguagens que usam chaves `{}` ou palavras-chave para definir blocos de código, Python usa **indentação** (espaços ou tabs). A indentação não é opcional — ela é parte obrigatória da sintaxe!

**Regras de indentação:**
- Use 4 espaços por nível de indentação (recomendado pelo PEP 8)
- Nunca misture tabs e espaços
- A indentação define blocos como funções, loops, condicionais

```python
# Exemplo de indentação correta
idade = 18
if idade >= 18:
    print("Você é maior de idade")
    print("Já pode votar!")
print("Este print está fora do if")
```

```python
# ATENÇÃO: indentação incorreta causa erro!
# Descomente o código abaixo para ver o erro:
# if True:
# print("Erro de indentação!")
```

### 2. Statements (Instruções)

Um **statement** é uma unidade de código que o interpretador Python pode executar. Em Python, normalmente cada linha é um statement. 

- Statements simples ocupam uma única linha
- Podemos usar `;` para múltiplos statements na mesma linha (não recomendado)
- Podemos quebrar linhas longas com `\` ou com parênteses

```python
# Statements em linhas separadas (mais legível)
nome = "Ana"
idade = 25
print(f"{nome} tem {idade} anos")
```

```python
# Quebra de linha com parênteses (recomendado)
total = (10 + 20 + 30 +
         40 + 50 + 60)
print(f"Total: {total}")
```

### 3. Variáveis em Python

Variáveis são como "caixinhas" que guardam dados na memória. Python é uma linguagem de **tipagem dinâmica** — você não precisa declarar o tipo da variável; ele é inferido automaticamente.

```python
# Criando variáveis - atribuição simples
nome = "Carlos"
idade = 30
altura = 1.75
estudante = True

print(f"Nome: {nome}")
print(f"Idade: {idade}")
print(f"Altura: {altura}")
print(f"Estudante: {estudante}")
```

### 4. Regras de Nomenclatura (Naming)

| Regra | Exemplo Correto | Exemplo Incorreto |
|-------|-----------------|-------------------|
| Deve começar com letra ou `_` | `nome`, `_valor` | `1nome`, `@nome` |
| Pode conter letras, números e `_` | `nome2`, `minha_var` | `nome-com`, `nome!` |
| Case-sensitive | `Nome` ≠ `nome` | — |
| Não pode ser palavra reservada | `nome` | `if`, `for`, `while` |
| Use snake_case (PEP 8) | `minha_variavel` | `minhaVariavel` |

```python
# Exemplos de nomes válidos
nome_completo = "Maria Souza"
_contador = 0
valor_total = 150.50
data_nascimento = "15/03/1990"

# Python é case-sensitive
Nome = "João"
nome = "Maria"
print(f"Nome (maiúsculo): {Nome}")
print(f"nome (minúsculo): {nome}")  # São variáveis diferentes!
```

### 5. Atribuição Múltipla

Python permite atribuir valores a várias variáveis em uma única linha, o que torna o código mais conciso.

```python
# Atribuição múltipla
x, y, z = 10, 20, 30
print(f"x = {x}, y = {y}, z = {z}")

# Mesmo valor para várias variáveis
a = b = c = 0
print(f"a = {a}, b = {b}, c = {c}")

# Troca de valores (swap)
x, y = y, x
print(f"Após swap: x = {x}, y = {y}")
```

### 6. A Função `print()` e Saída de Dados

A função `print()` é usada para exibir informações na tela. Ela pode receber múltiplos argumentos.

```python
# Diferentes formas de usar print
print("Olá, mundo!")                              # String simples
print("Python", "é", "incrível", sep=" - ")   # Separador personalizado
print("Linha 1", end=" -> ")                     # Sem quebra de linha
print("Linha 2")                                  # Continua na mesma linha
```

## 💻 Exemplos Práticos

```python
# Exemplo 1: Calculadora de média
nota1 = 8.5
nota2 = 7.0
nota3 = 9.2
media = (nota1 + nota2 + nota3) / 3
print(f"Média: {media:.2f}")

if media >= 7.0:
    print("Aluno aprovado! 🎉")
else:
    print("Aluno em recuperação.")
```

```python
# Exemplo 2: Conversor de temperatura
celsius = 25
fahrenheit = (celsius * 9/5) + 32
print(f"{celsius}°C equivale a {fahrenheit}°F")
```

```python
# Exemplo 3: Área de um círculo
pi = 3.14159
raio = 5
area = pi * raio ** 2
print(f"Área do círculo com raio {raio}: {area:.2f}")
```

## 📌 Resumo

- **Indentação**: Essencial em Python. Use 4 espaços para definir blocos de código.
- **Statements**: São instruções executáveis. Cada linha geralmente contém um statement.
- **Variáveis**: São criadas no momento da atribuição; Python infere o tipo automaticamente.
- **Naming**: Use `snake_case`, nomes descritivos, evite palavras reservadas.
- **Atribuição múltipla**: Permite atribuir valores a várias variáveis em uma linha.
- **case-sensitive**: `Nome` e `nome` são variáveis diferentes.
- **PEP 8**: Guia de estilo oficial do Python. Siga suas recomendações!

## 💡 Curiosidades

- **Por que Python usa indentação?** Guido van Rossum, criador do Python, acreditava que a legibilidade é fundamental. A indentação obrigatória força os programadores a escrever código mais limpo e organizado.

- O nome **snake_case** vem da aparência de cobra (snake) que os underscores (`_`) criam: `minha_variavel_exemplo`.

- Python foi inspirado pelo ABC, uma linguagem que também usava indentação. Na década de 1970, já existia essa ideia de que código bem indentado é mais fácil de entender.

- Ao contrário de linguagens como Java ou C, Python não precisa de `;` no final das linhas. Isso torna o código mais limpo!

## ✅ Boas Práticas

1. **Use 4 espaços para indentação** — não use tabs (configure seu editor para converter tab em espaços).
2. **Nomes significativos** — `salario_mensal` é melhor que `sm` ou `x`.
3. **Mantenha linhas com até 79 caracteres** (recomendação PEP 8).
4. **Use espaços ao redor de operadores** — `x = 5 + 3` e não `x=5+3`.
5. **Nunca use `;`** para colocar múltiplos statements na mesma linha.
6. **Siga o snake_case** para nomes de variáveis e funções.
7. **Evite nomes muito genéricos** como `dado`, `temp`, `var`.

## ⚠️ Erros Comuns

| Erro | Exemplo | Como Evitar |
|------|---------|-------------|
| Indentação inconsistente | Misturar tabs e espaços | Configure o editor para usar apenas espaços |
| Esquecer os dois pontos `:` | `if x > 5` sem `:` | Sempre adicione `:` após `if`, `for`, `while`, `def` |
| Nome de variável inválido | `2nome`, `meu-nome` | Use apenas letras, números e `_`; comece com letra ou `_` |
| Usar palavra reservada | `if = 10` | Consulte a lista de palavras reservadas do Python |
| Esquecer que é case-sensitive | Definir `Nome` e chamar `nome` | Seja consistente com a capitalização |
| Linhas muito longas | Código > 79 caracteres sem quebra | Use `()` ou `\` para quebrar linhas |
| Atribuição em vez de comparação | `if x = 5:` em vez de `if x == 5:` | Use `==` para comparação, `=` para atribuição |
