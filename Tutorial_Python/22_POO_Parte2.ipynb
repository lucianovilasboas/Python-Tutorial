## 📝 Introdução

Na primeira parte vimos os fundamentos da POO. Agora vamos explorar conceitos mais avançados que tornam a POO tão poderosa: **herança** (reutilizar e estender classes), **polimorfismo** (um mesmo método com comportamentos diferentes), **encapsulamento** (proteger dados internos) e **classes internas** (organização lógica).

Esses pilares permitem criar sistemas mais flexíveis, reutilizáveis e fáceis de manter.

## 📚 Explicação Detalhada

### Herança Simples

Uma classe **(filha)** herda atributos e métodos de outra classe **(pai)**. A classe filha pode:
- Usar tudo que a classe pai tem
- Adicionar novos atributos e métodos
- **Sobrescrever (override)** métodos existentes

```python
class Animal:
    def emitir_som(self):
        return "..."

class Cachorro(Animal):
    def emitir_som(self):
        return "Au au!"
```

### `super()`

Permite chamar métodos da classe pai dentro da classe filha. Essencial para não duplicar código.

### Herança Múltipla

Uma classe pode herdar de **várias classes** ao mesmo tempo. Python usa **MRO (Method Resolution Order)** para determinar qual método será chamado.

### Polimorfismo

Capacidade de um objeto poder ser tratado como se fosse de outro tipo. Em Python, o **duck typing** é a forma mais comum: "se anda como um pato e nada como um pato, então é um pato".

### Encapsulamento

Em Python, usa-se `__` (duplo underscore) para tornar um atributo "privado" (name mangling). O decorador `@property` permite criar getters/setters de forma elegante.

### Classes Internas (Nested Classes)

Classes definidas dentro de outras classes. Úteis para agrupar logicamente classes que só fazem sentido juntas.

## 💻 Exemplos

````python
# Exemplo 1: Herança simples
class Animal:
    def __init__(self, nome):
        self.nome = nome

    def mover(self):
        return f"{self.nome} está se movendo"

class Cachorro(Animal):
    def latir(self):
        return f"{self.nome} diz: Au au!"

rex = Cachorro("Rex")
print(rex.mover())  # herdado
print(rex.latir())   # próprio
````

````python
# Exemplo 2: super() para estender o __init__ do pai
class Animal:
    def __init__(self, nome):
        self.nome = nome

    def __str__(self):
        return f"Animal: {self.nome}"

class Cachorro(Animal):
    def __init__(self, nome, raca):
        super().__init__(nome)  # chama o __init__ de Animal
        self.raca = raca

    def __str__(self):
        return f"Cachorro: {self.nome} ({self.raca})"

rex = Cachorro("Rex", "Pastor Alemão")
print(rex)
````

````python
# Exemplo 3: Herança múltipla
class Trabalhador:
    def trabalhar(self):
        return "Trabalhando..."

class Estudante:
    def estudar(self):
        return "Estudando..."

class Estagiario(Trabalhador, Estudante):
    def __init__(self, nome):
        self.nome = nome

    def atividades(self):
        return f"{self.nome}: {self.trabalhar()} e {self.estudar()}"

joao = Estagiario("João")
print(joao.atividades())
print(Estagiario.__mro__)  # Method Resolution Order
````

````python
# Exemplo 4: Polimorfismo com sobrescrita
class Animal:
    def emitir_som(self):
        return "..."

class Cachorro(Animal):
    def emitir_som(self):
        return "Au au!"

class Gato(Animal):
    def emitir_som(self):
        return "Miau!"

class Vaca(Animal):
    def emitir_som(self):
        return "Muu!"

animais = [Cachorro(), Gato(), Vaca()]
for animal in animais:
    print(animal.emitir_som())
````

````python
# Exemplo 5: Duck typing (polimorfismo implícito)
class Pato:
    def fazer_som(self):
        return "Quack!"

class Sapo:
    def fazer_som(self):
        return "Croac!"

def imprimir_som(objeto):
    # Não importa a classe, importa se tem o método
    print(objeto.fazer_som())

imprimir_som(Pato())
imprimir_som(Sapo())
````

````python
# Exemplo 6: Encapsulamento com __ (name mangling)
class Conta:
    def __init__(self, titular, saldo):
        self.titular = titular
        self.__saldo = saldo  # atributo "privado"

    def get_saldo(self):
        return self.__saldo

    def depositar(self, valor):
        if valor > 0:
            self.__saldo += valor

c = Conta("Maria", 1000)
print(c.get_saldo())
try:
    print(c.__saldo)  # AttributeError!
except AttributeError as e:
    print(f"Erro: {e}")
print(c._Conta__saldo)  # name mangling (acesso possível, mas não recomendado)
````

````python
# Exemplo 7: @property (getter e setter)
class Pessoa:
    def __init__(self, nome, idade):
        self.nome = nome
        self._idade = idade

    @property
    def idade(self):
        return self._idade

    @idade.setter
    def idade(self, valor):
        if valor < 0 or valor > 150:
            raise ValueError("Idade inválida")
        self._idade = valor

    @property
    def maioridade(self):
        return self._idade >= 18

p = Pessoa("Ana", 25)
print(p.idade)  # getter
p.idade = 30    # setter
print(p.maioridade)  # propriedade computada
try:
    p.idade = 200
except ValueError as e:
    print(f"Erro: {e}")
````

````python
# Exemplo 8: Classes internas (nested classes)
class Universidade:
    def __init__(self, nome):
        self.nome = nome
        self.alunos = []

    class Aluno:
        def __init__(self, nome, matricula):
            self.nome = nome
            self.matricula = matricula

        def __str__(self):
            return f"Aluno({self.nome}, {self.matricula})"

    def matricular(self, nome, matricula):
        aluno = self.Aluno(nome, matricula)
        self.alunos.append(aluno)
        return aluno

    def listar_alunos(self):
        return [str(a) for a in self.alunos]

ufabc = Universidade("UFABC")
a1 = ufabc.matricular("Carlos", "2024001")
a2 = ufabc.matricular("Marina", "2024002")
print(ufabc.listar_alunos())
print(type(a1))  # <class '__main__.Universidade.Aluno'>
````

## ✅ Exercícios Resolvidos

````python
# Exercício 1: Hierarquia de formas geométricas
import math

class Forma:
    def area(self):
        return 0

    def __str__(self):
        return "Forma genérica"

class Circulo(Forma):
    def __init__(self, raio):
        self.raio = raio

    def area(self):
        return math.pi * self.raio ** 2

    def __str__(self):
        return f"Círculo (raio={self.raio})"

class Quadrado(Forma):
    def __init__(self, lado):
        self.lado = lado

    def area(self):
        return self.lado ** 2

    def __str__(self):
        return f"Quadrado (lado={self.lado})"

formas = [Circulo(5), Quadrado(4)]
for f in formas:
    print(f"{f} -> Área: {f.area():.2f}")
````

````python
# Exercício 2: Sistema de funcionários com herança
class Funcionario:
    def __init__(self, nome, salario):
        self.nome = nome
        self.salario = salario

    def calcular_bonus(self):
        return self.salario * 0.1

    def __str__(self):
        return f"{self.nome} - Salário: R$ {self.salario:.2f}"

class Gerente(Funcionario):
    def __init__(self, nome, salario, setor):
        super().__init__(nome, salario)
        self.setor = setor

    def calcular_bonus(self):
        return super().calcular_bonus() + self.salario * 0.15

    def __str__(self):
        return f"Gerente {self.nome} ({self.setor}) - Salário: R$ {self.salario:.2f}"

f = Funcionario("João", 3000)
g = Gerente("Maria", 5000, "TI")
print(f)
print(f"Bônus: R$ {f.calcular_bonus():.2f}")
print(g)
print(f"Bônus: R$ {g.calcular_bonus():.2f}")
````

````python
# Exercício 3: Classe com validação via @property
class Produto:
    def __init__(self, nome, preco):
        self.nome = nome
        self._preco = preco

    @property
    def preco(self):
        return self._preco

    @preco.setter
    def preco(self, valor):
        if valor <= 0:
            raise ValueError("Preço deve ser positivo")
        self._preco = valor

    def aplicar_desconto(self, percentual):
        if not 0 < percentual <= 100:
            raise ValueError("Percentual inválido")
        self._preco *= (100 - percentual) / 100

    def __str__(self):
        return f"Produto: {self.nome} - R$ {self._preco:.2f}"

p = Produto("Notebook", 3500)
print(p)
p.aplicar_desconto(10)
print(p)
p.preco = 2000
print(p)
````

## 📌 Resumo

- **Herança**: `class Filha(Pai):` — reutiliza e estende código
- **`super()`**: chama métodos da classe pai (evita duplicação)
- **Herança múltipla**: classe herda de várias ao mesmo tempo; MRO define ordem de resolução
- **Polimorfismo**: objetos de diferentes classes respondem ao mesmo método
- **Duck typing**: Python não verifica tipo, apenas se o método existe
- **Encapsulamento**: `__atributo` (name mangling) + `@property` (getter/setter)
- **Classes internas**: classes definidas dentro de outras para organização

## 💡 Curiosidades

- Python implementa herança múltipla de forma diferente de C++. O MRO (Method Resolution Order) usa o **Algoritmo C3 Linearization**.
- Em Python, **tudo é objeto** — até funções, classes e módulos são objetos.
- O name mangling (`__attr`) não é segurança real, é apenas uma convenção reforçada. O atributo ainda é acessível via `_Classe__attr`.
- Python não tem `private`/`protected`/`public` como Java. Tudo é público por padrão.
- O conceito de duck typing ficou famoso com o poema: "If it walks like a duck and quacks like a duck, it must be a duck".

## ✅ Boas Práticas

- Prefira **composição** sobre herança quando possível (menos acoplamento)
- Use `super().__init__()` sempre que estender uma classe
- Evite herança múltipla excessiva — pode gerar código confuso
- Documente a hierarquia de classes para facilitar manutenção
- Use `@property` em vez de `get_`/`set_` métodos (estilo Python)
- Não confie em atributos `__` para segurança — servem apenas para evitar conflitos
- Mantenha a hierarquia de herança rasa (poucos níveis)

## ⚠️ Erros Comuns

| Erro | Exemplo | Correção |
|------|---------|----------|
| Esquecer `()` no `super()` | `super().__init__` | `super().__init__()` |
| Herança circular | `A(B)` e `B(A)` | Redesenhar a hierarquia |
| Confundir `@property` com atributo | Chamar `obj.property()` | Usar `obj.property` (sem parênteses) |
| Não chamar `super()` no `__init__` | Variáveis do pai não inicializadas | Chamar `super().__init__()` |
| Acessar `__attr` diretamente | `obj.__attr` (AttributeError) | Usar getter ou `obj._Classe__attr` |
| Modificar `__dict__` diretamente | Quebrar encapsulamento manualmente | Usar métodos e propriedades |
