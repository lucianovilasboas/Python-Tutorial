## 📝 Introdução

Range, arrays, iteradores e geradores são ferramentas fundamentais para trabalhar com sequências em Python. Enquanto `range` gera sequências numéricas eficientes, iteradores e geradores permitem processar dados sob demanda sem carregar tudo na memória. Este notebook explora cada um desses conceitos com exemplos práticos.

## 📚 Explicação Detalhada

### Range
`range(start, stop, step)` gera uma sequência imutável de números. É lazy (não armazena todos os valores).

### Array
O módulo `array` fornece arrays tipados (mesmo tipo) mais eficientes que listas para grandes volumes de dados numéricos.

### Iteradores
Um iterador é um objeto que implementa `__iter__()` (retorna self) e `__next__()` (próximo item ou `StopIteration`).

### Geradores
Funções com `yield` que produzem uma sequência sob demanda. Generator expressions são como list comprehensions, mas lazy.

## 💻 Exemplos

```python
# Exemplo 1: Range básico
for i in range(5):
    print(i, end=" ")
```

```python
# Exemplo 2: Range com start, stop, step
print("range(2, 10, 2):", list(range(2, 10, 2)))
print("range(10, 0, -1):", list(range(10, 0, -1)))
```

```python
# Exemplo 3: Módulo array
from array import array

arr = array('i', [1, 2, 3, 4, 5])
print(f"Array: {arr}")
print(f"Tipo: {arr.typecode}")
arr.append(6)
print(f"Append: {arr}")
```

```python
# Exemplo 4: Array vs Lista — diferença de tipo
from array import array

arr = array('d', [1.5, 2.7, 3.14])
lista = [1.5, 2.7, 3.14]

print(f"Array: {arr} (tipo: {arr.typecode})")
print(f"Lista: {lista} (tipos misturados são permitidos)")
```

```python
# Exemplo 5: Criando um iterador personalizado
class Contador:
    def __init__(self, limite):
        self.limite = limite
        self.atual = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.atual >= self.limite:
            raise StopIteration
        self.atual += 1
        return self.atual

for num in Contador(5):
    print(num, end=" ")
```

```python
# Exemplo 6: Gerador com yield
def fibonacci_gen(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

for val in fibonacci_gen(10):
    print(val, end=" ")
```

```python
# Exemplo 7: Generator expression
quadrados = (x**2 for x in range(10))
print(f"Tipo: {type(quadrados)}")
print(f"Primeiros 5: {list(quadrados)[:5]}")
```

```python
# Exemplo 8: Iterando manualmente com iter e next
lista = ["a", "b", "c"]
it = iter(lista)

print(next(it))
print(next(it))
print(next(it))
# print(next(it))  # StopIteration
```

## ✅ Exercícios Resolvidos

```python
# Exercício Resolvido 1: Range decrescente
for i in range(20, 0, -2):
    print(i, end=" ")
```

```python
# Exercício Resolvido 2: Iterador personalizado — pares
class Pares:
    def __init__(self, limite):
        self.limite = limite
        self.atual = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.atual > self.limite:
            raise StopIteration
        valor = self.atual
        self.atual += 2
        return valor

for p in Pares(10):
    print(p, end=" ")
```

```python
# Exercício Resolvido 3: Gerador infinito com limite
def numeros_naturais():
    n = 0
    while True:
        yield n
        n += 1

gen = numeros_naturais()
for _ in range(10):
    print(next(gen), end=" ")
```

## 📌 Resumo

- `range(start, stop, step)` é lazy e eficiente para sequências numéricas
- `array('tipo', seq)` armazena dados homogêneos de forma compacta
- Iteradores implementam `__iter__` e `__next__`, lançando `StopIteration`
- Geradores com `yield` são iteradores lazy e elegantes
- Generator expressions são como comprehensions, mas com `()`
- `iter()` e `next()` permitem iteração manual
- Iteradores e geradores economizam memória processando sob demanda

## 💡 Curiosidades

- A função `range()` em Python 3 substituiu `xrange()` do Python 2
- O módulo `array` é mais eficiente que listas para grandes volumes, mas menos flexível
- Geradores podem representar sequências infinitas sem consumir memória infinita
- O protocolo de iteração é a base do funcionamento do `for` em Python
- Generator expressions são ligeiramente mais lentas que list comprehensions, mas usam menos memória

## ✅ Boas Práticas

- Prefira `range()` a listas grandes de números
- Use `array` para grandes volumes de dados numéricos homogêneos
- Prefira geradores a listas quando não precisar de todos os valores simultaneamente
- Use generator expressions para transformações simples e lazy
- Sempre levante `StopIteration` no `__next__` quando não houver mais itens

## ⚠️ Erros Comuns

| Erro | Descrição | Solução |
|------|-----------|---------|
| StopIteration não tratado | Chamar `next()` além do fim | Capturar `StopIteration` ou usar `for` |
| Range não é lista | `range(5)[1.5]` gera TypeError | Converter com `list()` se precisar de índices float |
| Array com tipo errado | `array('i', [1.5])` gera TypeError | Usar typecode compatível ('d' para float, 'i' para int) |
| Gerador só pode ser iterado uma vez | Consumir gerador duas vezes | Recriar o gerador se precisar iterar novamente |
| Confundir iterável com iterador | Nem todo iterável é iterador | Verificar com `hasattr(obj, '__next__')` |
| Esquecer yield | Função com yield vira gerador | Lembrar que qualquer função com yield é um gerador |
