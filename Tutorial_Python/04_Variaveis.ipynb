## 📝 Introdução

Variáveis são um dos conceitos mais fundamentais em qualquer linguagem de programação. Pense em uma variável como uma **caixa etiquetada** onde você guarda um valor. A **etiqueta** é o nome da variável, e o **conteúdo** é o valor armazenado.

Em Python, diferente de linguagens como Java ou C, você **não precisa declarar o tipo** da variável — o Python descobre automaticamente qual tipo de dado está sendo armazenado. Isso se chama **tipagem dinâmica**.

Além disso, em Python **tudo é objeto**, e as variáveis são na verdade "rótulos" (referências) para objetos na memória.

## 📚 Explicação Detalhada

### 1. O que são Variáveis?

Uma variável é um nome que **referencia** um valor armazenado na memória. Você pode pensar nela como um apelido para um dado.

Em Python, a variável é criada no momento em que você atribui um valor a ela pela primeira vez.

```python
# Criando (declarando) variáveis
nome = "Alice"       # String
idade = 25           # Inteiro (int)
altura = 1.68        # Ponto flutuante (float)
estudante = True     # Booleano (bool)

print(f"Nome: {nome}")
print(f"Idade: {idade}")
print(f"Altura: {altura}")
print(f"Estudante: {estudante}")
```

### 2. Regras de Nomenclatura

Python tem regras rígidas para nomes de variáveis:

| Regra | Detalhe |
|-------|---------|
| ✅ Começar com letra ou `_` | `nome`, `_contador` |
| ✅ Conter letras, números e `_` | `nota_final1` |
| ✅ Case-sensitive | `Nome` ≠ `nome` |
| ❌ Palavras reservadas | `if`, `for`, `while`, `class` |
| ❌ Caracteres especiais | `@`, `!`, `#`, `$`, `%` |
| ❌ Começar com número | `1nome` (invalido) |

```python
# Nomes VÁLIDOS
minha_variavel = 10
_valor_temporario = 5
nome_completo = "João Silva"
nota1 = 9.5
contador2 = 100

# Nomes INVÁLIDOS (descomente para ver o erro)
# 1nome = "erro"         # SyntaxError
# meu-nome = "erro"      # SyntaxError
# class = "erro"         # SyntaxError (palavra reservada)
# nome@ = "erro"         # SyntaxError

# Case-sensitive: variáveis diferentes
pessoa = "João"
Pessoa = "Maria"
PESSOA = "José"
print(pessoa, Pessoa, PESSOA)
```

### 3. Atribuição de Variáveis

Python oferece várias formas de atribuir valores a variáveis.

```python
# Atribuição simples
x = 10
print(f"x = {x}")

# Atribuição múltipla (mesmo valor)
a = b = c = 0
print(f"a = {a}, b = {b}, c = {c}")

# Atribuição múltipla (valores diferentes)
nome, idade, cidade = "Carlos", 32, "Rio de Janeiro"
print(f"{nome}, {idade} anos, mora no {cidade}")

# Troca de valores (swap) - específico do Python!
x, y = 5, 10
print(f"Antes: x = {x}, y = {y}")
x, y = y, x
print(f"Depois: x = {x}, y = {y}")
```

### 4. Desempacotamento de Sequências (Unpacking)

Python permite "desempacotar" sequências (listas, tuplas) em variáveis.

```python
# Desempacotamento de uma lista
frutas = ["maçã", "banana", "laranja"]
f1, f2, f3 = frutas
print(f"f1 = {f1}, f2 = {f2}, f3 = {f3}")

# Usando * para capturar múltiplos valores
primeiro, *resto = [1, 2, 3, 4, 5]
print(f"Primeiro: {primeiro}")
print(f"Resto: {resto}")

primeiro, *meio, ultimo = [1, 2, 3, 4, 5]
print(f"Primeiro: {primeiro}, Meio: {meio}, Último: {ultimo}")
```

### 5. Verificando o Tipo com `type()`

A função `type()` retorna o tipo do objeto armazenado em uma variável. É muito útil para depuração e para entender como Python classifica os dados.

```python
# Verificando tipos de variáveis
nome = "Alice"
idade = 25
altura = 1.68
estudante = True
valores = [1, 2, 3]

print(f"'{nome}' é do tipo: {type(nome)}")
print(f"{idade} é do tipo: {type(idade)}")
print(f"{altura} é do tipo: {type(altura)}")
print(f"{estudante} é do tipo: {type(estudante)}")
print(f"{valores} é do tipo: {type(valores)}")
```

### 6. Reatribuição e Tipagem Dinâmica

Em Python, uma variável pode mudar de tipo ao longo do programa — isso é a **tipagem dinâmica**.

```python
# Tipagem dinâmica: a mesma variável pode receber tipos diferentes
variavel = "Agora sou uma string"
print(f"Valor: '{variavel}' - Tipo: {type(variavel)}")

variavel = 42
print(f"Valor: {variavel} - Tipo: {type(variavel)}")

variavel = 3.14
print(f"Valor: {variavel} - Tipo: {type(variavel)}")

variavel = True
print(f"Valor: {variavel} - Tipo: {type(variavel)}")
```

### 7. Variáveis Globais vs Locais

O **escopo** de uma variável define onde ela pode ser acessada:
- **Global**: declarada fora de funções, acessível em todo o programa
- **Local**: declarada dentro de uma função, acessível apenas ali

Use a palavra-chave `global` para modificar uma variável global dentro de uma função.

```python
# Variável global
mensagem = "Olá do escopo global!"

def minha_funcao():
    # Variável local
    mensagem_local = "Olá do escopo local!"
    print(mensagem)          # Acessando variável global
    print(mensagem_local)    # Acessando variável local

minha_funcao()
# print(mensagem_local)  # Erro! Variável local não acessível fora da função
print(mensagem)  # OK - variável global
```

```python
# Modificando variável global com 'global'
contador = 0

def incrementar():
    global contador  # Declara que vamos usar a variável global
    contador += 1
    print(f"Contador dentro da função: {contador}")

incrementar()
incrementar()
incrementar()
print(f"Contador fora da função: {contador}")
```

### 8. Nomes de Variáveis e a Função `id()`

Cada objeto em Python tem um identificador único. Podemos ver isso com `id()`, que retorna o endereço de memória do objeto.

```python
# Entendendo referências com id()
x = 10
y = 10
z = x

print(f"id(x) = {id(x)}")
print(f"id(y) = {id(y)}")  # Mesmo id porque Python reutiliza objetos para números pequenos
print(f"id(z) = {id(z)}")  # Mesmo id de x (z é uma referência para o mesmo objeto)

print(f"x é y: {x is y}")  # True - mesmo objeto
print(f"x == y: {x == y}")  # True - mesmo valor
```

## 💻 Exemplos Práticos

```python
# Exemplo 1: Calculadora de Desconto
preco_original = 199.90
desconto_percentual = 15

valor_desconto = preco_original * (desconto_percentual / 100)
preco_final = preco_original - valor_desconto

print(f"Preço original: R$ {preco_original:.2f}")
print(f"Desconto ({desconto_percentual}%): R$ {valor_desconto:.2f}")
print(f"Preço final: R$ {preco_final:.2f}")
print(f"Tipo de preco_final: {type(preco_final)}")
```

```python
# Exemplo 2: Sistema de notas com variáveis
notas = [8.5, 7.0, 9.3, 6.8]
n1, n2, n3, n4 = notas  # Desempacotamento

media = (n1 + n2 + n3 + n4) / 4

print(f"Notas: {notas}")
print(f"Média: {media:.2f}")
print(f"Tipo de media: {type(media)}")

if media >= 7:
    situacao = "Aprovado(a)"
else:
    situacao = "Recuperação"

print(f"Situação: {situacao}")
```

```python
# Exemplo 3: Conversor de moedas
cotacao_dolar = 5.25
valor_em_dolar = 100

valor_em_real = valor_em_dolar * cotacao_dolar

print(f"US$ {valor_em_dolar:.2f} equivale a R$ {valor_em_real:.2f}")
print(f"Cotação usada: R$ {cotacao_dolar:.2f}")
```

```python
# Exemplo 4: Verificando tipos com type() em uma calculadora
def diagnosticar_variavel(valor):
    """Exibe o valor, tipo e ID de uma variável."""
    print(f"Valor: {valor}")
    print(f"Tipo: {type(valor)}")
    print(f"ID: {id(valor)}")
    print("-" * 30)


diagnosticar_variavel(42)
diagnosticar_variavel(3.14)
diagnosticar_variavel("Python")
diagnosticar_variavel([1, 2, 3])
```

## 🏆 Desafios

1. **Desafio do Sistema Bancário:** Crie um sistema simples com saldo global de R$ 1000. Crie funções `depositar(valor)` e `sacar(valor)` que atualizam o saldo global. Use `global` dentro das funções. No final, exiba o saldo.

2. **Desafio do Analisador de Variáveis:** Crie uma função `analisar(var)` que recebe uma variável e exibe: valor, tipo, ID, e se é um número (use `isinstance()`). Teste com string, int, float, bool, list.

3. **Desafio do Desempacotamento Avançado:** Dada a lista `dados = ["João", 25, "São Paulo", "SP", "Brasil", 01234-000]`, use desempacotamento com `*` para separar em: `nome`, `idade`, e `endereco` (que deve conter os demais itens).

```python
# 🏆 Desafio 1: Sistema Bancário
saldo = 1000.0

def depositar(valor):
    global saldo
    saldo += valor
    print(f"Depósito de R$ {valor:.2f} realizado. Saldo: R$ {saldo:.2f}")

def sacar(valor):
    global saldo
    if valor > saldo:
        print("Saldo insuficiente!")
    else:
        saldo -= valor
        print(f"Saque de R$ {valor:.2f} realizado. Saldo: R$ {saldo:.2f}")

depositar(500)
sacar(200)
sacar(2000)
print(f"Saldo final: R$ {saldo:.2f}")
print(f"Tipo de saldo: {type(saldo)}")
```

```python
# 🏆 Desafio 2: Analisador de Variáveis
def analisar(var):
    print(f"Valor: {var}")
    print(f"Tipo: {type(var)}")
    print(f"ID: {id(var)}")
    print(f"É número? {isinstance(var, (int, float))}")
    print("-" * 25)

analisar("Python")
analisar(42)
analisar(3.14)
analisar(True)
analisar([1, 2, 3])
```

```python
# 🏆 Desafio 3: Desempacotamento Avançado
dados = ["João", 25, "São Paulo", "SP", "Brasil", "01234-000"]
nome, idade, *endereco = dados
print(f"Nome: {nome}")
print(f"Idade: {idade}")
print(f"Endereço: {endereco}")
print(f"Tipo de nome: {type(nome)}")
print(f"Tipo de idade: {type(idade)}")
print(f"Tipo de endereco: {type(endereco)}")
```

## 📌 Resumo

- **Variáveis** são nomes que referenciam objetos na memória.
- **Tipagem dinâmica**: Python infere o tipo automaticamente; variáveis podem mudar de tipo.
- **Regras de nomenclatura**: letras, números e `_`; case-sensitive; não pode ser palavra reservada.
- **Atribuição múltipla**: `a, b = 1, 2` ou `a = b = c = 0`.
- **Desempacotamento**: `*var` captura múltiplos valores em uma lista.
- **`type()`**: retorna o tipo da variável.
- **`id()`**: retorna o identificador único do objeto.
- **Escopo**: variáveis globais (fora de funções) e locais (dentro de funções).
- Use `global` para modificar variáveis globais dentro de funções.
- **`snake_case`** é o padrão para nomes de variáveis em Python (PEP 8).

## 💡 Curiosidades

- Python usa **interning** para números pequenos (geralmente -5 a 256): variáveis com o mesmo valor apontam para o mesmo objeto na memória. É por isso que `x = 10; y = 10; x is y` retorna `True`.

- Diferente de muitas linguagens, Python não tem constantes nativas. Por convenção, usamos nomes **MAIÚSCULOS** para indicar que uma variável não deve ser alterada: `PI = 3.14159`.

- O Python é implementado em C (CPython). Quando você cria uma variável, o interpretador aloca um `PyObject` na memória heap.

- A função `type()` na verdade retorna um objeto `type`, que é a metaclasse de todos os objetos em Python. Sim, `type` é uma classe!

- Em Python, tudo é objeto: números, strings, funções, classes — até os próprios tipos são objetos!

## ✅ Boas Práticas

1. **Use nomes descritivos**: `salario_mensal` em vez de `sm` ou `x`.
2. **Siga o snake_case**: `minha_variavel` e não `minhaVariavel`.
3. **Constantes em MAIÚSCULAS**: `TAXA_JUROS = 0.05`.
4. **Evite nomes de uma letra** (exceto contadores em loops: `i`, `j`, `k`).
5. **Não use palavras reservadas**: `if`, `for`, `while`, `class`, `def`.
6. **Use `_` para variáveis temporárias/descartáveis**: `for _ in range(10)`.
7. **Evite variáveis globais** quando possível — prefira passar valores como parâmetros.
8. **Mantenha o escopo limitado**: declare variáveis perto de onde são usadas.

## ⚠️ Erros Comuns

| Erro | Exemplo | Como Evitar |
|------|---------|-------------|
| Nome inválido | `2nome`, `meu-nome` | Siga as regras: letras, números, `_` |
| Palavra reservada | `class = "A"` | Use `classe` ou `turma` em vez de `class` |
| Case-sensitive | `Nome` vs `nome` | Seja consistente com a capitalização |
| Variável não definida | `print(x)` sem `x` existir | Sempre inicialize variáveis antes de usar |
| Confundir `=` com `==` | `if x = 5:` | `=` atribui, `==` compara |
| Modificar global sem `global` | Tentar alterar global dentro de função | Use `global nome_var` dentro da função |
| Achar que `type()` retorna string | `type(5) == "int"` | Use `type(5) == int` ou `isinstance(5, int)` |
| Sobrescrever função embutida | `list = [1, 2, 3]` | Não use nomes como `list`, `str`, `int` |
