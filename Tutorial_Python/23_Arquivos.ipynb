## 📝 Introdução

Trabalhar com arquivos é essencial em qualquer linguagem de programação. Em Python, a manipulação de arquivos é simples e intuitiva. Seja para ler dados de um arquivo de configuração, processar um arquivo CSV, ou salvar resultados de uma execução, você precisará dominar a leitura e escrita de arquivos.

Python oferece funções poderosas como `open()`, o gerenciador de contexto `with` e o módulo `os` para operações com arquivos e diretórios.

## 📚 Explicação Detalhada

### A função `open()`

A função `open()` retorna um objeto arquivo. Sintaxe:

```python
arquivo = open(nome_arquivo, modo)
```

### Modos de abertura

| Modo | Descrição |
|------|-----------|
| `r` | Leitura (padrão). Erro se arquivo não existir |
| `w` | Escrita. Cria arquivo se não existir, sobrescreve se existir |
| `a` | Acrescentar (append). Cria se não existir, adiciona ao final |
| `x` | Criação exclusiva. Erro se arquivo já existir |
| `t` | Modo texto (padrão, combinado com r/w/a) |
| `b` | Modo binário |
| `+` | Leitura e escrita simultâneas |

### Leitura de Arquivos

- `read(n)` — lê todo o conteúdo ou `n` caracteres
- `readline()` — lê uma linha
- `readlines()` — retorna lista com todas as linhas
- Iterar diretamente com `for linha in arquivo:`

### Escrita em Arquivos

- `write(texto)` — escreve uma string
- `writelines(lista)` — escreve uma lista de strings

### Gerenciador de Contexto `with`

O `with` fecha o arquivo automaticamente, mesmo em caso de erro.

### Módulo `os`

- `os.remove(caminho)` — exclui arquivo
- `os.path.exists(caminho)` — verifica se existe
- `os.mkdir/path.mkdir` — criar diretório
- `os.listdir` — listar diretório

## 💻 Exemplos

````python
# Exemplo 1: Escrevendo em um arquivo (modo w)
arquivo = open("exemplo.txt", "w", encoding="utf-8")
arquivo.write("Olá, mundo!\n")
arquivo.write("Esta é a segunda linha.\n")
arquivo.close()
print("Arquivo criado com sucesso!")
````

````python
# Exemplo 2: Lendo um arquivo inteiro com read()
arquivo = open("exemplo.txt", "r", encoding="utf-8")
conteudo = arquivo.read()
arquivo.close()
print(repr(conteudo))
````

````python
# Exemplo 3: readline() e readlines()
# readline()
arquivo = open("exemplo.txt", "r", encoding="utf-8")
linha1 = arquivo.readline()
linha2 = arquivo.readline()
arquivo.close()
print(f"Linha 1: {linha1!r}")
print(f"Linha 2: {linha2!r}")

# readlines()
arquivo = open("exemplo.txt", "r", encoding="utf-8")
linhas = arquivo.readlines()
arquivo.close()
print(f"Lista de linhas: {linhas}")
````

````python
# Exemplo 4: Iterando sobre um arquivo com for
arquivo = open("exemplo.txt", "r", encoding="utf-8")
for i, linha in enumerate(arquivo, 1):
    linha = linha.rstrip("\n")
    print(f"{i}: {linha}")
arquivo.close()
````

````python
# Exemplo 5: Usando with (recomendado!)
with open("exemplo.txt", "a", encoding="utf-8") as f:
    f.write("Linha adicionada com with\n")
    f.write("Mais uma linha\n")
# f já foi fechado automaticamente

with open("exemplo.txt", "r", encoding="utf-8") as f:
    print(f.read())
````

````python
# Exemplo 6: writelines()
linhas = ["Maçã\n", "Banana\n", "Laranja\n"]
with open("frutas.txt", "w", encoding="utf-8") as f:
    f.writelines(linhas)

with open("frutas.txt", "r", encoding="utf-8") as f:
    print(f.read())
````

````python
# Exemplo 7: Manipulação com os (criar diretório, verificar, excluir)
import os

# Verificar se arquivo existe
print(f"exemplo.txt existe? {os.path.exists('exemplo.txt')}")
print(f"frutas.txt existe? {os.path.exists('frutas.txt')}")

# Excluir arquivo
if os.path.exists("frutas.txt"):
    os.remove("frutas.txt")
    print("frutas.txt removido")

# Listar diretório
print("Arquivos no diretório:")
for item in os.listdir("."):
    if os.path.isfile(item):
        print(f"  📄 {item}")
    else:
        print(f"  📁 {item}")
````

````python
# Exemplo 8: Modo x (criação exclusiva) e modo binário
import tempfile
import os

# Criação exclusiva - só funciona se o arquivo não existir
try:
    with open("novo_exclusivo.txt", "x", encoding="utf-8") as f:
        f.write("Arquivo criado com modo x\n")
    print("Arquivo criado com sucesso!")
    # Remove para não poluir
    os.remove("novo_exclusivo.txt")
except FileExistsError:
    print("Arquivo já existe!")

# Modo binário (útil para imagens, arquivos não-texto)
with open("exemplo_binario.bin", "wb") as f:
    f.write(b"\x48\x65\x6c\x6c\x6f")  # "Hello" em hex

with open("exemplo_binario.bin", "rb") as f:
    dados = f.read()
    print(f"Lido em binário: {dados}")
    print(f"Como string: {dados.decode('utf-8')}")

os.remove("exemplo_binario.bin")
````

## ✅ Exercícios Resolvidos

````python
# Exercício 1: Copiar conteúdo de um arquivo para outro
with open("exemplo.txt", "r", encoding="utf-8") as origem:
    conteudo = origem.read()

with open("copia.txt", "w", encoding="utf-8") as destino:
    destino.write(conteudo)

# Verificar
with open("copia.txt", "r", encoding="utf-8") as f:
    print(f"Conteúdo copiado:\n{f.read()}")

import os
os.remove("copia.txt")
````

````python
# Exercício 2: Contar linhas, palavras e caracteres de um arquivo
with open("exemplo.txt", "r", encoding="utf-8") as f:
    texto = f.read()

linhas = len(texto.split("\n"))
palavras = len(texto.split())
caracteres = len(texto)

print(f"Linhas: {linhas}")
print(f"Palavras: {palavras}")
print(f"Caracteres: {caracteres}")
````

````python
# Exercício 3: Escrever tabuada em arquivo
with open("tabuada_5.txt", "w", encoding="utf-8") as f:
    for i in range(1, 11):
        f.write(f"5 x {i:2d} = {5 * i:2d}\n")

print("Tabuada do 5:")
with open("tabuada_5.txt", "r", encoding="utf-8") as f:
    print(f.read())

import os
os.remove("tabuada_5.txt")
````

## 📌 Resumo

- **`open(caminho, modo)`**: abre um arquivo nos modos `r` (leitura), `w` (escrita), `a` (acrescentar), `x` (criação exclusiva)
- **`read()`**: lê o arquivo inteiro ou `n` caracteres
- **`readline()`**: lê uma linha
- **`readlines()`**: retorna lista de linhas
- **`write(texto)`**: escreve string no arquivo
- **`writelines(lista)`**: escreve lista de strings
- **`with open(...) as f:`**: gerencia fechamento automático
- **`os.remove()`**: exclui arquivo
- **`os.path.exists()`**: verifica existência
- **`os.listdir()`**: lista diretório
- Sempre feche arquivos ou use `with` para evitar vazamento de recursos

## 💡 Curiosidades

- O "`with`" é um **context manager** — um protocolo que define entrada e saída de um contexto. Você pode criar seus próprios context managers com `__enter__` e `__exit__`.
- Arquivos não fechados podem causar **vazamento de descritor de arquivo**, levando o sistema a ficar sem recursos.
- O módulo `pathlib` (Python 3.4+) oferece uma API orientada a objetos para manipular caminhos, mais moderna que `os.path`.
- Python lê arquivos texto usando o encoding do sistema por padrão. Especifique `encoding="utf-8"` para garantir compatibilidade.
- O maior arquivo que Python pode ler depende da memória RAM disponível. Para arquivos muito grandes, processe linha por linha.

## ✅ Boas Práticas

- **Sempre use `with`** para abrir arquivos — evita esquecer de fechar
- Especifique `encoding="utf-8"` para compatibilidade entre sistemas
- Use `os.path.exists()` antes de operações que podem falhar
- Para arquivos grandes, itere linha por linha em vez de usar `read()`
- Use `rstrip("\n")` ou `strip()` ao processar linhas
- Trate exceções com `try/except` ao manipular arquivos
- Prefira `pathlib` sobre `os.path` em código novo (Python 3.4+)

## ⚠️ Erros Comuns

| Erro | Exemplo | Correção |
|------|---------|----------|
| Esquecer de fechar o arquivo | `f = open(...)` sem `close()` | Usar `with open(...) as f:` |
| Esquecer encoding | `open("file.txt")` (pode quebrar acentos) | `open("file.txt", encoding="utf-8")` |
| FileNotFoundError | `open("inexistente.txt")` | Verificar com `os.path.exists()` |
| PermissionError | Tentar escrever em diretório sem permissão | Verificar permissões e usar try/except |
| Sobrescrever arquivo sem querer | Usar `"w"` em vez de `"a"` | Entender diferença entre w, a, x |
| `\n` não adicionado no write | `f.write("linha")` (sem quebra de linha) | Adicionar `\n`: `f.write("linha\n")` |
| Usar `read()` em arquivo muito grande | Travamento por falta de memória | Ler linha por linha com `for linha in f:` |
