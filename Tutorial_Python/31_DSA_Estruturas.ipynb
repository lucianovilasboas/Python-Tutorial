## 📝 Introducao

**DSA (Data Structures and Algorithms)** e o estudo de como organizar e processar dados de forma eficiente. Estruturas de dados bem escolhidas podem transformar um programa lento em um rapido.

### Por que estudar DSA?

- **Eficiencia**: escolher a estrutura certa pode reduzir tempo de O(n^2) para O(n log n)
- **Entrevistas**: e o topico mais cobrado em entrevistas tecnicas
- **Fundamentos**: entender como as coisas funcionam "por baixo dos panos"
- **Resolucao de problemas**: desenvolve pensamento computacional

## 📚 Explicacao Detalhada

### Listas e Arrays

- **Lista Python** (`list`): dinamica, heterogenea, implementada como array de ponteiros
- **Array** (`array.array` ou `numpy.array`): tamanho fixo, mesmo tipo, mais eficiente

| Operacao | Lista Python | Array (numpy) |
|----------|-------------|---------------|
| Acesso por indice | O(1) | O(1) |
| Insercao no final | O(1)* | O(n) (realocacao) |
| Insercao no inicio | O(n) | O(n) |
| Busca | O(n) | O(n) |

### Pilha (Stack) — LIFO

Ultimo a entrar, primeiro a sair (Last In, First Out).

- **push**: adicionar no topo
- **pop**: remover do topo
- **peek**: ver o topo sem remover

Em Python, listas funcionam como pilhas: `append()` = push, `pop()` = pop.

### Fila (Queue) — FIFO

Primeiro a entrar, primeiro a sair (First In, First Out).

- **enqueue**: adicionar no final
- **dequeue**: remover do inicio

Usar `collections.deque` para O(1) em ambas as pontas.

### Lista Encadeada (Linked List)

Cada no contem um valor e uma referencia ao proximo no. Vantagens: insercao e remocao O(1) em qualquer posicao. Desvantagem: acesso O(n) por indice.

### Tabela Hash (Hash Table)

Estrutura que mapeia chaves a valores usando uma funcao hash. Em Python: `dict` e `set`.

- Busca, insercao, remocao: O(1) medio

## 💻 Exemplos

```python
# Exemplo 1: Implementando uma Pilha (Stack)
class Pilha:
    def __init__(self):
        self.itens = []

    def push(self, item):
        self.itens.append(item)

    def pop(self):
        if not self.esta_vazia():
            return self.itens.pop()
        raise IndexError("Pilha vazia")

    def peek(self):
        if not self.esta_vazia():
            return self.itens[-1]
        raise IndexError("Pilha vazia")

    def esta_vazia(self):
        return len(self.itens) == 0

    def tamanho(self):
        return len(self.itens)

    def __str__(self):
        return str(self.itens)

pilha = Pilha()
pilha.push("A")
pilha.push("B")
pilha.push("C")
print(f"Pilha: {pilha}")
print(f"Topo: {pilha.peek()}")
print(f"Removendo: {pilha.pop()}")
print(f"Pilha apos pop: {pilha}")
```

```python
# Exemplo 2: Implementando uma Fila (Queue)
from collections import deque

class Fila:
    def __init__(self):
        self.itens = deque()

    def enqueue(self, item):
        self.itens.append(item)

    def dequeue(self):
        if not self.esta_vazia():
            return self.itens.popleft()
        raise IndexError("Fila vazia")

    def frente(self):
        if not self.esta_vazia():
            return self.itens[0]
        raise IndexError("Fila vazia")

    def esta_vazia(self):
        return len(self.itens) == 0

    def __str__(self):
        return str(list(self.itens))

fila = Fila()
fila.enqueue(1)
fila.enqueue(2)
fila.enqueue(3)
print(f"Fila: {fila}")
print(f"Removendo: {fila.dequeue()}")
print(f"Proximo: {fila.frente()}")
print(f"Fila apos dequeue: {fila}")
```

```python
# Exemplo 3: Aplicacao pratica — verificador de parenteses (usa pilha)
def parenteses_balanceados(expressao):
    pilha = []
    pares = {')': '(', ']': '[', '}': '{'}

    for char in expressao:
        if char in '([{':
            pilha.append(char)
        elif char in ')]}':
            if not pilha or pilha.pop() != pares[char]:
                return False
    return len(pilha) == 0

expressoes = [
    "(a + b) * [c - {d / e}]",
    "(a + b] * c",
    "((a + b)",
    "[()]{}"
]

for expr in expressoes:
    resultado = parenteses_balanceados(expr)
    print(f"'{expr:<25s}' -> {'Balanceado' if resultado else 'Nao balanceado'}")
```

```python
# Exemplo 4: Lista Encadeada (Linked List)
class No:
    def __init__(self, valor):
        self.valor = valor
        self.proximo = None

class ListaEncadeada:
    def __init__(self):
        self.cabeca = None

    def inserir_inicio(self, valor):
        novo = No(valor)
        novo.proximo = self.cabeca
        self.cabeca = novo

    def inserir_fim(self, valor):
        novo = No(valor)
        if not self.cabeca:
            self.cabeca = novo
            return
        atual = self.cabeca
        while atual.proximo:
            atual = atual.proximo
        atual.proximo = novo

    def remover(self, valor):
        if not self.cabeca:
            return
        if self.cabeca.valor == valor:
            self.cabeca = self.cabeca.proximo
            return
        atual = self.cabeca
        while atual.proximo and atual.proximo.valor != valor:
            atual = atual.proximo
        if atual.proximo:
            atual.proximo = atual.proximo.proximo

    def __str__(self):
        valores = []
        atual = self.cabeca
        while atual:
            valores.append(str(atual.valor))
            atual = atual.proximo
        return " -> ".join(valores) if valores else "vazia"

lista = ListaEncadeada()
lista.inserir_inicio(3)
lista.inserir_inicio(2)
lista.inserir_inicio(1)
lista.inserir_fim(4)
lista.inserir_fim(5)
print(f"Lista: {lista}")
lista.remover(3)
print(f"Apos remover 3: {lista}")
lista.remover(1)
print(f"Apos remover 1: {lista}")
```

```python
# Exemplo 5: Tabela Hash (dict) — operacoes e colisoes
# Python dict usa hash table internamente

# Demonstrando O(1) medio para busca
import time

n = 1_000_000
dicionario = {i: i * 2 for i in range(n)}
lista_grande = list(range(n))

# Busca no dicionario O(1)
inicio = time.perf_counter()
valor = dicionario[n - 1]
fim = time.perf_counter()
tempo_dict = fim - inicio

# Busca na lista O(n)
inicio = time.perf_counter()
indice = lista_grande.index(n - 1)
fim = time.perf_counter()
tempo_lista = fim - inicio

print(f"Tempo de busca no dict (O(1)): {tempo_dict*1000:.4f} ms")
print(f"Tempo de busca na lista (O(n)): {tempo_lista*1000:.4f} ms")
print(f"Dict foi {tempo_lista/tempo_dict:.0f}x mais rapido!")
```

```python
# Exemplo 6: Aplicacao de fila — simulador de fila de banco
import random

class SimuladorBanco:
    def __init__(self):
        self.fila = Fila()
        self.senha_atual = 1

    def novo_cliente(self):
        senha = self.senha_atual
        self.fila.enqueue(senha)
        self.senha_atual += 1
        return senha

    def chamar_proximo(self):
        if not self.fila.esta_vazia():
            return self.fila.dequeue()
        return None

    def pessoas_na_fila(self):
        return len(self.fila.itens)

banco = SimuladorBanco()
print("=== Simulador de Fila de Banco ===\n")

for i in range(5):
    senha = banco.novo_cliente()
    print(f"  Cliente {i+1} chegou — Senha: {senha:03d}")

print(f"\n  Pessoas na fila: {banco.pessoas_na_fila()}")
print()

for i in range(3):
    senha = banco.chamar_proximo()
    print(f"  Chamando senha: {senha:03d}")

print(f"\n  Restam na fila: {banco.pessoas_na_fila()} pessoas")
```

```python
# Exemplo 7: Implementando uma tabela hash simples
class TabelaHash:
    def __init__(self, tamanho=10):
        self.tamanho = tamanho
        self.tabela = [[] for _ in range(tamanho)]

    def _hash(self, chave):
        return hash(chave) % self.tamanho

    def inserir(self, chave, valor):
        indice = self._hash(chave)
        # Atualizar se chave ja existe
        for i, (k, v) in enumerate(self.tabela[indice]):
            if k == chave:
                self.tabela[indice][i] = (chave, valor)
                return
        self.tabela[indice].append((chave, valor))

    def buscar(self, chave):
        indice = self._hash(chave)
        for k, v in self.tabela[indice]:
            if k == chave:
                return v
        raise KeyError(chave)

    def remover(self, chave):
        indice = self._hash(chave)
        for i, (k, v) in enumerate(self.tabela[indice]):
            if k == chave:
                del self.tabela[indice][i]
                return
        raise KeyError(chave)

    def __str__(self):
        return str(self.tabela)

ht = TabelaHash(5)
ht.inserir("nome", "Joao")
ht.inserir("idade", 25)
ht.inserir("cidade", "BH")
print(f"Tabela Hash: {ht}")
print(f"Buscar 'nome': {ht.buscar('nome')}")
print(f"Buscar 'cidade': {ht.buscar('cidade')}")
ht.remover("idade")
print(f"Apos remover 'idade': {ht}")
```

```python
# Exercicio 1: Inverter string com pilha
def inverter_string(texto):
    pilha = list(texto)
    invertida = []
    while pilha:
        invertida.append(pilha.pop())
    return ''.join(invertida)

original = "Python DSA"
invertida = inverter_string(original)
print(f"Original: {original}")
print(f"Invertida: {invertida}")
```

```python
# Exercicio 2: Contagem de frequencia com tabela hash
def frequencia_palavras(texto):
    palavras = texto.lower().split()
    freq = {}
    for palavra in palavras:
        freq[palavra] = freq.get(palavra, 0) + 1
    return freq

texto = "o rato roeu a roupa do rei de roma o rato"
freq = frequencia_palavras(texto)

for palavra, contagem in sorted(freq.items(), key=lambda x: -x[1]):
    print(f"  {palavra}: {contagem}")
```

## 📌 Resumo

- **Stack (Pilha)**: LIFO — push/pop, util para backtracking, undo, parenteses
- **Queue (Fila)**: FIFO — enqueue/dequeue, util para simulacoes, BFS
- **Linked List**: nos encadeados — insercao O(1), busca O(n)
- **Hash Table**: dict/set — busca O(1) medio, mapeamento chave-valor
- **Array vs List**: array tem tipo fixo e e mais eficiente; lista e dinamica
- Estruturas lineares sao a base para estruturas mais complexas (arvores, grafos)

## 💡 Curiosidades

- O conceito de pilha foi formalizado por **Alan Turing** em 1946 e e a base da recursao (call stack).
- A fila de prioridade e implementada com **heap**, uma estrutura de arvore que garante que o menor (ou maior) elemento esta sempre no topo.
- O dicionario Python (dict) usa uma **tabela hash aberta com sondagem linear**, otimizada para minimizar colisoes.
- **Linked Lists** sao a base para implementar sistemas de arquivos (FAT, NTFS) e alocacao de memoria.
- O botao "voltar" do navegador e uma pilha; o botao "avancar" e outra pilha que armazena as paginas desempilhadas.

## ✅ Boas Praticas

- Use `collections.deque` para filas (O(1) em ambas as pontas)
- Use `list` como pilha com `append()`/`pop()`
- Prefira `dict`/`set` a `list` quando precisar de buscas frequentes
- Para grandes volumes de dados numericos, use `numpy.array` em vez de `list`
- Sempre trate operacoes em estruturas vazias (pop em pilha vazia, etc.)

## ⚠️ Erros Comuns

| Erro | Exemplo | Correcao |
|------|---------|----------|
| Pop em pilha vazia | `pilha.pop()` sem verificar | Verificar `esta_vazia()` antes |
| Usar lista como fila | `lista.pop(0)` e O(n) | Usar `collections.deque` |
| Iterar e modificar linked list | `while no: no = no.proximo` ao remover | Manter referencia ao no anterior |
| Hash de objetos mutaveis | `dict[lista]` (TypeError) | Usar apenas objetos imutaveis como chave |
| Colisoes nao tratadas em hash propria | Dados sobrescritos silenciosamente | Implementar encadeamento ou sondagem || Confundir LIFO com FIFO | Usar pilha quando precisa de fila | Entender a diferenca conceitual |
