## 📝 Introdução

Em Python, tudo é um **objeto** e cada objeto possui um **tipo** que determina o que podemos fazer com ele. Saber identificar e manipular tipos é essencial para escrever programas corretos. Nesta aula vamos explorar os tipos numéricos, o conceito de *casting* e os tipos de dados fundamentais da linguagem.

## 📚 Explicação Detalhada

### Tipos de Dados Básicos

| Tipo | Classe | Exemplo |
|------|--------|--------|
| `int` | Números inteiros | `42`, `-3`, `0` |
| `float` | Números decimais | `3.14`, `-0.5`, `1e3` |
| `complex` | Números complexos | `3+4j`, `-2j` |
| `str` | Texto (string) | `"Olá"`, `'Python'` |
| `bool` | Lógico | `True`, `False` |
| `list` | Lista ordenada | `[1, 2, 3]` |
| `tuple` | Tupla imutável | `(1, 2, 3)` |
| `set` | Conjunto sem duplicatas | `{1, 2, 3}` |
| `dict` | Dicionário chave-valor | `{"a": 1}` |

### Números

Python suporta três tipos numéricos:
- **int**: inteiros sem limite de tamanho
- **float**: números de ponto flutuante (precisão dupla)
- **complex**: parte real + imaginária (`j` suffix)

### Casting

Casting é a conversão explícita de um tipo para outro usando funções como `int()`, `float()` e `str()`.

## 💻 Exemplos

```python
# Exemplo 1: Usando type() para descobrir o tipo
print(type(10))
print(type(3.14))
print(type("Olá"))
print(type(True))
print(type([1, 2, 3]))
print(type({"nome": "Ana"}))
```

```python
# Exemplo 2: Números inteiros (int)
a = 42
b = -1000
c = 0
grande = 2 ** 100  # inteiro arbitrariamente grande
print(f"a = {a}, tipo: {type(a)}")
print(f"grande = {grande}, tipo: {type(grande)}")
```

```python
# Exemplo 3: Números float
x = 3.14159
y = -0.5
z = 1.2e3  # notação científica = 1200.0
print(f"x = {x}, y = {y}, z = {z}")
print(f"Tipo de x: {type(x)}")
```

```python
# Exemplo 4: Números complexos
c1 = 3 + 4j
c2 = -2j
print(f"c1 = {c1}, parte real: {c1.real}, imaginária: {c1.imag}")
print(f"c2 = {c2}, tipo: {type(c2)}")
```

```python
# Exemplo 5: Casting int() e float()
# float para int (trunca a parte decimal)
print(int(3.99))   # 3
print(int(-2.7))   # -2

# str numérica para int/float
print(int("42") + 8)
print(float("3.14") * 2)

# int para float
print(float(10))
```

```python
# Exemplo 6: Casting str()
idade = 25
altura = 1.75
texto = "Tenho " + str(idade) + " anos e " + str(altura) + " m de altura."
print(texto)

# str() com booleanos
print(str(True) + " é um valor lógico.")
```

```python
# Exemplo 7: Números aleatórios com random
import random

print("Aleatório entre 1 e 10:", random.randint(1, 10))
print("Aleatório entre 0.0 e 1.0:", random.random())
print("Aleatório entre 5.5 e 10.5:", random.uniform(5.5, 10.5))
print("Escolha de uma lista:", random.choice(["maçã", "banana", "cereja"]))
```

```python
# Exemplo 8: Booleanos como números
print("True + True =", True + True)   # 2
print("False * 10 =", False * 10)    # 0
print("int(True) =", int(True))      # 1
print("bool(1) =", bool(1))          # True
print("bool(0) =", bool(0))          # False
print("bool('') =", bool(""))        # False
print("bool('Python') =", bool("Python"))  # True
```

## ✅ Exercícios Resolvidos

```python
# Exercício Resolvido 1: Conversão de temperatura
# Converter 25°C para Fahrenheit e exibir o tipo do resultado
celsius = 25
fahrenheit = (celsius * 9/5) + 32
print(f"{celsius}°C = {fahrenheit}°F")
print(f"Tipo do resultado: {type(fahrenheit)}")
```

```python
# Exercício Resolvido 2: Soma de dígitos de um número
# Solicita um número inteiro e soma seus dígitos (usando casting)
numero = input("Digite um número inteiro: ")
soma = 0
for digito in numero:
    soma += int(digito)
print(f"A soma dos dígitos de {numero} é {soma}")
```

```python
# Exercício Resolvido 3: Jogo de adivinhação simples
import random

sorteado = random.randint(1, 10)
palpite = int(input("Adivinhe o número entre 1 e 10: "))
if palpite == sorteado:
    print("Parabéns! Você acertou!")
else:
    print(f"Errou! O número era {sorteado}.")
```

## 📌 Resumo

- Python possui tipos dinâmicos: a variável assume o tipo do valor atribuído
- Use `type()` para inspecionar o tipo de qualquer objeto
- **int**, **float** e **complex** são os tipos numéricos
- Casting explícito com `int()`, `float()`, `str()`, `bool()`
- O módulo `random` oferece funções para gerar números aleatórios
- Booleanos são um subtipo de inteiros (`True=1`, `False=0`)

## 💡 Curiosidades

- Python pode lidar com números inteiros **gigantescos** (limitados apenas pela memória).
- `0.1 + 0.2` não é exatamente `0.3` devido à representação binária de ponto flutuante (IEEE 754).
- O módulo `decimal` permite aritmética decimal exata para aplicações financeiras.

## ✅ Boas Práticas

- Sempre use `int()` ou `float()` para converter entradas do usuário (`input()` retorna string)
- Prefira f-strings para formatar números como strings
- Evite comparar floats com `==`; use `abs(a - b) < 1e-9`
- Documente o tipo esperado em funções usando type hints (`def soma(a: int, b: int) -> int:`)

## ⚠️ Erros Comuns

| Erro | Exemplo | Solução |
|------|---------|--------|
| Concatenar str + int | `"Idade: " + 25` | Use `str(25)` ou f-string |
| Converter string inválida | `int("abc")` | Verifique com `.isdigit()` ou use try/except |
| Perda de precisão | `int(3.999)` → `3` | Não assuma arredondamento; use `round()` |
| Confundir `random.random()` com `random.randint()` | `random.random()` retorna float | Leia a documentação de cada função |
