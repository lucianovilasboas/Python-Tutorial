## 🎯 Objetivos de Aprendizagem

- Criar multiplos graficos na mesma figura com `subplot()`

- Desenhar graficos de barras e barras horizontais

- Personalizar cada tipo de grafico com cores, legendas e formatacao

## 📝 Introducao

Alem dos graficos de linha basicos, o Matplotlib oferece uma rica variedade de tipos de graficos para diferentes necessidades de visualizacao:

- **Subplots**: organizam varios graficos lado a lado em uma unica figura
- **Scatter**: mostram a relacao entre duas variaveis numericas (dispersao)
- **Bars**: comparam categorias ou grupos (graficos de barras)
- **Histograms**: visualizam a distribuicao de frequencia de dados
- **Pie Charts**: mostram proporcoes e percentuais de um todo

Dominar esses tipos de graficos e essencial para analise de dados, relatorios e dashboards.

## 📚 Explicacao Detalhada

### Subplots

`plt.subplot(linhas, colunas, posicao)` cria uma grade de graficos. Alternativa moderna: `plt.subplots()`.

```python
plt.subplot(2, 2, 1)  # 2 linhas, 2 colunas, posicao 1
plt.plot(x, y)

plt.subplot(2, 2, 2)  # posicao 2
plt.plot(x, z)
```

### Scatter Plot (Grafico de Dispersao)

`plt.scatter(x, y)` plota pontos individuais sem conecta-los. Util para ver correlacoes.

- `c` ou `color`: cor dos pontos
- `s` ou `size`: tamanho dos pontos
- `alpha`: transparencia (0 a 1)
- `cmap`: mapa de cores para representar uma terceira variavel

### Grafico de Barras (Bar Chart)

`plt.bar(x, y)` desenha barras verticais. `plt.barh(x, y)` para barras horizontais.

- `width`: largura da barra
- `color`: cor da barra
- `edgecolor`: cor da borda

### Histograma

`plt.hist(dados)` agrupa dados em intervalos (bins) e mostra a frequencia.

- `bins`: numero de intervalos
- `alpha`: transparencia
- `edgecolor`: cor da borda das barras

### Grafico de Pizza (Pie Chart)

`plt.pie(valores)` mostra proporcoes como fatias de uma pizza.

- `labels`: nome de cada fatia
- `autopct`: formato do percentual (ex: `'%1.1f%%'`)
- `explode`: destacar fatias
- `shadow`: sombra 3D

## 💻 Exemplos

````python
# Exemplo 1: Subplots (2x2)
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

# Grade 2x2
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x), 'r-')
plt.title('Seno')
plt.grid(True, alpha=0.3)

plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), 'b-')
plt.title('Cosseno')
plt.grid(True, alpha=0.3)

plt.subplot(2, 2, 3)
plt.plot(x, x**2, 'g-')
plt.title('x^2')
plt.grid(True, alpha=0.3)

plt.subplot(2, 2, 4)
plt.plot(x, np.sqrt(x), 'm-')
plt.title('sqrt(x)')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

````python
# Exemplo 2: Scatter Plot basico
np.random.seed(42)
x = np.random.normal(50, 15, 200)
y = np.random.normal(50, 15, 200)
cores = np.random.rand(200)
tamanhos = np.random.randint(20, 200, 200)

plt.figure(figsize=(8, 6))
scatter = plt.scatter(x, y, c=cores, s=tamanhos, alpha=0.6, cmap='viridis')
plt.colorbar(scatter, label='Intensidade')
plt.xlabel('Variavel X')
plt.ylabel('Variavel Y')
plt.title('Scatter Plot com Cor e Tamanho Variaveis')
plt.grid(True, alpha=0.3)
plt.show()
````

````python
# Exemplo 3: Grafico de Barras
categorias = ['Python', 'Java', 'JavaScript', 'C++', 'C#', 'PHP']
popularidade = [30, 25, 22, 18, 15, 10]
cores = ['#3776AB', '#ED8B00', '#F7DF1E', '#00599C', '#68217A', '#777BB4']

plt.figure(figsize=(10, 6))
barras = plt.bar(categorias, popularidade, color=cores, edgecolor='black', linewidth=0.5)

# Adicionar valores no topo das barras
for barra, valor in zip(barras, popularidade):
    plt.text(barra.get_x() + barra.get_width()/2, barra.get_height() + 0.5,
             str(valor), ha='center', fontweight='bold')

plt.xlabel('Linguagem')
plt.ylabel('Popularidade (%)')
plt.title('Popularidade de Linguagens de Programacao')
plt.ylim(0, 35)
plt.grid(axis='y', alpha=0.3)
plt.show()
````

````python
# Exemplo 4: Barras Horizontais (barh)
disciplinas = ['Matematica', 'Portugues', 'Historia', 'Fisica', 'Quimica']
notas = [8.5, 7.2, 9.0, 6.8, 8.1]

plt.figure(figsize=(8, 5))
cores = plt.cm.viridis(np.linspace(0.2, 0.8, len(disciplinas)))
plt.barh(disciplinas, notas, color=cores, edgecolor='black')
plt.xlabel('Nota Media')
plt.title('Notas Medias por Disciplina')
plt.xlim(0, 10)
plt.grid(axis='x', alpha=0.3)

for i, (disciplina, nota) in enumerate(zip(disciplinas, notas)):
    plt.text(nota + 0.1, i, f'{nota:.1f}', va='center', fontweight='bold')

plt.show()
````

````python
# Exemplo 5: Histograma
np.random.seed(42)
idades = np.random.normal(25, 8, 1000)  # media 25, desvio 8
idades = np.clip(idades, 0, 60)

plt.figure(figsize=(10, 6))
plt.hist(idades, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.title('Distribuicao de Idades')

# Linhas de referencia
plt.axvline(np.mean(idades), color='red', linestyle='--', linewidth=2, label=f'Media: {np.mean(idades):.1f}')
plt.axvline(np.median(idades), color='green', linestyle='-', linewidth=2, label=f'Mediana: {np.median(idades):.1f}')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
````

````python
# Exemplo 6: Histograma comparativo (sobreposicao)
np.random.seed(42)
grupo_A = np.random.normal(60, 10, 500)
grupo_B = np.random.normal(70, 15, 500)

plt.figure(figsize=(10, 6))
plt.hist(grupo_A, bins=30, alpha=0.5, color='blue', edgecolor='black', label='Grupo A')
plt.hist(grupo_B, bins=30, alpha=0.5, color='red', edgecolor='black', label='Grupo B')
plt.xlabel('Valor')
plt.ylabel('Frequencia')
plt.title('Comparacao de Distribuicoes')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
````

````python
# Exemplo 7: Grafico de Pizza (Pie Chart)
categorias = ['Alimentacao', 'Transporte', 'Moradia', 'Lazer', 'Saude', 'Educacao']
gastos = [800, 300, 1200, 400, 350, 250]
cores = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7', '#DDA0DD']
destaque = (0, 0, 0.1, 0, 0, 0)  # destacar Moradia

plt.figure(figsize=(8, 8))
plt.pie(gastos, labels=categorias, autopct='%1.1f%%', explode=destaque,
        colors=cores, shadow=True, startangle=90,
        textprops={'fontsize': 11})
plt.title('Distribuicao de Gastos Mensais', fontsize=14, fontweight='bold')
plt.show()
````

````python
# Exemplo 8: Subplots com tipos diferentes de graficos
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Diferentes Tipos de Graficos', fontsize=16, fontweight='bold')

np.random.seed(42)

# 1. Linha
x = np.linspace(0, 10, 50)
axes[0, 0].plot(x, np.sin(x), 'r-o', markersize=4)
axes[0, 0].set_title('Linha (Seno)')
axes[0, 0].grid(True, alpha=0.3)

# 2. Scatter
x_scatter = np.random.rand(50) * 100
y_scatter = np.random.rand(50) * 100
axes[0, 1].scatter(x_scatter, y_scatter, c=np.random.rand(50), alpha=0.7, cmap='plasma')
axes[0, 1].set_title('Scatter')
axes[0, 1].grid(True, alpha=0.3)

# 3. Barras
cats = ['A', 'B', 'C', 'D', 'E']
vals = np.random.randint(10, 50, 5)
axes[0, 2].bar(cats, vals, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7'])
axes[0, 2].set_title('Barras')
axes[0, 2].grid(axis='y', alpha=0.3)

# 4. Histograma
data = np.random.normal(0, 1, 500)
axes[1, 0].hist(data, bins=25, color='purple', edgecolor='black', alpha=0.7)
axes[1, 0].set_title('Histograma')
axes[1, 0].grid(axis='y', alpha=0.3)

# 5. Pizza
sizes = [25, 20, 18, 15, 12, 10]
axes[1, 1].pie(sizes, labels=['A', 'B', 'C', 'D', 'E', 'F'], autopct='%1.0f%%',
               colors=plt.cm.Set3(np.linspace(0, 1, 6)))
axes[1, 1].set_title('Pizza')

# 6. Barras horizontais
axes[1, 2].barh(['X', 'Y', 'Z', 'W'], [30, 25, 20, 15], color='teal', edgecolor='black')
axes[1, 2].set_title('Barra Horizontal')
axes[1, 2].grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()
````

````python
# Exercicio 1: Subplot com 3 graficos (seno, cosseno, tangente)
x = np.linspace(-np.pi, np.pi, 100)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].plot(x, np.sin(x), 'r-')
axes[0].set_title('Seno')
axes[0].grid(True, alpha=0.3)
axes[0].axhline(y=0, color='black', linewidth=0.5)

axes[1].plot(x, np.cos(x), 'b-')
axes[1].set_title('Cosseno')
axes[1].grid(True, alpha=0.3)
axes[1].axhline(y=0, color='black', linewidth=0.5)

axes[2].plot(x, np.tan(x), 'g-')
axes[2].set_title('Tangente')
axes[2].set_ylim(-5, 5)
axes[2].grid(True, alpha=0.3)
axes[2].axhline(y=0, color='black', linewidth=0.5)

plt.tight_layout()
plt.show()
````

````python
# Exercicio 2: Scatter plot de correlacao altura x peso
np.random.seed(42)
altura = np.random.normal(170, 10, 100)
peso = altura * 0.65 + np.random.normal(0, 5, 100)  # correlacao positiva

# Calcular linha de tendencia
coef = np.polyfit(altura, peso, 1)
linha_tendencia = np.poly1d(coef)
x_linha = np.linspace(altura.min(), altura.max(), 100)

plt.figure(figsize=(8, 6))
plt.scatter(altura, peso, alpha=0.6, c='steelblue', edgecolors='black')
plt.plot(x_linha, linha_tendencia(x_linha), 'r--', linewidth=2, label='Tendencia')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Relacao Altura x Peso')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

````python
# Exercicio 3: Pie chart de despesas com destaque
despesas = [1200, 500, 300, 200, 400, 250]
rotulos = ['Aluguel', 'Alimentacao', 'Transporte', 'Internet', 'Lazer', 'Outros']
destaques = [0.1, 0, 0, 0, 0, 0]

plt.figure(figsize=(8, 8))
plt.pie(despesas, labels=rotulos, autopct='%1.1f%%', explode=destaques,
        shadow=True, startangle=90,
        colors=plt.cm.Paired(np.linspace(0, 1, len(despesas))))
plt.title('Despesas Mensais', fontsize=14, fontweight='bold')
plt.show()
````

## 📌 Resumo

- **Subplots**: `plt.subplot(lin, col, pos)` ou `plt.subplots()` para multiplos graficos
- **Scatter**: `plt.scatter(x, y)` — pontos com cor (`c`), tamanho (`s`), transparencia (`alpha`)
- **Barras**: `plt.bar(x, y)` verticais, `plt.barh(x, y)` horizontais
- **Histograma**: `plt.hist(dados, bins=N)` — distribuicao de frequencia
- **Pizza**: `plt.pie(valores)` — proporcoes com `labels`, `autopct`, `explode`
- Use `plt.tight_layout()` para ajustar espacamento automaticamente
- `plt.figure(figsize=(w, h))` define o tamanho da figura

## 💡 Curiosidades

- O scatter plot foi inventado por **John Herschel** em 1833, muito antes dos computadores.
- O grafico de pizza (pie chart) foi popularizado por **William Playfair** em 1801, considerado o pai dos graficos estatisticos.
- **Florence Nightingale** usou graficos de pizza polares (rosas) para mostrar que mais soldados morriam de doencas do que em batalha durante a Guerra da Crimeia.
- Matplotlib pode emular estilos visuais como `ggplot` (do R), `seaborn`, `fivethirtyeight` e outros com `plt.style.use('nome')`.
- O histograma foi introduzido por **Karl Pearson** em 1895.

## ✅ Boas Praticas

- Use `plt.tight_layout()` ou `constrained_layout=True` ao criar subplots
- Adicione `alpha` (transparencia) em scatter plots com muitos pontos sobrepostos
- Ordene as fatias do pie chart do maior para o menor
- Use cores com contraste adequado e acessiveis para daltonicos
- Sempre rotule os eixos — um grafico sem rotulos e inutil
- Prefira barras horizontais quando os rotulos das categorias forem longos
- Adicione barras de erro (`yerr`) quando apropriado em barras e linhas

## ⚠️ Erros Comuns

| Erro | Exemplo | Correcao |
|------|---------|----------|
| Esquecer `plt.show()` | Grafico nao renderiza | Sempre chamar `plt.show()` |
| Subplot sobreposto | Graficos se sobrepoem | Usar `plt.tight_layout()` |
| Muitas fatias no pie chart | Grafico ilegivel | Agrupar categorias pequenas em "Outros" |
| Bins inadequados no histograma | Distribuicao nao aparece bem | Testar diferentes valores de `bins` |
| Tamanho de figura padrao pequeno | Grafico apertado | Usar `figsize=(largura, altura)` |
| Nao usar `np.array()` | Erro ao plotar listas vs arrays | Converter com `np.array(dados)` |
