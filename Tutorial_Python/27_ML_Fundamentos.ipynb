## 📝 Introducao

**Machine Learning (Aprendizado de Maquina)** e um ramo da inteligencia artificial que permite que sistemas aprendam e melhorem a partir da experiencia, sem serem explicitamente programados.

Antes de aplicar algoritmos de ML, e fundamental entender os dados. Isso inclui medidas estatisticas basicas como media, mediana, moda, desvio padrao e percentis, que nos ajudam a resumir e interpretar conjuntos de dados.

### Tipos de Machine Learning

- **Supervisionado**: o modelo aprende com dados rotulados (ex: classificacao, regressao)
- **Nao supervisionado**: o modelo encontra padroes em dados nao rotulados (ex: clustering)
- **Por reforco**: o modelo aprende por tentativa e erro, recebendo recompensas

## 📚 Explicacao Detalhada

### Media (Mean)

A soma de todos os valores dividida pelo numero de elementos. Representa o valor "tipico" do conjunto.

```python
media = sum(valores) / len(valores)
# ou
media = np.mean(valores)
```

### Mediana (Median)

O valor central quando os dados estao ordenados. Menos sensivel a outliers que a media.

```python
mediana = np.median(valores)
```

### Moda (Mode)

O valor que aparece com mais frequencia no conjunto de dados.

```python
from scipy import stats
moda = stats.mode(valores)
```

### Desvio Padrao (Standard Deviation)

Mede a dispersao dos dados em relacao a media. Um desvio padrao baixo indica que os dados estao proximos da media; um alto indica dados espalhados.

```python
desvio = np.std(valores)
variancia = np.var(valores)
```

### Percentil (Percentile)

O valor abaixo do qual uma certa porcentagem dos dados cai. Ex: o percentil 25 (Q1) e o valor abaixo do qual 25% dos dados estao.

```python
p25 = np.percentile(valores, 25)
p50 = np.percentile(valores, 50)  # = mediana
p75 = np.percentile(valores, 75)  # Q3
```

### Distribuicao de Dados

Conjuntos de dados podem ter diferentes formatos de distribuicao. A **distribuicao normal** (curva de sino ou Gaussiana) e a mais importante em ML, caracterizada por:
- Simetria em torno da media
- 68% dos dados dentro de 1 desvio padrao da media
- 95% dentro de 2 desvios padrao
- 99.7% dentro de 3 desvios padrao

## 💻 Exemplos

````python
# Exemplo 1: Media, Mediana e Moda
import numpy as np
from scipy import stats

notas = np.array([7.0, 8.5, 9.0, 6.5, 7.0, 8.0, 9.5, 7.0, 6.0, 8.5])

media = np.mean(notas)
mediana = np.median(notas)
moda_result = stats.mode(notas, keepdims=True)

print(f"Notas: {notas}")
print(f"Media: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Moda: {moda_result.mode[0]} (aparece {moda_result.count[0]} vezes)")
````

````python
# Exemplo 2: Impacto de outliers na media vs mediana
salarios = np.array([2000, 2200, 2500, 2800, 3000, 3200, 3500, 50000])  # um CEO ganha 50k

print(f"Salarios: {salarios}")
print(f"Media (com outlier): R$ {np.mean(salarios):.2f}")
print(f"Mediana (com outlier): R$ {np.median(salarios):.2f}")

salarios_sem_outlier = salarios[:-1]
print(f"\nSem o CEO:")
print(f"Media (sem outlier): R$ {np.mean(salarios_sem_outlier):.2f}")
print(f"Mediana (sem outlier): R$ {np.median(salarios_sem_outlier):.2f}")
````

````python
# Exemplo 3: Desvio Padrao e Variancia
np.random.seed(42)

dados_pouca_variacao = np.random.normal(50, 3, 1000)
dados_muita_variacao = np.random.normal(50, 15, 1000)

print("Dados com POUCA variacao:")
print(f"  Media: {np.mean(dados_pouca_variacao):.2f}")
print(f"  Desvio Padrao: {np.std(dados_pouca_variacao):.2f}")
print(f"  Variancia: {np.var(dados_pouca_variacao):.2f}")

print("\nDados com MUITA variacao:")
print(f"  Media: {np.mean(dados_muita_variacao):.2f}")
print(f"  Desvio Padrao: {np.std(dados_muita_variacao):.2f}")
print(f"  Variancia: {np.var(dados_muita_variacao):.2f}")
````

````python
# Exemplo 4: Percentis e Quartis
np.random.seed(42)
idades = np.random.normal(35, 12, 1000)
idades = np.clip(idades, 0, 100)

q1 = np.percentile(idades, 25)
q2 = np.percentile(idades, 50)  # mediana
q3 = np.percentile(idades, 75)
p90 = np.percentile(idades, 90)
p10 = np.percentile(idades, 10)

print(f"Idade minima: {np.min(idades):.0f}")
print(f"Percentil 10: {p10:.1f}")
print(f"Q1 (25%): {q1:.1f}")
print(f"Q2 / Mediana (50%): {q2:.1f}")
print(f"Q3 (75%): {q3:.1f}")
print(f"Percentil 90: {p90:.1f}")
print(f"Idade maxima: {np.max(idades):.0f}")
print(f"\nIntervalo interquartil (IQR): {q3 - q1:.1f}")
````

````python
# Exemplo 5: Visualizando a Distribuicao Normal
import matplotlib.pyplot as plt

np.random.seed(42)
dados_normal = np.random.normal(0, 1, 10000)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(dados_normal, bins=50, color='steelblue', edgecolor='black', alpha=0.7, density=True)

# Curva teorica da normal
x = np.linspace(-4, 4, 100)
y = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)
plt.plot(x, y, 'r-', linewidth=2, label='Curva Normal Teorica')

plt.axvline(0, color='red', linestyle='--', label='Media = 0')
plt.axvline(-1, color='green', linestyle=':', label='±1 Desvio')
plt.axvline(1, color='green', linestyle=':')
plt.title('Distribuicao Normal (Histograma)')
plt.legend()
plt.grid(axis='y', alpha=0.3)

# Boxplot
plt.subplot(1, 2, 2)
plt.boxplot(dados_normal, vert=True, patch_artist=True,
            boxprops=dict(facecolor='steelblue', alpha=0.7))
plt.title('Boxplot da Distribuicao Normal')
plt.ylabel('Valor')
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
````

````python
# Exemplo 6: Diferentes tipos de distribuicao
np.random.seed(42)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Distribuicao Uniforme
uniformes = np.random.uniform(0, 10, 5000)
axes[0].hist(uniformes, bins=40, color='coral', edgecolor='black', alpha=0.7)
axes[0].set_title('Distribuicao Uniforme')
axes[0].set_xlabel('Valor')
axes[0].set_ylabel('Frequencia')

# Distribuicao Exponencial
exponenciais = np.random.exponential(2, 5000)
axes[1].hist(exponenciais, bins=40, color='mediumseagreen', edgecolor='black', alpha=0.7)
axes[1].set_title('Distribuicao Exponencial')
axes[1].set_xlabel('Valor')

# Distribuicao Normal (Bimodal)
bimodal = np.concatenate([np.random.normal(0, 1, 2500), np.random.normal(4, 1, 2500)])
axes[2].hist(bimodal, bins=40, color='mediumpurple', edgecolor='black', alpha=0.7)
axes[2].set_title('Distribuicao Bimodal')
axes[2].set_xlabel('Valor')

plt.tight_layout()
plt.show()
````

````python
# Exemplo 7: Regra Empirica 68-95-99.7
np.random.seed(42)
dados = np.random.normal(100, 15, 10000)

media = np.mean(dados)
desvio = np.std(dados)

dentro_1 = np.sum((dados >= media - desvio) & (dados <= media + desvio))
dentro_2 = np.sum((dados >= media - 2*desvio) & (dados <= media + 2*desvio))
dentro_3 = np.sum((dados >= media - 3*desvio) & (dados <= media + 3*desvio))

print(f"Media = {media:.2f}, Desvio Padrao = {desvio:.2f}")
print(f"Dentro de 1 desvio (±{desvio:.0f}): {dentro_1/len(dados)*100:.1f}% (esperado: ~68%)")
print(f"Dentro de 2 desvios (±{2*desvio:.0f}): {dentro_2/len(dados)*100:.1f}% (esperado: ~95%)")
print(f"Dentro de 3 desvios (±{3*desvio:.0f}): {dentro_3/len(dados)*100:.1f}% (esperado: ~99.7%)")
````

````python
# Exemplo 8: Resumo estatistico completo (describe)
import pandas as pd

np.random.seed(42)
dados_complexos = {
    'Idade': np.random.normal(30, 10, 200).astype(int),
    'Salario': np.random.normal(5000, 2000, 200).astype(int),
    'Horas_Trabalho': np.random.normal(40, 5, 200).astype(int),
    'Score': np.random.uniform(0, 100, 200).round(1)
}

df = pd.DataFrame(dados_complexos)
print("Resumo Estatistico:\n")
print(df.describe())

print(f"\nCorrelacao Salario x Idade: {df['Salario'].corr(df['Idade']):.3f}")
print(f"Correlacao Salario x Score: {df['Salario'].corr(df['Score']):.3f}")
````

## ✅ Exercicios Resolvidos

````python
# Exercicio 1: Analise estatistica de um dataset
np.random.seed(42)
alturas = np.random.normal(170, 8, 500)

print(f"Media de altura: {np.mean(alturas):.1f} cm")
print(f"Mediana de altura: {np.median(alturas):.1f} cm")
print(f"Desvio Padrao: {np.std(alturas):.2f} cm")
print(f"Altura minima: {np.min(alturas):.1f} cm")
print(f"Altura maxima: {np.max(alturas):.1f} cm")
print(f"Q1: {np.percentile(alturas, 25):.1f} cm")
print(f"Q3: {np.percentile(alturas, 75):.1f} cm")
print(f"Pessoas acima de 180cm: {np.sum(alturas > 180)} ({np.mean(alturas > 180)*100:.1f}%)")
````

````python
# Exercicio 2: Verificando se dados sao normais com histograma e QQ-plot
from scipy import stats as sp_stats

np.random.seed(42)

# Dados normalmente distribuidos
dados_normais = np.random.normal(0, 1, 500)

# Dados nao normais (exponencial)
dados_nao_normais = np.random.exponential(1, 500)

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

axes[0, 0].hist(dados_normais, bins=30, color='steelblue', edgecolor='black', alpha=0.7)
axes[0, 0].set_title('Histograma - Dados Normais')
sp_stats.probplot(dados_normais, dist="norm", plot=axes[0, 1])
axes[0, 1].set_title('QQ-Plot - Dados Normais')

axes[1, 0].hist(dados_nao_normais, bins=30, color='coral', edgecolor='black', alpha=0.7)
axes[1, 0].set_title('Histograma - Dados Nao-Normais')
sp_stats.probplot(dados_nao_normais, dist="norm", plot=axes[1, 1])
axes[1, 1].set_title('QQ-Plot - Dados Nao-Normais')

plt.tight_layout()
plt.show()
````

## 📌 Resumo

- **Media** (`np.mean()`): soma dos valores / quantidade — sensivel a outliers
- **Mediana** (`np.median()`): valor central — robusta a outliers
- **Moda** (`stats.mode()`): valor mais frequente
- **Desvio Padrao** (`np.std()`): medida de dispersao em torno da media
- **Variancia** (`np.var()`): quadrado do desvio padrao
- **Percentil** (`np.percentile()`): valor abaixo do qual X% dos dados estao
- **IQR**: Q3 - Q1 (intervalo interquartil)
- **Distribuicao Normal**: curva de sino, regra 68-95-99.7
- ML supervisionado usa dados rotulados; nao supervisionado encontra padroes

## 💡 Curiosidades

- O termo "Machine Learning" foi cunhado por **Arthur Samuel** em 1959 enquanto trabalhava na IBM.
- A distribuicao normal e chamada de "Gaussiana" em homenagem a **Carl Friedrich Gauss**.
- O **Teorema Central do Limite** afirma que a media de muitas variaveis independentes tende a uma distribuicao normal, independentemente da distribuicao original. Isso e a base de muitos metodos estatisticos.
- Outliers podem distorcer completamente a media. Em 2013, a cidade de Bell (California) descobriu que seu administrador ganhava $787,637/ano — um outlier que aumentava artificialmente a media salarial.
- Francis Galton, primo de Darwin, foi pioneiro no uso de percentis e no conceito de regressao a media.

## ✅ Boas Praticas

- Sempre explore os dados com estatisticas descritivas antes de aplicar ML
- Use `df.describe()` do Pandas para um resumo rapido
- Visualize distribuicoes com histogramas e boxplots
- Identifique e trate outliers antes de treinar modelos
- Verifique a escala das variaveis — ML e sensivel a escalas diferentes
- Calcule a matriz de correlacao para identificar variaveis redundantes

## ⚠️ Erros Comuns

| Erro | Exemplo | Correcao |
|------|---------|----------|
| Usar media com outliers | Media salarial distorcida por CEO | Usar mediana ou remover outliers |
| Ignorar distribuicao dos dados | Aplicar modelos que assumem normalidade | Verificar distribuicao primeiro |
| Confundir desvio padrao com variancia | Interpretar variancia como dispersao na unidade original | Desvio padrao esta na mesma unidade dos dados |
| Nao normalizar dados | Variaveis em escalas diferentes dominam o modelo | Usar StandardScaler ou MinMaxScaler |
| Bins inadequados no histograma | Nao revela a forma real da distribuicao | Testar varios valores de bins |
