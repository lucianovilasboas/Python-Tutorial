## 📝 Introducao

**Algoritmos de busca** encontram um elemento em uma colecao. **Algoritmos de ordenacao** organizam elementos em uma ordem especifica.

### Resumo de Complexidades

| Algoritmo | Melhor Caso | Medio | Pior Caso | Espaco |
|-----------|------------|-------|-----------|--------|
| Linear Search | O(1) | O(n) | O(n) | O(1) |
| Binary Search | O(1) | O(log n) | O(log n) | O(1) |
| Bubble Sort | O(n) | O(n^2) | O(n^2) | O(1) |
| Selection Sort | O(n^2) | O(n^2) | O(n^2) | O(1) |
| Insertion Sort | O(n) | O(n^2) | O(n^2) | O(1) |
| Quick Sort | O(n log n) | O(n log n) | O(n^2) | O(log n) |
| Merge Sort | O(n log n) | O(n log n) | O(n log n) | O(n) |
| Counting Sort | O(n+k) | O(n+k) | O(n+k) | O(k) |
| Radix Sort | O(nk) | O(nk) | O(nk) | O(n+k) |

## 💻 Exemplos

```python
# Exemplo 1: Busca Linear
def busca_linear(arr, alvo):
    for i, valor in enumerate(arr):
        if valor == alvo:
            return i
    return -1

arr = [5, 3, 8, 1, 9, 2, 7]
print(f"Array: {arr}")
print(f"Buscar 9: indice {busca_linear(arr, 9)}")
print(f"Buscar 4: indice {busca_linear(arr, 4)}")
```

```python
# Exemplo 2: Busca Binaria (requer array ordenado)
def busca_binaria(arr, alvo):
    esquerda, direita = 0, len(arr) - 1
    while esquerda <= direita:
        meio = (esquerda + direita) // 2
        if arr[meio] == alvo:
            return meio
        elif arr[meio] < alvo:
            esquerda = meio + 1
        else:
            direita = meio - 1
    return -1

arr_ord = sorted(arr)
print(f"Array ordenado: {arr_ord}")
print(f"Buscar 7: indice {busca_binaria(arr_ord, 7)}")
print(f"Buscar 4: indice {busca_binaria(arr_ord, 4)}")
```

```python
# Exemplo 3: Algoritmos O(n^2) - Bubble, Selection, Insertion
import numpy as np
import time
import matplotlib.pyplot as plt

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

def insertion_sort(arr):
    for i in range(1, len(arr)):
        chave = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > chave:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = chave
    return arr

original = [64, 34, 25, 12, 22, 11, 90, 45, 33, 77]
print(f"Original:          {original}")
print(f"Bubble Sort:       {bubble_sort(original.copy())}")
print(f"Selection Sort:    {selection_sort(original.copy())}")
print(f"Insertion Sort:    {insertion_sort(original.copy())}")
```

```python
# Exemplo 4: Quick Sort
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivo = arr[len(arr) // 2]
    menores = [x for x in arr if x < pivo]
    iguais = [x for x in arr if x == pivo]
    maiores = [x for x in arr if x > pivo]
    return quick_sort(menores) + iguais + quick_sort(maiores)

arr_teste = [38, 27, 43, 3, 9, 82, 10]
print(f"Original:    {arr_teste}")
print(f"Quick Sort:  {quick_sort(arr_teste)}")
```

```python
# Exemplo 5: Merge Sort
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    meio = len(arr) // 2
    esquerda = merge_sort(arr[:meio])
    direita = merge_sort(arr[meio:])

    return _merge(esquerda, direita)

def _merge(esquerda, direita):
    resultado = []
    i = j = 0
    while i < len(esquerda) and j < len(direita):
        if esquerda[i] <= direita[j]:
            resultado.append(esquerda[i])
            i += 1
        else:
            resultado.append(direita[j])
            j += 1
    resultado.extend(esquerda[i:])
    resultado.extend(direita[j:])
    return resultado

print(f"Original:   {arr_teste}")
print(f"Merge Sort: {merge_sort(arr_teste)}")
```

```python
# Exemplo 6: Counting Sort (valores inteiros em range conhecido)
def counting_sort(arr):
    if not arr:
        return arr
    max_val = max(arr)
    min_val = min(arr)
    range_val = max_val - min_val + 1

    count = [0] * range_val
    output = [0] * len(arr)

    for num in arr:
        count[num - min_val] += 1

    for i in range(1, len(count)):
        count[i] += count[i - 1]

    for num in reversed(arr):
        output[count[num - min_val] - 1] = num
        count[num - min_val] -= 1

    return output

arr_cont = [4, 2, 2, 8, 3, 3, 1, 7, 5, 6]
print(f"Original:       {arr_cont}")
print(f"Counting Sort:  {counting_sort(arr_cont)}")
```

```python
# Exemplo 7: Radix Sort
def counting_sort_for_radix(arr, exp):
    n = len(arr)
    output = [0] * n
    count = [0] * 10

    for i in range(n):
        index = (arr[i] // exp) % 10
        count[index] += 1

    for i in range(1, 10):
        count[i] += count[i - 1]

    for i in range(n - 1, -1, -1):
        index = (arr[i] // exp) % 10
        output[count[index] - 1] = arr[i]
        count[index] -= 1

    for i in range(n):
        arr[i] = output[i]

def radix_sort(arr):
    if not arr:
        return arr
    max_val = max(arr)
    exp = 1
    while max_val // exp > 0:
        counting_sort_for_radix(arr, exp)
        exp *= 10
    return arr

arr_radix = [170, 45, 75, 90, 802, 24, 2, 66]
print(f"Original:   {arr_radix}")
print(f"Radix Sort: {radix_sort(arr_radix.copy())}")
```

```python
# Exemplo 8: Benchmark comparativo de algoritmos de ordenacao
np.random.seed(42)
tamanhos = [100, 500, 1000, 2000]
algoritmos = {
    'Bubble Sort': bubble_sort,
    'Selection Sort': selection_sort,
    'Insertion Sort': insertion_sort,
    'Quick Sort': quick_sort,
    'Merge Sort': merge_sort,
    'Python sorted()': sorted,
}

resultados = {nome: [] for nome in algoritmos}

for n in tamanhos:
    dados = np.random.randint(0, 10000, n).tolist()
    for nome, func in algoritmos.items():
        inicio = time.perf_counter()
        func(dados.copy())
        fim = time.perf_counter()
        resultados[nome].append((fim - inicio) * 1000)

print(f"{'Algoritmo':<20s}", end='')
for t in tamanhos:
    print(f"{t:>8d}", end='')
print("\n" + "-" * 52)

for nome, tempos in resultados.items():
    print(f"{nome:<20s}", end='')
    for t in tempos:
        print(f"{t:>7.2f}ms", end='')
    print()

print(f"\nNota: O(n^2) cresce muito mais rapido que O(n log n)!")
```

## ✅ Exercicios Resolvidos

```python
# Exercicio 1: Busca binaria recursiva
def busca_binaria_rec(arr, alvo, esq=0, dir=None):
    if dir is None:
        dir = len(arr) - 1
    if esq > dir:
        return -1
    meio = (esq + dir) // 2
    if arr[meio] == alvo:
        return meio
    if arr[meio] < alvo:
        return busca_binaria_rec(arr, alvo, meio + 1, dir)
    return busca_binaria_rec(arr, alvo, esq, meio - 1)

print(f"Busca binaria recursiva de 5: indice {busca_binaria_rec(arr_ord, 5)}")
```

```python
# Exercicio 2: Ordenar por chave especifica (lista de tuplas)
alunos = [("Joao", 8.5), ("Maria", 9.2), ("Pedro", 7.0), ("Ana", 9.5)]

# Quick sort adaptado para ordenar por nota (segundo elemento)
def quick_sort_tuplas(arr):
    if len(arr) <= 1:
        return arr
    pivo = arr[len(arr) // 2][1]
    menores = [x for x in arr if x[1] < pivo]
    iguais = [x for x in arr if x[1] == pivo]
    maiores = [x for x in arr if x[1] > pivo]
    return quick_sort_tuplas(menores) + iguais + quick_sort_tuplas(maiores)

print("Alunos ordenados por nota (Quick Sort adaptado):")
for nome, nota in quick_sort_tuplas(alunos):
    print(f"  {nome}: {nota}")

print(f"\nUsando sorted() nativo (mais rapido):")
for nome, nota in sorted(alunos, key=lambda x: x[1]):
    print(f"  {nome}: {nota}")
```

## 📌 Resumo

- **Busca Linear**: O(n) — percorre todos os elementos. Funciona em qualquer array.
- **Busca Binaria**: O(log n) — requer array ordenado. Muito mais rapida para grandes volumes.
- **Bubble Sort**: O(n^2) — simples, didatico, ineficiente. Compara pares adjacentes.
- **Selection Sort**: O(n^2) — encontra o menor e coloca na posicao correta.
- **Insertion Sort**: O(n^2) — eficiente para arrays quase ordenados. Insere cada elemento na posicao correta.
- **Quick Sort**: O(n log n) medio — divide e conquista com pivo. Muito usado na pratica.
- **Merge Sort**: O(n log n) garantido — divide ao meio, ordena recursivamente, mescla.
- **Counting Sort**: O(n+k) — para inteiros em range pequeno. Nao usa comparacoes.
- **Radix Sort**: O(nk) — ordena por digitos. Para numeros ou strings de tamanho fixo.

## 💡 Curiosidades

- **Quick Sort** foi desenvolvido por **Tony Hoare** em 1959 e e ate hoje um dos algoritmos mais usados. O `sorted()` do Python usa **Timsort**, uma mistura de Merge Sort e Insertion Sort.
- **Bubble Sort** e famoso por ser ensinado em todo curso de programacao, mas quase nunca usado em producao.
- **Counting Sort** foi inventado por **Harold H. Seward** em 1954 e e a base do Radix Sort.
- O Python `sorted()` usa o algoritmo **Timsort**, criado por **Tim Peters** em 2002, que e hibrido (Merge Sort + Insertion Sort) e otimizado para dados do mundo real.
- **Radix Sort** era usado em maquinas de cartao perfurado (Hollerith) no seculo XIX para ordenar dados do censo americano.

## ✅ Boas Praticas

- Para arrays pequenos (n < 50), Insertion Sort pode ser mais rapido que Quick Sort
- Use `sorted()` ou `.sort()` do Python — sao implementacoes otimizadas em C
- Busca binaria so funciona em arrays ORDENADOS — se nao estiver, ordene primeiro
- Algoritmos estaveis (Merge, Insertion) mantem a ordem relativa de elementos iguais
- Para inteiros em range pequeno, Counting/Radix Sort sao imbatíveis
- Quick Sort no pior caso (O(n^2)) ocorre com pivo mal escolhido — use pivo aleatorio ou mediana de tres

## ⚠️ Erros Comuns

| Erro | Exemplo | Correcao |
|------|---------|----------|
| Busca binaria em array nao ordenado | Resultado imprevisivel | Ordenar antes de usar busca binaria |
| Bubble Sort sem flag de otimizacao | Sempre O(n^2) mesmo ja ordenado | Adicionar flag `trocou` para early exit |
| Quick Sort com pivo fixo no primeiro | Pior caso O(n^2) em array ordenado | Pivo aleatorio ou mediana de tres |
| Counting Sort com range enorme | Memoria explode | Usar outro algoritmo se max-min for grande |
| Merge Sort in-place | Tentar fazer sem array auxiliar | Merge Sort precisa de O(n) de memoria extra |
| Modificar array original sem querer | Funcao altera o array de entrada | Fazer copia: `arr.copy()` |
