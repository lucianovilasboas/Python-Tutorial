
### Principais Operacoes SQL

| Operacao | SQL | Descricao |
|----------|-----|-----------|
| Criar BD | `CREATE DATABASE nome` | Cria um novo banco de dados |
| Criar Tabela | `CREATE TABLE nome (...)` | Cria tabela com colunas |
| Inserir | `INSERT INTO tabela VALUES (...)` | Adiciona registros |
| Selecionar | `SELECT colunas FROM tabela` | Consulta registros |
| Filtrar | `WHERE condicao` | Filtra resultados |
| Ordenar | `ORDER BY coluna [ASC/DESC]` | Ordena resultados |
| Atualizar | `UPDATE tabela SET col=val WHERE...` | Modifica registros |
| Excluir | `DELETE FROM tabela WHERE...` | Remove registros |
| Limitar | `LIMIT n` ou `LIMIT offset, n` | Pagina resultados |
| Juntar | `JOIN tabela ON condicao` | Combina tabelas |
| Deletar Tabela | `DROP TABLE nome` | Remove tabela |

### Prevencao de SQL Injection

Sempre use **parametros preparados** (`%s`):

```python
# SEGURO:
cursor.execute("SELECT * FROM users WHERE name = %s", (nome,))

# INSEGURO (NUNCA faca isso!):
cursor.execute(f"SELECT * FROM users WHERE name = '{nome}'")
```

## 💻 Exemplos

````python
# Exemplo 1: Conectando e criando um banco de dados
import mysql.connector

# Conectar ao MySQL (sem especificar database ainda)
try:
    conexao = mysql.connector.connect(
        host="localhost",
        user="root",
        password="sua_senha"
    )
    cursor = conexao.cursor()
    cursor.execute("CREATE DATABASE IF NOT EXISTS escola_python")
    print("Banco de dados 'escola_python' criado/verificado com sucesso!")
    cursor.close()
    conexao.close()
except mysql.connector.Error as err:
    print(f"Erro de conexao: {err}")
    print("(Certifique-se de que o MySQL esta rodando e as credenciais estao corretas)")
````

````python
# Exemplo 2: Criando tabela
try:
    conexao = mysql.connector.connect(
        host="localhost",
        user="root",
        password="sua_senha",
        database="escola_python"
    )
    cursor = conexao.cursor()

    cursor.execute("""
        CREATE TABLE IF NOT EXISTS alunos (
            id INT AUTO_INCREMENT PRIMARY KEY,
            nome VARCHAR(100) NOT NULL,
            idade INT,
            curso VARCHAR(100),
            nota DECIMAL(4,2)
        )
    """)
    print("Tabela 'alunos' criada/verificada com sucesso!")

    # Mostrar estrutura
    cursor.execute("DESCRIBE alunos")
    print("\nEstrutura da tabela:")
    for coluna in cursor.fetchall():
        print(f"  {coluna[0]:<15s} {coluna[1]:<15s} {coluna[2]}")

    cursor.close()
    conexao.close()
except mysql.connector.Error as err:
    print(f"Erro: {err}")
````

````python
# Exemplo 3: INSERT - Inserindo dados
try:
    conexao = mysql.connector.connect(
        host="localhost", user="root", password="sua_senha",
        database="escola_python"
    )
    cursor = conexao.cursor()

    # Inserir um unico registro
    sql = "INSERT INTO alunos (nome, idade, curso, nota) VALUES (%s, %s, %s, %s)"
    valores = ("Maria Silva", 20, "Ciencia da Computacao", 8.75)
    cursor.execute(sql, valores)

    # Inserir multiplos registros
    varios_alunos = [
        ("Joao Santos", 22, "Sistemas de Informacao", 7.50),
        ("Ana Pereira", 19, "Engenharia de Software", 9.20),
        ("Pedro Costa", 25, "Ciencia da Computacao", 6.80),
        ("Lucia Oliveira", 21, "Sistemas de Informacao", 8.90),
    ]
    cursor.executemany(sql, varios_alunos)

    conexao.commit()
    print(f"{cursor.rowcount} registros inseridos com sucesso!")

    cursor.close()
    conexao.close()
except mysql.connector.Error as err:
    print(f"Erro: {err}")
````

````python
# Exemplo 4: SELECT e fetchall()
try:
    conexao = mysql.connector.connect(
        host="localhost", user="root", password="sua_senha",
        database="escola_python"
    )
    cursor = conexao.cursor()

    cursor.execute("SELECT * FROM alunos")
    resultados = cursor.fetchall()

    print(f"{'ID':<5s} {'Nome':<20s} {'Idade':<8s} {'Curso':<30s} {'Nota':<8s}")
    print("-" * 75)
    for linha in resultados:
        print(f"{linha[0]:<5d} {linha[1]:<20s} {linha[2]:<8d} {linha[3]:<30s} {linha[4]:<8.2f}")

    cursor.close()
    conexao.close()
except mysql.connector.Error as err:
    print(f"Erro: {err}")
````

````python
# Exemplo 5: SELECT com WHERE, ORDER BY, LIMIT
try:
    conexao = mysql.connector.connect(
        host="localhost", user="root", password="sua_senha",
        database="escola_python"
    )
    cursor = conexao.cursor()

    # WHERE - Filtrar por curso
    print("1. Alunos de Ciencia da Computacao:")
    cursor.execute("SELECT nome, nota FROM alunos WHERE curso = %s", ("Ciencia da Computacao",))
    for nome, nota in cursor.fetchall():
        print(f"   {nome}: {nota}")

    # ORDER BY - Notas em ordem decrescente
    print("\n2. Top 3 melhores notas:")
    cursor.execute("SELECT nome, nota FROM alunos ORDER BY nota DESC LIMIT 3")
    for nome, nota in cursor.fetchall():
        print(f"   {nome}: {nota}")

    # WHERE + operadores
    print("\n3. Alunos com nota >= 7.0 e idade > 20:")
    cursor.execute("SELECT nome, idade, nota FROM alunos WHERE nota >= %s AND idade > %s", (7.0, 20))
    for nome, idade, nota in cursor.fetchall():
        print(f"   {nome} (idade {idade}): {nota}")

    # LIMIT com offset (paginacao)
    print("\n4. Pagina 1 (2 registros por pagina):")
    cursor.execute("SELECT nome FROM alunos LIMIT 2 OFFSET 0")
    for (nome,) in cursor.fetchall():
        print(f"   {nome}")

    cursor.close()
    conexao.close()
except mysql.connector.Error as err:
    print(f"Erro: {err}")
````

````python
# Exemplo 6: UPDATE e DELETE
try:
    conexao = mysql.connector.connect(
        host="localhost", user="root", password="sua_senha",
        database="escola_python"
    )
    cursor = conexao.cursor()

    # UPDATE - Atualizar nota
    cursor.execute("UPDATE alunos SET nota = %s WHERE nome = %s", (9.50, "Maria Silva"))
    print(f"Nota atualizada: {cursor.rowcount} registro(s) afetado(s)")

    # DELETE - Remover registro
    cursor.execute("DELETE FROM alunos WHERE nome = %s", ("Pedro Costa",))
    print(f"Registro removido: {cursor.rowcount} registro(s) afetado(s)")

    conexao.commit()
    cursor.close()
    conexao.close()
except mysql.connector.Error as err:
    print(f"Erro: {err}")
````

````python
# Exemplo 7: JOIN entre tabelas
try:
    conexao = mysql.connector.connect(
        host="localhost", user="root", password="sua_senha",
        database="escola_python"
    )
    cursor = conexao.cursor()

    # Criar tabela de cursos
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS cursos (
            id INT AUTO_INCREMENT PRIMARY KEY,
            nome VARCHAR(100),
            carga_horaria INT,
            professor VARCHAR(100)
        )
    """)

    cursor.executemany(
        "INSERT IGNORE INTO cursos (nome, carga_horaria, professor) VALUES (%s, %s, %s)",
        [
            ("Ciencia da Computacao", 3600, "Dr. Carlos Mendes"),
            ("Sistemas de Informacao", 3200, "Dra. Patricia Lima"),
            ("Engenharia de Software", 3400, "Dr. Roberto Alves"),
        ]
    )
    conexao.commit()

    # JOIN: alunos com detalhes do curso
    cursor.execute("""
        SELECT a.nome, a.nota, c.carga_horaria, c.professor
        FROM alunos a
        JOIN cursos c ON a.curso = c.nome
        ORDER BY a.nota DESC
    """)

    print("Alunos com detalhes do curso (JOIN):")
    print(f"{'Nome':<20s} {'Nota':<8s} {'Carga Hor.':<12s} {'Professor':<25s}")
    print("-" * 65)
    for nome, nota, carga, prof in cursor.fetchall():
        print(f"{nome:<20s} {nota:<8.2f} {carga:<12d} {prof:<25s}")

    cursor.close()
    conexao.close()
except mysql.connector.Error as err:
    print(f"Erro: {err}")
````

````python
# Exercicio 1: Inserir e consultar com seguranca
def inserir_aluno(nome, idade, curso, nota):
    try:
        conexao = mysql.connector.connect(
            host="localhost", user="root", password="sua_senha",
            database="escola_python"
        )
        cursor = conexao.cursor()
        sql = "INSERT INTO alunos (nome, idade, curso, nota) VALUES (%s, %s, %s, %s)"
        cursor.execute(sql, (nome, idade, curso, nota))
        conexao.commit()
        print(f"Aluno '{nome}' inserido com sucesso! ID: {cursor.lastrowid}")
        return cursor.lastrowid
    except mysql.connector.Error as err:
        print(f"Erro: {err}")
        return None
    finally:
        if 'conexao' in locals() and conexao.is_connected():
            cursor.close()
            conexao.close()

# Exemplo de uso (descomente para testar):
# inserir_aluno("Test User", 18, "Ciencia da Computacao", 8.0)
````

````python
# Exercicio 2: Relatorio de media por curso
try:
    conexao = mysql.connector.connect(
        host="localhost", user="root", password="sua_senha",
        database="escola_python"
    )
    cursor = conexao.cursor()

    cursor.execute("""
        SELECT curso,
               COUNT(*) AS total_alunos,
               ROUND(AVG(nota), 2) AS media_nota,
               MIN(nota) AS menor_nota,
               MAX(nota) AS maior_nota
        FROM alunos
        GROUP BY curso
        ORDER BY media_nota DESC
    """)

    print("Relatorio de Notas por Curso:")
    print(f"{'Curso':<30s} {'Alunos':<10s} {'Media':<10s} {'Min':<8s} {'Max':<8s}")
    print("-" * 65)
    for curso, total, media, min_n, max_n in cursor.fetchall():
        print(f"{curso:<30s} {total:<10d} {media:<10.2f} {min_n:<8.2f} {max_n:<8.2f}")

    cursor.close()
    conexao.close()
except mysql.connector.Error as err:
    print(f"Erro: {err}")
````

## 📌 Resumo

- **`mysql.connector`**: biblioteca oficial para conectar Python ao MySQL
- **`connect()`**: estabelece conexao com host, user, password, database
- **`cursor()`**: objeto para executar comandos SQL
- **`execute(sql, params)`**: executa SQL com parametros seguros (%s)
- **`executemany(sql, lista)`**: executa mesma query para varios registros
- **`commit()`**: confirma alteracoes no banco
- **`fetchall()` / `fetchone()`**: recupera resultados
- **CRUD**: Create (INSERT), Read (SELECT), Update (UPDATE), Delete (DELETE)
- **Sempre use parametros preparados** para evitar SQL Injection

## 💡 Curiosidades

- MySQL foi criado por **Michael Widenius** e **David Axmark** em 1995. O "My" vem do nome da filha de Widenius, "My".
- SQL Injection e uma das vulnerabilidades mais antigas e perigosas da web. Em 2012, um hacker usou SQL Injection para roubar 450.000 senhas do Yahoo.
- Alem do `mysql-connector`, existem outras bibliotecas como `PyMySQL` (puro Python) e `SQLAlchemy` (ORM).
- O comando `TRUNCATE TABLE` limpa todos os dados mas mantem a estrutura — diferente de `DROP TABLE` que remove tudo.
- **Indices** em MySQL funcionam como o indice de um livro: aceleram buscas mas ocupam espaco e tornam insercoes mais lentas.

## ✅ Boas Praticas

- **Sempre use parametros preparados** (%s) — nunca concatene strings SQL
- Feche conexoes e cursores com `close()` ou use `with` (context manager)
- Use `try/except/finally` para tratar erros de conexao
- Limite colunas no SELECT (`SELECT nome, idade` em vez de `SELECT *`)
- Use indices para colunas frequentemente usadas em WHERE e JOIN
- Nao armazene senhas de banco em codigo — use variaveis de ambiente

## ⚠️ Erros Comuns

| Erro | Exemplo | Correcao |
|------|---------|----------|
| SQL Injection | `f"SELECT * FROM users WHERE name='{nome}'"` | `cursor.execute("SELECT * FROM users WHERE name=%s", (nome,))` |
| Esquecer commit() | Dados nao persistem apos INSERT/UPDATE/DELETE | Chamar `conexao.commit()` |
| Conexao nao fechada | Vazamento de conexoes, limite do MySQL | `conexao.close()` ou context manager |
| Nao tratar encoding | Caracteres acentuados quebram | Usar `charset='utf8mb4'` na conexao |
| Acesso negado (Access Denied) | Credenciais erradas ou host nao permitido | Verificar user/password e privilegios |
