## 📝 Introdução

Comentários são textos escritos no código que o interpretador Python **ignora completamente**. Eles existem exclusivamente para seres humanos — programadores que lerão o código no futuro (incluindo você mesmo semanas depois!).

Um bom comentário explica **por que** o código foi escrito de determinada forma, não **o que** o código faz (isso deve ser evidente pelo próprio código). Comentários são ferramentas de comunicação entre desenvolvedores.

## 📚 Explicação Detalhada

### 1. Comentários de Linha Única (`#`)

O símbolo `#` indica que tudo à sua frente naquela linha é um comentário. É o tipo mais comum de comentário em Python.

````python
# Isto é um comentário de linha única
print("Olá, mundo!")  # Comentário no final da linha

# Podemos "comentar" um bloco inteiro
# colocando # no início de cada linha
# Útil para desabilitar código temporariamente
# print("Esta linha está comentada")
````

### 2. Comentários Multilinha (`'''` e `"""`)

Python não tem um símbolo específico para comentários multilinha. No entanto, strings delimitadas por triplas aspas (`'''` ou `"""`) que não são atribuídas a nenhuma variável funcionam como comentários, pois são ignoradas pelo interpretador.

````python
'''
Este é um comentário multilinha.
Tudo aqui é ignorado pelo Python.
Útil para explicações longas.
'''

print("O comentário acima não interfere no código")

"""
Também podemos usar aspas duplas.
O resultado é o mesmo.
"""
````

### 3. Docstrings (Documentação de Código)

**Docstrings** são strings de documentação usadas para explicar o propósito de módulos, funções, classes e métodos. Diferente de comentários comuns, docstrings são **acessíveis em tempo de execução** através do atributo `__doc__` ou da função `help()`.

````python
# Docstring de uma função
def saudacao(nome):
    """Exibe uma saudação personalizada para o usuário.
    
    Args:
        nome (str): O nome da pessoa a ser saudada.
        
    Returns:
        str: A mensagem de saudação formatada.
    """
    return f"Olá, {nome}! Seja bem-vindo(a)!"


print(saudacao("Ana"))
print(saudacao.__doc__)  # Acessando a docstring
````

### 4. Docstring de Módulo

Colocamos uma docstring no início do arquivo para documentar o propósito do módulo inteiro.

````python
"""Módulo de operações matemáticas básicas.

Este módulo contém funções simples para realizar
operações aritméticas como soma, subtração,
multiplicação e divisão.
"""

def somar(a, b):
    """Retorna a soma de a e b."""
    return a + b

def subtrair(a, b):
    """Retorna a subtração de b de a."""
    return a - b

print(somar(10, 5))
print(subtrair(10, 5))
````

### 5. Quando Comentar (e Quando NÃO Comentar)

**Comente quando:**
- O código resolve um problema complexo e não óbvio
- Há uma decisão de design importante (por que escolheu A em vez de B)
- A lógica precisa de contexto que não está no código
- Você está documentando uma API pública (docstrings)

**NÃO comente quando:**
- O código é autoexplicativo (`x += 1  # incrementa x em 1` é inútil!)
- Você está apenas "enchendo linguiça" para parecer mais profissional
- O comentário vai ficar desatualizado quando o código mudar

````python
# ❌ Comentário ruim (redundante)
x = x + 1  # Adiciona 1 a x

# ✅ Comentário útil
taxa_desconto = 0.10  # 10% de desconto para pagamento à vista

# ✅ Comentário explicando decisão
# Usamos sorted() em vez de .sort() para preservar a lista original
lista_ordenada = sorted([3, 1, 4, 1, 5])
````

## 💻 Exemplos Práticos

````python
# Exemplo 1: Comentários em um programa de cálculo de frete
distancia_km = 150
peso_kg = 3.5

# Taxa base: R$ 5,00 para os primeiros 50 km
# Taxa adicional: R$ 0,50 por km excedente
# Taxa de peso: R$ 2,00 por kg

taxa_base = 5.00
taxa_adicional = max(0, distancia_km - 50) * 0.50
taxa_peso = peso_kg * 2.00

frete_total = taxa_base + taxa_adicional + taxa_peso
print(f"Frete total: R$ {frete_total:.2f}")
````

````python
# Exemplo 2: Docstring completa seguindo padrão Google
def calcular_imc(peso, altura):
    """Calcula o Índice de Massa Corporal (IMC).

    O IMC é calculado dividindo o peso (kg) pelo
    quadrado da altura (m).

    Args:
        peso (float): Peso em quilogramas.
        altura (float): Altura em metros.

    Returns:
        float: O valor do IMC calculado.

    Example:
        >>> calcular_imc(70, 1.75)
        22.857142857142858
    """
    return peso / (altura ** 2)


imc = calcular_imc(70, 1.75)
print(f"IMC: {imc:.2f}")
print(f"Classificação: ", end="")
if imc < 18.5:
    print("Abaixo do peso")
elif imc < 25:
    print("Peso normal")
elif imc < 30:
    print("Sobrepeso")
else:
    print("Obesidade")
````

**Exercício 2:** Crie uma função com docstring completa que converta temperaturas de Celsius para Fahrenheit.

````python
def celsius_para_fahrenheit(celsius):
    """Converte uma temperatura de Celsius para Fahrenheit.
    
    A fórmula de conversão é: F = (C * 9/5) + 32
    
    Args:
        celsius (float): Temperatura em graus Celsius.
        
    Returns:
        float: Temperatura em graus Fahrenheit.
        
    Example:
        >>> celsius_para_fahrenheit(0)
        32.0
        >>> celsius_para_fahrenheit(100)
        212.0
    """
    return (celsius * 9 / 5) + 32


print(f"0°C = {celsius_para_fahrenheit(0)}°F")
print(f"100°C = {celsius_para_fahrenheit(100)}°F")
````

````python
# 🟡 Exercício 6 (Médio): Função area_circulo com docstring
def area_circulo(raio):
    """Calcula a área de um círculo.
    
    A fórmula é: A = π * r²
    
    Args:
        raio (float): O raio do círculo.
        
    Returns:
        float: A área calculada.
        
    Example:
        >>> area_circulo(5)
        78.53975
    """
    pi = 3.14159
    return pi * raio ** 2


print(f"Área: {area_circulo(5):.2f}")
````

3. **Desafio do Refactoring:** Pegue o código abaixo, renomeie as variáveis e adicione comentários significativos:
   ```python
   a = 5

   d = ((a + b) * c) / (a - b)
   ```

````python
# 🏆 Desafio 2: Calculadora Científica Documentada
"""Módulo de calculadora científica com funções básicas."""


def somar(a, b):
    """Retorna a soma de dois números.
    
    Args:
        a (float): Primeiro número.
        b (float): Segundo número.
        
    Returns:
        float: Resultado de a + b.
    """
    return a + b


def subtrair(a, b):
    """Retorna a subtração de dois números.
    
    Args:
        a (float): Primeiro número.
        b (float): Segundo número.
        
    Returns:
        float: Resultado de a - b.
    """
    return a - b


def multiplicar(a, b):
    """Retorna a multiplicação de dois números."""
    return a * b


def dividir(a, b):
    """Retorna a divisão de dois números.
    
    Raises:
        ValueError: Se b for zero.
    """
    if b == 0:
        raise ValueError("Divisão por zero não permitida")
    return a / b


# Exibindo documentação de cada função
funcoes = [somar, subtrair, multiplicar, dividir]
for func in funcoes:
    print(f"{func.__name__}: {func.__doc__}\n")
````

## 📌 Resumo

- **Comentários de linha única**: Use `#` para comentar uma linha ou parte dela.
- **Comentários multilinha**: Use `'''` ou `""""` para blocos de texto que serão ignorados.
- **Docstrings**: Use `"""` dentro de funções, classes ou módulos para documentação oficial — acessível via `help()` ou `__doc__`.
- **Comentários explicam o PORQUÊ**, não o O QUÊ — o código já mostra o que faz.
- **Mantenha comentários atualizados**: Um comentário desatualizado é pior que nenhum.
- **Evite comentários óbvios**: `x += 1  # incrementa x` é ruído, não informação.
- Use marcadores como `TODO:`, `FIXME:`, `NOTE:` para anotações importantes.

## 💡 Curiosidades

- O símbolo `#` é chamado de **octothorpe** ou **hash** em inglês. No Brasil, é conhecido como **jogo da velha** ou **cerquilha**.

- Python é uma das poucas linguagens que tem docstrings como parte da especificação da linguagem. O PEP 257 define como devem ser usadas.

- O famoso comentário `# This is the answer to life, the universe, and everything` aparece em muitos códigos Python com o número 42.

- Guido van Rossum, criador do Python, disse: "Código é lido muito mais vezes do que escrito" — por isso a importância de comentários claros!

- Existem ferramentas como **Sphinx** e **pydoc** que geram documentação automática a partir das docstrings do seu código.

## ✅ Boas Práticas

1. **Comente o porquê, não o como** — o código já mostra como.
2. **Use frases completas** para comentários que explicam blocos de código.
3. **Mantenha comentários alinhados** com o código que eles descrevem.
4. **Atualize comentários** quando o código mudar.
5. **Use docstrings para APIs públicas** — funções que outros programadores vão usar.
6. **Evite comentários óbvios** — eles poluem o código.
7. **Use um espaço após `#`** — `# Comentário` e não `#Comentário`.
8. **Comentários não substituem código limpo** — prefira nomes de variáveis significativos.

## ⚠️ Erros Comuns

| Erro | Exemplo | Como Evitar |
|------|---------|-------------|
| Comentário redundante | `x = x + 1  # incrementa x` | Comente apenas o que não é óbvio |
| Comentário desatualizado | Código mudou mas comentário não | Sempre revise comentários ao alterar código |
| Falta de docstring em funções | Função sem documentação | Sempre adicione docstring em funções públicas |
| Comentário mentiroso | Comentário que não reflete o código | Teste seus comentários como testa seu código |
| Comentários em excesso | Cada linha tem um comentário | Confie em código limpo, não em comentários |
| Comentário mal formatado | Sem espaços, sem pontuação | Siga as convenções de formatação |
| Usar comentário onde deveria usar docstring | `# Função que soma` vs docstring | Use `"""` dentro da função para documentação |
