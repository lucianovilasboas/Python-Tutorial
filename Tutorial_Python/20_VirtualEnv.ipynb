## 📝 Introdução

Quando você começa a instalar pacotes Python com `pip`, eles vão para o diretório global do sistema. Isso pode causar conflitos entre projetos que precisam de versões diferentes da mesma biblioteca. Os **ambientes virtuais** resolvem esse problema criando espaços isolados onde cada projeto tem suas próprias dependências. O módulo `venv` é a ferramenta oficial do Python para criar esses ambientes. Vamos aprender a criar, ativar, usar e compartilhar ambientes virtuais!

## 📚 Explicação Detalhada

### O que é um Ambiente Virtual?

Um ambiente virtual é uma cópia isolada do interpretador Python com seu próprio diretório de pacotes (`site-packages`). Cada projeto pode ter seu próprio ambiente, com versões específicas de bibliotecas.

### Principais comandos

| Comando | Descrição |
|---------|-----------|
| `python -m venv venv` | Cria o ambiente virtual na pasta `venv/` |
| `venv\Scripts\activate` (Windows) | Ativa o ambiente |
| `source venv/bin/activate` (Linux/Mac) | Ativa o ambiente |
| `deactivate` | Desativa o ambiente |
| `pip install pacote` | Instala pacote no ambiente ativo |
| `pip freeze` | Lista pacotes instalados |
| `pip freeze > requirements.txt` | Exporta dependências |
| `pip install -r requirements.txt` | Instala dependências de um arquivo |

### requirements.txt

Arquivo de texto que lista todas as dependências do projeto. Exemplo:
```
requests==2.31.0
numpy>=1.24.0
flask
```

## 💻 Exemplos

### Exemplo 1: Criando um ambiente virtual

````python
# No terminal, execute:
# python -m venv meu_ambiente

# Isso cria a pasta meu_ambiente/ com:
# - meu_ambiente/Scripts/  (Windows)
# - meu_ambiente/bin/      (Linux/Mac)
# - meu_ambiente/Lib/
# - meu_ambiente/include/

print("Ambiente virtual criado com sucesso!")
print("Execute o comando no terminal, não aqui no Jupyter.")
````

### Exemplo 2: Ativando o ambiente

````python
# Windows (Prompt / PowerShell):
# meu_ambiente\Scripts\activate

# Linux/Mac:
# source meu_ambiente/bin/activate

# Após ativar, você verá (meu_ambiente) no início do prompt.
print("Ambiente ativado! Agora os pacotes serão instalados aqui.")
````

### Exemplo 3: Instalando pacotes no ambiente

````python
# Com o ambiente ativo, use pip normalmente:
# pip install requests
# pip install numpy pandas matplotlib

# Para versão específica:
# pip install flask==2.3.3

print("Pacotes instalados no ambiente virtual.")
````

### Exemplo 4: Verificando pacotes instalados com pip freeze

````python
# pip freeze lista todos os pacotes com suas versões
# Comando no terminal:
# pip freeze

# Exemplo de saída:
# certifi==2024.2.2
# charset-normalizer==3.3.2
# idna==3.6
# requests==2.31.0
# urllib3==2.2.1

print("Use 'pip freeze' no terminal para ver os pacotes.")
````

### Exemplo 5: Exportando dependências para requirements.txt

````python
# No terminal, com ambiente ativo:
# pip freeze > requirements.txt

# Isso cria um arquivo requirements.txt com todas as dependências.
# Conteúdo típico:
# requests==2.31.0
# numpy==1.26.4
# pandas==2.2.0

print("Arquivo requirements.txt gerado!")
print("Compartilhe esse arquivo com outros desenvolvedores.")
````

### Exemplo 6: Instalando dependências de um requirements.txt

````python
# Em um novo ambiente, instale todas as dependências:
# pip install -r requirements.txt

# O pip lerá o arquivo e instalará cada pacote na versão especificada.
print("Dependências instaladas a partir do requirements.txt!")
````

### Exemplo 7: Desativando o ambiente

````python
# Para sair do ambiente virtual:
# deactivate

# O prompt volta ao normal (sem o prefixo do ambiente).
print("Ambiente desativado. Voltando ao Python global.")
````

### Exemplo 8: Removendo um ambiente virtual

````python
# Para remover completamente, basta deletar a pasta:
# Windows (PowerShell):
# Remove-Item -Recurse -Force meu_ambiente

# Linux/Mac:
# rm -rf meu_ambiente

print("Ambiente virtual removido com sucesso!")
````

### Exercício 3: Verificar diferença entre ambiente global e virtual

````python
import sys
print("Python executável:", sys.executable)
print("Caminho dos pacotes:")
for p in sys.path:
    print(f"  - {p}")

# Com o ambiente ativo, sys.executable aponta para dentro da pasta venv
# Sem ambiente, aponta para o Python global do sistema
````

## 📌 Resumo

- Ambientes virtuais isolam dependências entre projetos
- `python -m venv nome_do_ambiente` cria o ambiente
- Ative com `nome\Scripts\activate` (Windows) ou `source nome/bin/activate` (Linux/Mac)
- `pip freeze` lista pacotes; `pip freeze > requirements.txt` exporta
- `pip install -r requirements.txt` instala dependências de um arquivo
- `deactivate` sai do ambiente
- Inclua `venv/` no `.gitignore` — nunca versionar a pasta do ambiente

## 💡 Curiosidades

- O módulo `venv` substituiu o antigo `virtualenv` (que era um pacote de terceiros) a partir do Python 3.3
- Existe também o `pipenv` e o `poetry` que combinam gerenciamento de dependências com ambientes virtuais
- O `conda` (do Anaconda) também cria ambientes, mas funciona de forma diferente do `venv`
- No VS Code, você pode selecionar o interpretador do ambiente virtual com `Ctrl+Shift+P` → "Python: Select Interpreter"

## ✅ Boas Práticas

- Sempre crie um ambiente virtual para cada projeto Python
- Nomeie a pasta como `venv` ou `.venv` por convenção
- Adicione `venv/` ao `.gitignore` para não versionar o ambiente
- Mantenha o `requirements.txt` atualizado
- Use `python -m venv` em vez de `virtualenv` (já vem com o Python)
- Documente no README como configurar o ambiente

## ⚠️ Erros Comuns

| Erro | Causa | Solução |
|------|-------|---------|
| `'pip' não é reconhecido` | Python não está no PATH ou ambiente não ativado | Ative o ambiente ou reinstale o Python com a opção "Add to PATH" |
| `Error: [WinError 5] Acesso negado` | Tentar instalar pacote sem permissão no Python global | Use ambiente virtual (não precisa de admin) |
| `pip freeze` vazio | Ambiente não ativado | Ative o ambiente primeiro |
| Esquecer de ativar o ambiente | Pacotes instalados globalmente em vez de no ambiente | Verifique se o prompt mostra `(venv)` |
| `ModuleNotFoundError` após ativar | Pacote instalado no ambiente errado | Verifique se o ambiente ativo é o correto com `sys.executable` |
| Versionar a pasta `venv/` | Pasta grande (centenas de MBs) desnecessária | Adicione `venv/` ao `.gitignore` e use `requirements.txt` |
