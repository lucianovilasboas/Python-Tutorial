## 📝 Introdução

Booleanos são a base da lógica computacional. Em Python, `True` e `False` são objetos do tipo `bool` que controlam o fluxo do programa. Os operadores permitem combinar e comparar valores para tomar decisões.

## 📚 Explicação Detalhada

### Booleanos
- `True` e `False` (sempre com iniciais maiúsculas)
- `bool(0)`, `bool("")`, `bool([])`, `bool(None)` → `False`
- Qualquer outro valor → `True`

### Operadores

| Categoria | Operadores |
|-----------|-----------|
| Aritméticos | `+`, `-`, `*`, `/`, `//`, `%`, `**` |
| Atribuição | `=`, `+=`, `-=`, `*=`, `/=` … |
| Comparação | `==`, `!=`, `>`, `<`, `>=`, `<=` |
| Lógicos | `and`, `or`, `not` |
| Identidade | `is`, `is not` |
| Associação | `in`, `not in` |
| Bitwise | `&`, `|`, `^`, `~`, `<<`, `>>` |

### Precedência (do maior para menor)
1. `()` parênteses
2. `**`
3. `~`, `+x`, `-x`
4. `*`, `/`, `//`, `%`
5. `+`, `-`
6. `<<`, `>>`
7. `&`
8. `^`
9. `|`
10. `==`, `!=`, `>`, `<`, `>=`, `<=`, `is`, `in`
11. `not`
12. `and`
13. `or`

## 💻 Exemplos

```python
# Exemplo 1: Booleanos básicos
print(True)
print(False)
print(type(True))
print(bool(1))
print(bool(0))
print(bool(""))
print(bool("Texto"))
```

```python
# Exemplo 2: Operadores aritméticos
a, b = 15, 4
print(f"Soma: {a + b}")
print(f"Subtração: {a - b}")
print(f"Multiplicação: {a * b}")
print(f"Divisão: {a / b}")
print(f"Divisão inteira: {a // b}")
print(f"Resto: {a % b}")
print(f"Potência: {a ** b}")
```

```python
# Exemplo 3: Operadores de atribuição compostos
x = 10
x += 5   # x = x + 5
print(f"x += 5 → {x}")
x *= 2   # x = x * 2
print(f"x *= 2 → {x}")
x //= 3
print(f"x //= 3 → {x}")
```

```python
# Exemplo 4: Operadores de comparação
print(f"5 == 5 → {5 == 5}")
print(f"5 != 5 → {5 != 5}")
print(f"10 > 5 → {10 > 5}")
print(f"3 < 2 → {3 < 2}")
print(f"5 >= 5 → {5 >= 5}")
print(f"4 <= 3 → {4 <= 3}")
```

```python
# Exemplo 5: Operadores lógicos
idade = 25
tem_carteira = True
print(idade >= 18 and tem_carteira)  # True
print(idade >= 18 or tem_carteira)   # True
print(not tem_carteira)              # False

# Curto-circuito
print(False and print("Isso não executa"))  # False (sem executar o print)
```

```python
# Exemplo 6: Operador de identidade (is)
a = [1, 2, 3]
b = [1, 2, 3]
c = a
print(f"a is c → {a is c}")   # True (mesmo objeto)
print(f"a is b → {a is b}")   # False (objetos diferentes)
print(f"a == b → {a == b}")   # True (mesmo valor)
print(f"a is not b → {a is not b}")  # True
```

```python
# Exemplo 7: Operadores de associação (in / not in)
frase = "Python é incrível"
print(f"'Python' in frase → {'Python' in frase}")
print(f"'Java' in frase → {'Java' in frase}")
print(f"'Java' not in frase → {'Java' not in frase}")

lista = [10, 20, 30]
print(f"20 in lista → {20 in lista}")
print(f"50 in lista → {50 in lista}")
```

```python
# Exemplo 8: Operadores bitwise
x, y = 5, 3  # 5 = 0101, 3 = 0011
print(f"x & y (AND): {x & y}")   # 0101 & 0011 = 0001 = 1
print(f"x | y (OR):  {x | y}")   # 0101 | 0011 = 0111 = 7
print(f"x ^ y (XOR): {x ^ y}")   # 0101 ^ 0011 = 0110 = 6
print(f"~x (NOT):    {~x}")       # ~0101 = ...1010 = -6
print(f"x << 1:      {x << 1}")   # 0101 << 1 = 1010 = 10
print(f"x >> 1:      {x >> 1}")   # 0101 >> 1 = 0010 = 2
```

## ✅ Exercícios Resolvidos

```python
# Exercício Resolvido 1: Verificar se um ano é bissexto
ano = int(input("Digite um ano: "))
bissexto = (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0)
print(f"O ano {ano} é bissexto? {bissexto}")
```

```python
# Exercício Resolvido 2: Validação de senha
senha = input("Crie uma senha (mín. 8 caracteres): ")
tem_tamanho = len(senha) >= 8
tem_numero = any(c.isdigit() for c in senha)
tem_maiuscula = any(c.isupper() for c in senha)
valida = tem_tamanho and tem_numero and tem_maiuscula
print(f"Tamanho >= 8: {tem_tamanho}")
print(f"Contém número: {tem_numero}")
print(f"Contém maiúscula: {tem_maiuscula}")
print(f"Senha válida? {valida}")
```

```python
# Exercício Resolvido 3: Média ponderada com aprovação
n1 = float(input("Nota 1 (peso 2): "))
n2 = float(input("Nota 2 (peso 3): "))
n3 = float(input("Nota 3 (peso 5): "))
media = (n1 * 2 + n2 * 3 + n3 * 5) / 10
aprovado = media >= 6.0
print(f"Média: {media:.1f} — {'Aprovado' if aprovado else 'Reprovado'}")
```

## 📌 Resumo

- `True` e `False` são os valores booleanos (subtipos de `int`)
- Valores vazios/zero são `False`; os demais são `True`
- Operadores aritméticos: `+`, `-`, `*`, `/`, `//`, `%`, `**`
- Operadores de comparação retornam booleanos
- `and` / `or` / `not` para lógica combinacional
- `is` compara identidade (mesmo objeto); `==` compara valor
- `in` verifica associação em sequências
- Use parênteses para tornar a precedência explícita

## 💡 Curiosidades

- `True + True` retorna `2` — porque `bool` é subclasse de `int`.
- O operador `:=` (walrus operator) permite atribuir e avaliar em uma expressão.
- `None` é o único valor do tipo `NoneType` e é avaliado como `False` em contextos booleanos.

## ✅ Boas Práticas

- Não compare booleanos com `== True`; basta `if variavel:`
- Evite encadeamentos muito longos de `and`/`or`; quebre em variáveis
- Use `is None` para verificar `None`, não `== None`
- Parênteses são seus aliados para precedência clara

## ⚠️ Erros Comuns

| Erro | Exemplo | Solução |
|------|---------|--------|
| Usar `=` em vez de `==` | `if x = 5:` | Use `==` para comparação |
| Confundir `and`/`or` | `if idade >= 18 or tem_carteira:` | Entenda a tabela verdade |
| Comparar com `is` para valores | `x is 5` (funciona, mas não é ideal) | Use `==` para comparar valores |
| Precedência incorreta | `5 + 3 * 2` → 11 (não 16) | Use `(5 + 3) * 2` se desejar 16 |
