## 🎯 Objetivos

- Entender a regra LEGB (Local, Enclosing, Global, Built-in)
- Diferenciar escopo local de escopo global
- Usar `global` para modificar variáveis globais
- Usar `nonlocal` em funções aninhadas
- Compreender o tempo de vida (lifetime) das variáveis

## 📝 Introdução

Escopo (scope) determina onde uma variável pode ser acessada no código. Python segue a regra LEGB: Local → Enclosing (função externa) → Global → Built-in. Entender escopo é crucial para evitar bugs sutis e escrever código limpo e previsível.

## 📚 Explicação Detalhada

### LEGB — Ordem de busca do Python
1. **L**ocal: dentro da função atual
2. **E**nclosing: funções externas (closures)
3. **G**lobal: módulo (nível do arquivo)
4. **B**uilt-in: nomes embutidos (`print`, `len`, etc.)

### global
Declara que uma variável pertence ao escopo global, permitindo modificá-la dentro de uma função.

### nonlocal
Usado em funções aninhadas para modificar variáveis do escopo enclosing.

### Lifetime
Variáveis locais são criadas na chamada da função e destruídas ao final. Variáveis globais persistem durante toda a execução do módulo.

## 💻 Exemplos

```python
# Exemplo 1: Escopo local vs global
x = 10  # global

def funcao():
    x = 5  # local
    print(f"Dentro da função: x = {x}")

funcao()
print(f"Fora da função: x = {x}")
```

```python
# Exemplo 2: Modificando global com 'global'
contador = 0

def incrementar():
    global contador
    contador += 1

incrementar()
incrementar()
print(f"Contador final: {contador}")
```

```python
# Exemplo 3: Escopo enclosing com nonlocal
def externa():
    x = "externa"

    def interna():
        nonlocal x
        x = "modificada pela interna"

    interna()
    print(f"x na função externa: {x}")

externa()
```

```python
# Exemplo 4: Escopo built-in
print("A função 'print' está no escopo built-in")
print(f"len([1,2,3]) = {len([1,2,3])}")
```

```python
# Exemplo 5: Sombra (shadowing) — variável local esconde a global
nome = "Global"

def saudacao():
    nome = "Local"  # sombreia a variável global
    print(f"Dentro: {nome}")

saudacao()
print(f"Fora: {nome}")
```

```python
# Exemplo 6: Escopo em funções aninhadas (LEGB na prática)
z = "global"

def funcao_externa():
    y = "enclosing"

    def funcao_interna():
        x = "local"
        print(f"Local: {x}")
        print(f"Enclosing: {y}")
        print(f"Global: {z}")

    funcao_interna()

funcao_externa()
```

```python
# Exemplo 7: Variável global mutável (lista) pode ser modificada sem 'global'
lista = [1, 2, 3]

def adicionar(item):
    lista.append(item)  # modifica o objeto, não a referência

adicionar(4)
print(lista)
```

```python
# Exemplo 8: Lifetime de variável local
def criar_variavel():
    temp = 42
    return temp

resultado = criar_variavel()
print(f"Valor retornado: {resultado}")
# print(temp)  # NameError: variável 'temp' não existe fora da função
```

## ✅ Exercícios Resolvidos

```python
# Exercício Resolvido 1: Contador com global
acessos = 0

def registrar_acesso():
    global acessos
    acessos += 1
    print(f"Acesso registrado. Total: {acessos}")

registrar_acesso()
registrar_acesso()
registrar_acesso()
```

```python
# Exercício Resolvido 2: Closure com nonlocal
def multiplicador(fator):
    def multiplicar(numero):
        return numero * fator
    return multiplicar

dobrar = multiplicador(2)
triplicar = multiplicador(3)

print(f"Dobro de 5: {dobrar(5)}")
print(f"Triplo de 5: {triplicar(5)}")
```

```python
# Exercício Resolvido 3: Sombreamento e boas práticas
total = 100

def calcular_desconto(percentual):
    total = 0  # variável local (não afeta a global)
    total = 1000 - (1000 * percentual / 100)
    return total

print(f"Preço com desconto: R${calcular_desconto(10):.2f}")
print(f"Total global inalterado: {total}")
```

## 📌 Resumo

- Python busca variáveis na ordem LEGB: Local → Enclosing → Global → Built-in
- `global` permite modificar uma variável global dentro de uma função
- `nonlocal` permite modificar uma variável do escopo enclosing em funções aninhadas
- Variáveis locais são criadas na chamada e destruídas ao final da função
- Variáveis globais persistem durante toda a execução do módulo
- Objetos mutáveis (listas, dicts) podem ser modificados sem `global`, mas reatribuição requer `global`
- Shadowing ocorre quando uma variável local tem o mesmo nome de uma global

## 💡 Curiosidades

- Python não cria um novo escopo para blocos `if`, `for`, `while` (diferente de C/Java)
- O escopo built-in é implementado no módulo `builtins`
- Closures são amplamente usadas em decorators, callbacks e programação funcional
- Em Python, funções lembram o escopo onde foram definidas (lexical scoping)

## ✅ Boas Práticas

- Evite usar `global` sempre que possível — prefira parâmetros e retorno
- Use `global` para constantes ou contadores globais quando realmente necessário
- Prefira `nonlocal` a variáveis globais em closures
- Evite sombrear nomes built-in (não crie uma variável chamada `list` ou `print`)
- Mantenha funções puras (sem efeitos colaterais) sempre que possível

## ⚠️ Erros Comuns

| Erro | Descrição | Solução |
|------|-----------|---------|
| UnboundLocalError | Tentar ler e depois atribuir variável global | Usar `global` antes de usar |
| Esquecer `global` | Atribuição cria variável local em vez de modificar global | Declarar `global var` |
| Esquecer `nonlocal` | Tentar modificar variável enclosing | Usar `nonlocal var` |
| Shadowing acidental | Variável local com mesmo nome de função built-in | Renomear a variável |
| Modificar global sem necessidade | Efeitos colaterais difíceis de rastrear | Usar `return` em vez de `global` |
| Confundir escopo em blocos | Achar que `if`/`for` cria escopo | Lembrar que apenas funções e classes criam escopo |
