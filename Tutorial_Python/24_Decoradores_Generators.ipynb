## 📝 Introdução

**Decoradores** e **Generators** são dois recursos poderosos e elegantes de Python que permitem escrever código mais limpo, reutilizável e eficiente.

- **Decoradores** permitem modificar ou estender o comportamento de funções e métodos sem alterar seu código fonte.
- **Generators** permitem criar sequências de dados sob demanda, economizando memória e permitindo processamento lazy (preguiçoso).

## 📚 Explicação Detalhada

### Decoradores

Um **decorador** é uma função que recebe outra função como argumento e retorna uma nova função com comportamento estendido.

```python
@meu_decorador
def minha_funcao():
    pass
```

É equivalente a: `minha_funcao = meu_decorador(minha_funcao)`

### `functools.wraps`

Preserva o nome, docstring e outros metadados da função original na função decorada.

### Decoradores Nativos

- **`@staticmethod`**: método que não recebe `self` nem `cls` — funciona como função comum dentro da classe
- **`@classmethod`**: método que recebe `cls` (a classe) em vez de `self` — pode acessar/modificar atributos de classe
- **`@property`**: transforma um método em atributo somente-leitura (ou com getter/setter)

### Generators

Funções que usam `yield` em vez de `return`. Elas retornam um objeto **generator** que pode ser iterado.

- **`yield`**: pausa a execução, retorna um valor e salva o estado
- **`generator.send(valor)`**: envia um valor para dentro do generator
- **`yield from`**: delega a iteração para outro iterável ou generator

### Expressões Generator

Semelhantes a list comprehensions, mas usam `()` em vez de `[]`. São lazy: não alocam toda a lista na memória.

## 💻 Exemplos

````python
# Exemplo 1: Criando um decorador simples
def meu_decorador(func):
    def wrapper():
        print("Antes da função")
        func()
        print("Depois da função")
    return wrapper

@meu_decorador
def dizer_oi():
    print("Olá!")

dizer_oi()
````

````python
# Exemplo 2: Decorador com functools.wraps
from functools import wraps

def log_execucao(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print(f"Executando: {func.__name__}")
        resultado = func(*args, **kwargs)
        print(f"Finalizado: {func.__name__}")
        return resultado
    return wrapper

@log_execucao
def somar(a, b):
    """Soma dois números"""
    return a + b

print(somar(3, 5))
print(f"Nome: {somar.__name__}")  # Preservado graças a @wraps
print(f"Doc: {somar.__doc__}")
````

````python
# Exemplo 3: Decorador com argumentos
def repetir(n):
    def decorador(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorador

@repetir(3)
def mensagem(texto):
    print(f"Mensagem: {texto}")

mensagem("Python é incrível!")
````

````python
# Exemplo 4: @staticmethod e @classmethod
class Matematica:
    pi = 3.14159

    @staticmethod
    def somar(a, b):
        return a + b

    @classmethod
    def area_circulo(cls, raio):
        return cls.pi * raio ** 2

    def metodo_instancia(self, x):
        return x * 2

# Usando sem instanciar
print(Matematica.somar(10, 20))
print(Matematica.area_circulo(5))

# Também funciona com instância
m = Matematica()
print(m.somar(1, 2))
print(m.metodo_instancia(5))
````

````python
# Exemplo 5: @property (revisão e aprofundamento)
class Temperatura:
    def __init__(self, celsius):
        self._celsius = celsius

    @property
    def celsius(self):
        return self._celsius

    @celsius.setter
    def celsius(self, valor):
        if valor < -273.15:
            raise ValueError("Temperatura abaixo do zero absoluto")
        self._celsius = valor

    @property
    def fahrenheit(self):
        return self._celsius * 9 / 5 + 32

    @property
    def kelvin(self):
        return self._celsius + 273.15

t = Temperatura(25)
print(f"{t.celsius}°C = {t.fahrenheit:.1f}°F = {t.kelvin:.1f}K")
t.celsius = 100
print(f"{t.celsius}°C = {t.fahrenheit:.1f}°F")
````

````python
# Exemplo 6: Generator básico com yield
def contar_ate(n):
    contador = 1
    while contador <= n:
        yield contador
        contador += 1

# Criando o generator
gen = contar_ate(5)
print(type(gen))

# Iterando
for numero in gen:
    print(numero, end=" ")
````

````python
# Exemplo 7: Generator vs Lista (economia de memória)
import sys


def gerar_numeros(n):
    for i in range(n):
        yield i

def lista_numeros(n):
    return list(range(n))

n = 100000
gen = gerar_numeros(n)
lista = lista_numeros(n)

print(f"Tamanho do generator: {sys.getsizeof(gen)} bytes")
print(f"Tamanho da lista: {sys.getsizeof(lista)} bytes")
print(f"Diferença: {sys.getsizeof(lista) / sys.getsizeof(gen):.0f}x maior!")
````

````python
# Exemplo 8: send(), close() e yield from
def corrotina():
    print("Iniciando corrotina...")
    valor = yield "Pronto"
    print(f"Recebido: {valor}")
    yield "Processado"
    print("Finalizando...")

c = corrotina()
print(next(c))         # Inicia e obtém "Pronto"
print(c.send("Olá!")) # Envia valor e obtém "Processado"

# yield from
def gerar_pares(n):
    for i in range(n):
        if i % 2 == 0:
            yield i

def gerar_impares(n):
    for i in range(n):
        if i % 2 != 0:
            yield i

def gerar_tudo(n):
    yield from gerar_pares(n)
    yield from gerar_impares(n)

print("Pares e ímpares com yield from:")
print(list(gerar_tudo(10)))
````

## ✅ Exercícios Resolvidos

````python
# Exercício 1: Decorador para medir tempo de execução
import time
from functools import wraps

def medir_tempo(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        inicio = time.perf_counter()
        resultado = func(*args, **kwargs)
        fim = time.perf_counter()
        print(f"{func.__name__} levou {fim - inicio:.6f} segundos")
        return resultado
    return wrapper

@medir_tempo
def calcular_soma(n):
    return sum(range(n))

resultado = calcular_soma(1000000)
print(f"Resultado: {resultado}")
````

````python
# Exercício 2: Generator da sequência de Fibonacci
def fibonacci(n):
    a, b = 0, 1
    contador = 0
    while contador < n:
        yield a
        a, b = b, a + b
        contador += 1

print("Fibonacci (10 primeiros):")
for i, num in enumerate(fibonacci(10)):
    print(f"F{i} = {num}")
````

````python
# Exercício 3: Expressão generator para números primos
def eh_primo(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Expressão generator
primos_ate_50 = (x for x in range(50) if eh_primo(x))
print("Primos até 50:")
print(list(primos_ate_50))

# Comparação com list comprehension
import sys
lista_primos = [x for x in range(1000) if eh_primo(x)]
gen_primos = (x for x in range(1000) if eh_primo(x))
print(f"Lista: {sys.getsizeof(lista_primos)} bytes")
print(f"Generator: {sys.getsizeof(gen_primos)} bytes")
````

## 📌 Resumo

- **Decoradores**: funções que modificam outras funções. Sintaxe `@decorador`
- **`functools.wraps`**: preserva metadados da função original
- **`@staticmethod`**: método sem acesso a `self` ou `cls`
- **`@classmethod`**: método que recebe a classe (`cls`) como primeiro parâmetro
- **`@property`**: transforma método em atributo (com getter/setter opcionais)
- **`yield`**: pausa a função e retorna um valor, preservando o estado
- **`generator.send()`**: envia valor de volta para o generator
- **`yield from`**: delega iteração para outro iterável
- **Expressões generator**: `(x for x in ...)` — lazy, economizam memória

## 💡 Curiosidades

- O açúcar sintático `@decorador` foi adicionado no **Python 2.4** (2005), inspirado em Java annotations e outras linguagens.
- O padrão **Decorator** é um dos 23 padrões de design do GoF (Gang of Four).
- `yield from` foi introduzido no **Python 3.3** e simplificou muito o trabalho com generators aninhados.
- Generators são a base do **asyncio** em Python — corrotinas usam `yield` internamente (em Python < 3.5) e depois evoluíram para `async def`/`await`.
- A função `range()` em Python 3 é um generator (lazy), ao contrário do Python 2 onde `range()` retornava uma lista.

## ✅ Boas Práticas

- Use `@wraps` em todos os decoradores que você criar
- Decore apenas quando necessário — decoradores demais podem dificultar o debugging
- Prefira `@classmethod` em vez de métodos de instância que só usam `cls`
- Use generators para processar grandes volumes de dados (economia de memória)
- Use expressões generator para transformações simples em vez de list comprehensions quando não precisar da lista inteira
- Documente decoradores complexos com exemplos de uso
- Não abuse de `send()` — casos de uso são raros e o código fica menos legível

## ⚠️ Erros Comuns

| Erro | Exemplo | Correção |
|------|---------|----------|
| Esquecer `()` no decorador com args | `@repetir(3)` vs `@repetir` (sem argumentos) | Usar `()` se o decorador aceita args |
| Esquecer `@wraps` | Perde nome e docstring da função | Sempre usar `@wraps(func)` no wrapper |
| Confundir `@staticmethod` com `@classmethod` | `staticmethod` não tem acesso à classe | Usar `@classmethod` se precisar de `cls` |
| Retornar `return` em generator | Misturar `return` e `yield` causa `StopIteration` | Usar só `yield` (ou `return` vazio para finalizar) |
| Usar generator uma vez só | `list(gen); list(gen)` — segunda vez vazio | Criar novo generator ou converter para lista |
| Chamar função decorada sem `()` | `func` em vez de `func()` | Invocar a função com `()` |
| Achar que `@property` deixa privado | `@property` não cria atributo privado | Combinar com `_atributo` + `@property` + `@xxx.setter` |
