## 📝 Introducao

**Arvores** e **Grafos** sao estruturas de dados nao lineares essenciais para modelar hierarquias, redes, rotas e relacoes complexas.

- **Arvore**: estrutura hierarquica com um no raiz, nos filhos e folhas. Nao tem ciclos.
- **Grafo**: conjunto de vertices conectados por arestas. Pode ter ciclos e ser direcionado ou nao.

### Aplicacoes no mundo real:

- **BST**: indices de banco de dados, sistemas de arquivos
- **AVL**: dicionarios, sistemas que precisam de busca rapida balanceada
- **Grafos**: Google Maps (rotas), redes sociais, internet

## 📚 Explicacao Detalhada

### Arvore Binaria

Cada no tem no maximo 2 filhos: esquerdo e direito.

**Travessias (formas de percorrer):**
- **Pre-ordem**: raiz → esquerda → direita
- **Em-ordem**: esquerda → raiz → direita (BST: ordem crescente!)
- **Pos-ordem**: esquerda → direita → raiz

### Arvore Binaria de Busca (BST)

Regra: **esquerda < raiz < direita** para todos os nos.

- **Busca**: O(log n) medio, O(n) pior caso (arvore degenerada)
- **Insercao**: O(log n) medio
- **Remocao**: 3 casos — no folha, no com 1 filho, no com 2 filhos

### Arvore AVL

BST **auto-balanceada**. A diferenca de altura entre subarvores esquerda e direita (fator de balanceamento) nao pode exceder 1.

- Rotacoes mantem o balanceamento: simples (esquerda/direita) e duplas
- Garante O(log n) para todas as operacoes

### Grafos

Representacoes em Python:
- **Matriz de adjacencia**: matriz `n x n` (bom para grafos densos)
- **Lista de adjacencia**: dict de listas (bom para grafos esparsos, mais comum)

**Travessias:**
- **BFS (Busca em Largura)**: usa fila, encontra caminho mais curto em grafos nao ponderados
- **DFS (Busca em Profundidade)**: usa pilha/recursao, explora um caminho ate o fim

## 💻 Exemplos

```python
# Exemplo 1: Arvore Binaria e travessias
class No:
    def __init__(self, valor):
        self.valor = valor
        self.esquerda = None
        self.direita = None

class ArvoreBinaria:
    def __init__(self):
        self.raiz = None

    def pre_ordem(self, no=None):
        if no is None:
            no = self.raiz
        if no:
            print(no.valor, end=' ')
            self.pre_ordem(no.esquerda)
            self.pre_ordem(no.direita)

    def em_ordem(self, no=None):
        if no is None:
            no = self.raiz
        if no:
            self.em_ordem(no.esquerda)
            print(no.valor, end=' ')
            self.em_ordem(no.direita)

    def pos_ordem(self, no=None):
        if no is None:
            no = self.raiz
        if no:
            self.pos_ordem(no.esquerda)
            self.pos_ordem(no.direita)
            print(no.valor, end=' ')

arvore = ArvoreBinaria()
arvore.raiz = No(1)
arvore.raiz.esquerda = No(2)
arvore.raiz.direita = No(3)
arvore.raiz.esquerda.esquerda = No(4)
arvore.raiz.esquerda.direita = No(5)

print("Pre-ordem: ", end='')
arvore.pre_ordem()
print("\nEm-ordem: ", end='')
arvore.em_ordem()
print("\nPos-ordem: ", end='')
arvore.pos_ordem()
print()
```

```python
# Exemplo 2: Arvore Binaria de Busca (BST)
class BST:
    def __init__(self):
        self.raiz = None

    def inserir(self, valor):
        if not self.raiz:
            self.raiz = No(valor)
        else:
            self._inserir(self.raiz, valor)

    def _inserir(self, no, valor):
        if valor < no.valor:
            if no.esquerda:
                self._inserir(no.esquerda, valor)
            else:
                no.esquerda = No(valor)
        else:
            if no.direita:
                self._inserir(no.direita, valor)
            else:
                no.direita = No(valor)

    def buscar(self, valor):
        return self._buscar(self.raiz, valor)

    def _buscar(self, no, valor):
        if no is None:
            return False
        if valor == no.valor:
            return True
        if valor < no.valor:
            return self._buscar(no.esquerda, valor)
        return self._buscar(no.direita, valor)

    def em_ordem(self):
        resultado = []
        self._em_ordem(self.raiz, resultado)
        return resultado

    def _em_ordem(self, no, resultado):
        if no:
            self._em_ordem(no.esquerda, resultado)
            resultado.append(no.valor)
            self._em_ordem(no.direita, resultado)

bst = BST()
valores = [50, 30, 70, 20, 40, 60, 80]
for v in valores:
    bst.inserir(v)

print(f"BST em ordem: {bst.em_ordem()}")
print(f"Buscar 40: {bst.buscar(40)}")
print(f"Buscar 25: {bst.buscar(25)}")
```

```python
# Exemplo 3: Implementacao basica de AVL
class NoAVL:
    def __init__(self, valor):
        self.valor = valor
        self.esquerda = None
        self.direita = None
        self.altura = 1

class AVL:
    def altura(self, no):
        return no.altura if no else 0

    def fator_balanceamento(self, no):
        return self.altura(no.esquerda) - self.altura(no.direita) if no else 0

    def rotacionar_direita(self, y):
        x = y.esquerda
        T2 = x.direita
        x.direita = y
        y.esquerda = T2
        y.altura = max(self.altura(y.esquerda), self.altura(y.direita)) + 1
        x.altura = max(self.altura(x.esquerda), self.altura(x.direita)) + 1
        return x

    def rotacionar_esquerda(self, x):
        y = x.direita
        T2 = y.esquerda
        y.esquerda = x
        x.direita = T2
        x.altura = max(self.altura(x.esquerda), self.altura(x.direita)) + 1
        y.altura = max(self.altura(y.esquerda), self.altura(y.direita)) + 1
        return y

    def inserir(self, no, valor):
        if not no:
            return NoAVL(valor)

        if valor < no.valor:
            no.esquerda = self.inserir(no.esquerda, valor)
        else:
            no.direita = self.inserir(no.direita, valor)

        no.altura = max(self.altura(no.esquerda), self.altura(no.direita)) + 1
        balance = self.fator_balanceamento(no)

        # Rotacoes
        if balance > 1 and valor < no.esquerda.valor:
            return self.rotacionar_direita(no)
        if balance < -1 and valor > no.direita.valor:
            return self.rotacionar_esquerda(no)
        if balance > 1 and valor > no.esquerda.valor:
            no.esquerda = self.rotacionar_esquerda(no.esquerda)
            return self.rotacionar_direita(no)
        if balance < -1 and valor < no.direita.valor:
            no.direita = self.rotacionar_direita(no.direita)
            return self.rotacionar_esquerda(no)

        return no

avl = AVL()
raiz = None
for v in [10, 20, 30, 40, 50, 25]:
    raiz = avl.inserir(raiz, v)

print(f"AVL balanceada — Fator da raiz: {avl.fator_balanceamento(raiz)}")
print(f"Raiz: {raiz.valor}")
print(f"Altura da arvore: {raiz.altura}")
```

```python
# Exemplo 4: Grafo com lista de adjacencia
from collections import deque

class Grafo:
    def __init__(self):
        self.adjacencias = {}

    def adicionar_vertice(self, v):
        if v not in self.adjacencias:
            self.adjacencias[v] = []

    def adicionar_aresta(self, v1, v2, direcionado=False):
        self.adicionar_vertice(v1)
        self.adicionar_vertice(v2)
        self.adjacencias[v1].append(v2)
        if not direcionado:
            self.adjacencias[v2].append(v1)

    def bfs(self, inicio):
        visitados = set()
        fila = deque([inicio])
        ordem = []
        while fila:
            v = fila.popleft()
            if v not in visitados:
                visitados.add(v)
                ordem.append(v)
                for vizinho in self.adjacencias[v]:
                    if vizinho not in visitados:
                        fila.append(vizinho)
        return ordem

    def dfs(self, inicio):
        visitados = set()
        ordem = []

        def _dfs(v):
            visitados.add(v)
            ordem.append(v)
            for vizinho in self.adjacencias[v]:
                if vizinho not in visitados:
                    _dfs(vizinho)

        _dfs(inicio)
        return ordem

g = Grafo()
g.adicionar_aresta('A', 'B')
g.adicionar_aresta('A', 'C')
g.adicionar_aresta('B', 'D')
g.adicionar_aresta('B', 'E')
g.adicionar_aresta('C', 'F')
g.adicionar_aresta('E', 'F')

print(f"BFS a partir de A: {g.bfs('A')}")
print(f"DFS a partir de A: {g.dfs('A')}")
```

```python
# Exemplo 5: Grafo — encontrando caminho mais curto (BFS)
def caminho_mais_curto(grafo, inicio, destino):
    fila = deque([[inicio]])
    visitados = {inicio}

    while fila:
        caminho = fila.popleft()
        ultimo = caminho[-1]

        if ultimo == destino:
            return caminho

        for vizinho in grafo.adjacencias[ultimo]:
            if vizinho not in visitados:
                visitados.add(vizinho)
                fila.append(caminho + [vizinho])

    return None

mapa = Grafo()
mapa.adicionar_aresta('Casa', 'Padaria')
mapa.adicionar_aresta('Casa', 'Escola')
mapa.adicionar_aresta('Padaria', 'Mercado')
mapa.adicionar_aresta('Escola', 'Mercado')
mapa.adicionar_aresta('Mercado', 'Trabalho')
mapa.adicionar_aresta('Escola', 'Parque')
mapa.adicionar_aresta('Parque', 'Trabalho')

caminho = caminho_mais_curto(mapa, 'Casa', 'Trabalho')
print(f"Caminho mais curto da Casa ao Trabalho:")
print(" -> ".join(caminho))
```

```python
# Exercicio 1: Encontrar menor valor em uma BST
def menor_valor(raiz):
    atual = raiz
    while atual and atual.esquerda:
        atual = atual.esquerda
    return atual.valor if atual else None

def maior_valor(raiz):
    atual = raiz
    while atual and atual.direita:
        atual = atual.direita
    return atual.valor if atual else None

print(f"BST - Menor valor: {menor_valor(bst.raiz)}")
print(f"BST - Maior valor: {maior_valor(bst.raiz)}")
```

```python
# Exercicio 2: Contar vertices e arestas de um grafo
def info_grafo(grafo):
    vertices = len(grafo.adjacencias)
    arestas = sum(len(vizinhos) for vizinhos in grafo.adjacencias.values())
    # Em grafo nao direcionado, cada aresta e contada 2x
    return vertices, arestas // 2

v, a = info_grafo(mapa)
print(f"Grafo do mapa: {v} vertices, {a} arestas")
print(f"Adjacencias: {dict(mapa.adjacencias)}")
```

## 📌 Resumo

- **Arvore**: estrutura hierarquica; nos, arestas, raiz, folhas
- **Arvore Binaria**: cada no tem ate 2 filhos
- **Travessias**: pre-ordem (raiz primeiro), em-ordem (esquerda→raiz→direita), pos-ordem
- **BST**: esquerda < raiz < direita — busca O(log n)
- **AVL**: BST balanceada com rotacoes — garante O(log n)
- **Grafo**: vertices + arestas; representado por lista de adjacencia
- **BFS**: nivel por nivel (fila) — caminho mais curto em grafos nao ponderados
- **DFS**: profundidade primeiro (pilha/recursao) — explora ate o fim

## 💡 Curiosidades

- O algoritmo **Dijkstra** foi criado por Edsger Dijkstra em 1956, em apenas 20 minutos, enquanto tomava cafe com sua noiva.
- O conceito de arvore em computacao e invertido: a raiz fica no topo e as folhas embaixo!
- **AVL** foi nomeada em homenagem aos seus inventores: **A**delson-**V**elsky e **L**andis (1962).
- Grafos sao usados pelo algoritmo **PageRank** do Google para rankear paginas web.
- A estrutura de **heap** (base para filas de prioridade) e implementada como uma arvore binaria completa.

## ✅ Boas Praticas

- Use lista de adjacencia para grafos (mais eficiente em memoria que matriz)
- Recursao e natural para travessias de arvores, mas cuidado com profundidade
- Balanceie a BST (ou use AVL) se houver muitas insercoes e remocoes
- BFS usa fila; DFS usa pilha — escolha baseado no problema
- Para grafos muito grandes, use `networkx` (biblioteca Python especializada)

## ⚠️ Erros Comuns

| Erro | Exemplo | Correcao |
|------|---------|----------|
| BST degenerada | Inserir [1,2,3,4,5] vira lista encadeada | Usar AVL ou embaralhar insercoes |
| Esquecer caso base na recursao | Stack overflow em arvore profunda | Sempre verificar `if no is None` |
| BFS sem controle de visitados | Loop infinito em grafos com ciclos | Manter conjunto de visitados |
| Confundir arvore binaria com BST | Qualquer arvore binaria nao e BST | Validar propriedade: esq < no < dir |
| Nao balancear AVL apos remocao | Arvore fica desbalanceada | Rotacionar apos cada remocao || Criar grafo sem adicionar vertices primeiro | KeyError ao acessar adjacencia | Usar `setdefault` ou verificar existencia |
