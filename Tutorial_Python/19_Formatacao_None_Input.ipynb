## 📝 Introdução

Três tópicos fundamentais para qualquer programa Python: **formatar strings** para exibir dados de forma elegante, entender o **`None`** (ausência de valor) e **coletar entradas do usuário** com `input()`. As f-strings tornaram a formatação muito mais simples e legível. O `None` é essencial para representar "nada" em funções e estruturas de dados. E o `input()` permite criar programas interativos. Vamos dominar esses conceitos!

## 📚 Explicação Detalhada

### Formatação de Strings

| Método | Exemplo | Descrição |
|--------|---------|-----------|
| f-strings | `f"Olá, {nome}"` | Mais moderna e legível (Python 3.6+) |
| `.format()` | `"Olá, {}".format(nome)` | Versátil, permite posições e nomes |
| `%` | `"Olá, %s" % nome` | Estilo antigo (C/printf) |

**Especificadores de formato**:
- `:.2f` → duas casas decimais
- `:d` → inteiro
- `:>10` → alinhado à direita em 10 espaços
- `:<10` → alinhado à esquerda
- `:^10` → centralizado
- `:,.2f` → separador de milhares + 2 casas

### None

- `None` é um objeto singleton do tipo `NoneType`
- Representa ausência de valor (não é zero, nem string vazia)
- Use `is None` ou `is not None` para comparar (não `==`)
- Comum como valor padrão em parâmetros de funções e retorno de funções sem `return`

### input()

- `input(prompt)` → sempre retorna uma **string**
- Para obter número: `int(input(...))` ou `float(input(...))`
- Para booleanos: comparar a string com `"True"` ou `"sim"`

## 💻 Exemplos

### Exemplo 1: f-strings básicas

```python
nome = "Maria"
idade = 25
altura = 1.68

print(f"Olá, meu nome é {nome}.")
print(f"Tenho {idade} anos e {altura:.2f}m de altura.")
print(f"Ano que vem terei {idade + 1} anos.")
```

### Exemplo 2: f-strings com expressões e formatação

```python
preco = 49.9
quantidade = 3
total = preco * quantidade

print(f"Preço unitário: R$ {preco:.2f}")
print(f"Quantidade: {quantidade:3d}")
print(f"Total: R$ {total:>8.2f}")
print(f"Preço com desconto (10%): R$ {total * 0.9:.2f}")
print(f"Total formatado: R$ {total:,.2f}")
```

### Exemplo 3: Método .format()

```python
print("Nome: {}, Idade: {}".format("João", 30))
print("Nome: {nome}, Idade: {idade}".format(nome="Ana", idade=22))
print("Produto: {0:15} | Preço: R$ {1:>8.2f}".format("Notebook", 3500.0))
print("{:<10} {:^10} {:>10}".format("Esquerda", "Centro", "Direita"))
```

### Exemplo 4: Operador % (estilo antigo)

```python
nome = "Carlos"
nota = 8.5
print("O aluno %s tirou %.1f." % (nome, nota))
print("Inteiro: %d | Float: %.3f | String: %s" % (42, 3.14159, "Python"))
```

### Exemplo 5: Trabalhando com None

```python
valor = None
print(f"Tipo de None: {type(valor)}")
print(f"None é igual a 0? {None == 0}")
print(f"None é igual a ''? {None == ''}")
print(f"None é None? {valor is None}")
print(f"None não é None? {valor is not None}")
```

### Exemplo 6: None como valor padrão em funções

```python
def saudacao(nome=None):
    if nome is None:
        return "Olá, visitante!"
    return f"Olá, {nome}!"

print(saudacao())
print(saudacao("Lucas"))
print(saudacao(None))
```

### Exemplo 7: input() básico

```python
nome = input("Digite seu nome: ")
print(f"Prazer em conhecê-lo, {nome}!")
```

### Exemplo 8: Conversão de tipos no input

```python
try:
    idade_str = input("Qual sua idade? ")
    idade = int(idade_str)
    altura = float(input("Qual sua altura (em metros)? "))
    resposta = input("Você estuda Python? (sim/não): ").strip().lower()
    estuda_python = resposta in ("sim", "s", "yes")

    print(f"\n--- Dados Cadastrados ---")
    print(f"Idade: {idade} anos")
    print(f"Altura: {altura:.2f}m")
    print(f"Estuda Python: {'Sim' if estuda_python else 'Não'}")
except ValueError:
    print("Erro: certifique-se de digitar números válidos!")
```

## ✅ Exercícios Resolvidos

### Exercício 1: Cadastro formatado com f-strings

```python
def cadastrar_usuario():
    nome = input("Nome: ").strip()
    idade = int(input("Idade: "))
    salario = float(input("Salário: R$ "))

    print("\n" + "=" * 40)
    print(f"{'DADOS DO USUÁRIO':^40}")
    print("=" * 40)
    print(f"{'Nome:':<10} {nome}")
    print(f"{'Idade:':<10} {idade} anos")
    print(f"{'Salário:':<10} R$ {salario:>8.2f}")
    print("=" * 40)

cadastrar_usuario()
```

### Exercício 2: Função com None para busca

```python
def buscar_contato(agenda, nome):
    for contato in agenda:
        if contato["nome"].lower() == nome.lower():
            return contato
    return None

agenda = [
    {"nome": "Ana", "telefone": "11999999999"},
    {"nome": "João", "telefone": "21988888888"},
]

nome_busca = input("Nome para buscar: ").strip()
resultado = buscar_contato(agenda, nome_busca)

if resultado is not None:
    print(f"Telefone de {resultado['nome']}: {resultado['telefone']}")
else:
    print("Contato não encontrado.")
```

### Exercício 3: Tabela formatada com .format()

```python
print("=" * 35)
print("{:<15} {:>10} {:>8}".format("Produto", "Qtd", "Preço"))
print("=" * 35)
print("{:<15} {:>10} {:>8}".format("Arroz", 5, "R$ 25.00"))
print("{:<15} {:>10} {:>8}".format("Feijão", 3, "R$ 18.50"))
print("{:<15} {:>10} {:>8}".format("Macarrão", 10, "R$ 8.90"))
print("=" * 35)
```

## 📌 Resumo

- **f-strings** (`f"...{var}"`) são a forma mais moderna e recomendada de formatar strings
- **`.format()`** oferece mais opções de posicionamento e nomeação
- **Operador `%`** é o método mais antigo, ainda presente em código legado
- `None` representa ausência de valor, compare com `is None` (não `== None`)
- `input()` sempre retorna string — converta com `int()` ou `float()` quando necessário
- Trate conversões com `try/except` para evitar `ValueError`

## 💡 Curiosidades

- f-strings foram introduzidas no Python 3.6 e são mais rápidas que `.format()` e `%`
- `None` é um singleton — só existe uma instância de `NoneType` na memória
- Você pode usar `f"{2 * 10}"` dentro de f-strings — expressões são avaliadas
- Em Python 3.8+, f-strings suportam `=` para depuração: `f"{var=}"` exibe `var=valor`

## ✅ Boas Práticas

- Prefira f-strings sempre que possível (Python 3.6+)
- Use `is None` em vez de `== None` para comparação
- Sempre valide e converta entradas do usuário com `try/except`
- Use `.strip()` em strings de `input()` para remover espaços extras
- Use `None` como valor padrão para parâmetros mutáveis (evita bugs com listas/dicts)

## ⚠️ Erros Comuns

| Erro | Causa | Solução |
|------|-------|---------|
| `ValueError: invalid literal for int()` | Tentar converter string não numérica com `int()` | Use `try/except` ou valide antes |
| Comparar com `None` usando `==` | `None == 0` é `False`, mas `None == None` funciona | Use `x is None` em vez de `x == None` |
| `TypeError: unsupported format character` | Usar formatação `%` com tipo errado | Verifique se o especificador corresponde ao tipo |
| `AttributeError: 'NoneType' object has no attribute '...'` | Tentar acessar método em valor `None` | Verifique se o valor não é `None` antes de usar |
| `NameError` ao usar input | Esquecer de atribuir o retorno de `input()` a uma variável | Sempre faça `var = input(...)` |
