## 📝 Introdução

Funções são blocos de código reutilizáveis que executam uma tarefa específica. Em Python, funções são objetos de primeira classe: podem ser passadas como argumentos, retornadas de outras funções e armazenadas em variáveis. Este notebook cobre desde a definição básica até conceitos avançados como `*args`, `**kwargs`, funções lambda e recursão.

## 📚 Explicação Detalhada

### def e return
Use `def nome_funcao(parametros):` para definir. `return` envia um valor de volta (ou `None` se omitido).

### Parâmetros e Argumentos
- Parâmetros posicionais, nomeados (keyword), valores padrão
- `*args`: tupla com argumentos posicionais extras
- `**kwargs`: dicionário com argumentos nomeados extras

### Funções Lambda
Funções anônimas de uma única expressão: `lambda args: expressão`

### Recursão
Função que chama a si mesma. Precisa de um caso base para terminar.

## 💻 Exemplos

```python
# Exemplo 1: Função simples
def saudacao(nome):
    return f"Olá, {nome}!"

print(saudacao("Maria"))
```

```python
# Exemplo 2: Argumento padrão
def potencia(base, expoente=2):
    return base ** expoente

print(potencia(3))
print(potencia(3, 3))
```

```python
# Exemplo 3: *args (argumentos posicionais variáveis)
def soma_todos(*args):
    return sum(args)

print(soma_todos(1, 2, 3, 4, 5))
print(soma_todos(10, 20))
```

```python
# Exemplo 4: **kwargs (argumentos nomeados variáveis)
def exibir_info(**kwargs):
    for chave, valor in kwargs.items():
        print(f"{chave}: {valor}")

exibir_info(nome="Ana", idade=30, cidade="São Paulo")
```

```python
# Exemplo 5: Lambda com map
numeros = [1, 2, 3, 4, 5]
quadrados = list(map(lambda x: x**2, numeros))
print(quadrados)
```

```python
# Exemplo 6: Lambda com filter
numeros = [10, 15, 22, 33, 40, 55]
pares = list(filter(lambda x: x % 2 == 0, numeros))
print(pares)
```

```python
# Exemplo 7: Recursão — Fatorial
def fatorial(n):
    if n <= 1:
        return 1
    return n * fatorial(n - 1)

print(f"Fatorial de 5: {fatorial(5)}")
print(f"Fatorial de 7: {fatorial(7)}")
```

```python
# Exemplo 8: Recursão — Fibonacci
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

for i in range(10):
    print(f"F({i}) = {fibonacci(i)}")
```

## ✅ Exercícios Resolvidos

```python
# Exercício Resolvido 1: Função que conta vogais
def contar_vogais(texto):
    vogais = "aeiouAEIOU"
    return sum(1 for char in texto if char in vogais)

print(contar_vogais("Python é incrível!"))
```

```python
# Exercício Resolvido 2: Função com *args para média
def media(*notas):
    if len(notas) == 0:
        return 0
    return sum(notas) / len(notas)

print(f"Média: {media(7.5, 8.0, 9.2, 6.8)}")
print(f"Média vazia: {media()}")
```

```python
# Exercício Resolvido 3: Lambda para ordenar por último nome
nomes = ["Ana Sousa", "Bruno Costa", "Carla Alves"]
nomes.sort(key=lambda nome: nome.split()[-1])
print(nomes)
```

## 📌 Resumo

- Funções são definidas com `def` e retornam valores com `return`
- Parâmetros podem ter valores padrão
- `*args` coleta argumentos posicionais extras; `**kwargs` coleta nomeados
- Lambda são funções anônimas de uma linha
- `map()` aplica uma função a cada item; `filter()` filtra itens
- Recursão: função que chama a si mesma, precisa de caso base
- Funções em Python são objetos de primeira classe

## 💡 Curiosidades

- O termo "Lambda" vem do cálculo lambda, um sistema formal da lógica matemática
- Python limita a recursão a ~1000 chamadas por padrão (`sys.getrecursionlimit()`)
- `*args` e `**kwargs` são convenções de nome; você pode usar qualquer nome (ex.: `*numeros`)
- O `map()` da Python 3 é lazy — retorna um iterador, não uma lista

## ✅ Boas Práticas

- Use nomes descritivos para funções (verbos: `calcular_media`, `obter_usuario`)
- Mantenha funções pequenas e com uma única responsabilidade
- Documente funções com docstrings (`"""descrição"""`)
- Evite funções lambda complexas — prefira `def` para lógicas com mais de uma linha
- Cuidado com argumentos mutáveis como valor padrão (`def f(x=[])` causa efeito colateral)

## ⚠️ Erros Comuns

| Erro | Descrição | Solução |
|------|-----------|---------|
| Esquecer `return` | Função retorna `None` | Adicionar `return` com o valor desejado |
| Argumento mutável padrão | `def f(x=[])` reusa a mesma lista | Usar `None` e criar dentro: `if x is None: x = []` |
| Confundir `*args` com `**kwargs` | Ordem errada na definição | `def f(pos, *args, padrao=True, **kwargs)` |
| Recursão infinita | Esquecer o caso base | Sempre definir condição de parada |
| Modificar variável global | Efeito colateral indesejado | Usar `return` em vez de alterar globais |
| TypeError na lambda | Expression muito complexa | Usar `def` para lógica com múltiplas linhas |
