## 📝 Introdução

Strings são sequências de caracteres usadas para representar texto. Em Python, strings são imutáveis — uma vez criadas, não podem ser alteradas no lugar. Dominar a manipulação de strings é fundamental para processamento de texto, formatação de saída e desenvolvimento web.

## 📚 Explicação Detalhada

### Criando Strings
- Aspas simples: `'texto'`
- Aspas duplas: `"texto"`
- Aspas triplas (multilinha): `'''texto'''` ou `"""texto"""`

### Indexação e Slicing
- Cada caractere possui um índice (começa em 0)
- Índices negativos contam do final (`-1` é o último)
- Slicing: `string[inicio:fim:passo]`

### Métodos Comuns
| Método | Descrição |
|--------|-----------|
| `upper()` | Converte para maiúsculas |
| `lower()` | Converte para minúsculas |
| `strip()` | Remove espaços nas bordas |
| `replace(a, b)` | Substitui `a` por `b` |
| `split()` | Divide em lista |
| `join()` | Junta lista em string |
| `find()` | Busca substring |
| `count()` | Conta ocorrências |

## 💻 Exemplos

```python
# Exemplo 1: Criando strings
simples = 'Olá'
duplas = "Mundo"
multilinha = """Este é um texto
com múltiplas
linhas."""
print(simples, duplas)
print(multilinha)
```

```python
# Exemplo 2: Indexação
texto = "Python"
print(f"Primeiro caractere: {texto[0]}")     # P
print(f"Terceiro caractere: {texto[2]}")     # t
print(f"Último caractere: {texto[-1]}")      # n
print(f"Penúltimo: {texto[-2]}")             # o
print(f"Tamanho: {len(texto)}")
```

```python
# Exemplo 3: Fatiamento (slicing)
s = "Programação"
print(s[0:6])    # 'Progra'  — índices 0 a 5
print(s[:4])     # 'Prog'    — do início até 3
print(s[6:])     # 'mação'   — do índice 6 até o fim
print(s[-5:])    # 'mação'   — últimos 5 caracteres
print(s[::2])    # 'Pormço'  — pulando de 2 em 2
print(s[::-1])   # 'oãçamargorP' — invertida
```

```python
# Exemplo 4: Métodos de modificação
frase = "  Aprender Python é divertido!  "
print(frase.upper())
print(frase.lower())
print(frase.strip())          # remove espaços nas pontas
print(frase.replace("Python", "Java"))
print(frase.replace(" ", ""))  # remove todos os espaços
```

```python
# Exemplo 5: split() e join()
nomes = "Ana, Bruno, Carla, Daniel"
lista = nomes.split(", ")
print(lista)

numeros = ["um", "dois", "três"]
juncao = " - ".join(numeros)
print(juncao)
```

```python
# Exemplo 6: Concatenação
a = "Python"
b = "é"
c = "legal"
resultado = a + " " + b + " " + c
print(resultado)
print("Olá " * 3)  # repetição
```

```python
# Exemplo 7: Formatação com f-strings
nome = "Maria"
idade = 28
altura = 1.65
print(f"Meu nome é {nome}, tenho {idade} anos e {altura:.2f} m de altura.")

# Operações dentro das chaves
a, b = 10, 3
print(f"{a} + {b} = {a + b}")
print(f"{a} / {b} = {a / b:.3f}")
```

```python
# Exemplo 8: Formatação com .format()
print("Meu nome é {} e tenho {} anos.".format("João", 32))
print("Produto: {prod}, Preço: R$ {preco:.2f}".format(prod="Mouse", preco=49.90))

# Caracteres de escape
print("Linha 1\nLinha 2")      # \n = nova linha
print("Coluna1\tColuna2")      # \t = tab
print("Ele disse: \"Python é ótimo!\"")  # \" = aspas literais
```

## ✅ Exercícios Resolvidos

```python
# Exercício Resolvido 1: Inverter nome completo
nome_completo = "João Silva Santos"
partes = nome_completo.split()
invertido = " ".join([partes[-1]] + partes[:-1])
print(f"Original: {nome_completo}")
print(f"Invertido (último nome primeiro): {invertido}")
```

```python
# Exercício Resolvido 2: Contador de vogais
frase = input("Digite uma frase: ").lower()
vogais = "aeiou"
contador = 0
for letra in frase:
    if letra in vogais:
        contador += 1
print(f"A frase contém {contador} vogal(is).")
```

```python
# Exercício Resolvido 3: Validador de palíndromo
palavra = input("Digite uma palavra: ").strip().lower()
invertida = palavra[::-1]
if palavra == invertida:
    print(f"'{palavra}' é um palíndromo!")
else:
    print(f"'{palavra}' não é um palíndromo. Invertido: {invertida}")
```

## 📌 Resumo

- Strings são sequências imutáveis de caracteres
- Indexação começa em 0; índices negativos contam do final
- Slicing permite extrair substrings: `[inicio:fim:passo]`
- Métodos como `upper()`, `lower()`, `strip()`, `replace()` retornam **novas** strings
- `split()` divide em lista; `join()` une lista em string
- Prefira **f-strings** para formatação moderna de strings

## 💡 Curiosidades

- Strings em Python são armazenadas como Unicode (UTF-8), suportando acentos e emojis.
- O operador `in` verifica se uma substring existe: `"Py" in "Python"` → `True`.
- O método `isalpha()` verifica se a string contém apenas letras.

## ✅ Boas Práticas

- Prefira f-strings em vez de concatenação com `+`
- Use `.strip()` ao processar entradas do usuário
- Evite strings multilinha com `\n` explícito; prefira aspas triplas
- Use métodos encadeados com moderação para não prejudicar a legibilidade

## ⚠️ Erros Comuns

| Erro | Exemplo | Solução |
|------|---------|--------|
| IndexError | `texto[10]` em string curta | Verifique `len()` antes |
| Tentar modificar string | `s[0] = "x"` | Crie nova string: `"x" + s[1:]` |
| Esquecer que slicing exclui o fim | `s[0:4]` pega índices 0,1,2,3 | Lembre-se: `[inicio:fim)` |
| Confundir `split()` com `join()` | Usar `split()` para unir | `split()` divide; `join()` une |
