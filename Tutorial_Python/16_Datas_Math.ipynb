## 📚 Explicação Detalhada

### Módulo `datetime`

O módulo `datetime` fornece classes para manipular datas e horas:

| Classe       | Descrição                                 | Exemplo               |
|--------------|-------------------------------------------|-----------------------|
| `date`       | Data (ano, mês, dia)                      | `date(2025, 12, 25)`  |
| `time`       | Hora (hora, minuto, segundo, microssegundo)| `time(14, 30, 0)`    |
| `datetime`   | Combina data + hora                       | `datetime(2025, 12, 25, 14, 30)` |
| `timedelta`  | Diferença entre duas datas/tempos         | `timedelta(days=7)`   |

**`strftime`**: formata um objeto datetime em string
**`strptime`**: converte uma string em objeto datetime

### Módulo `math`

Funções matemáticas comuns:

| Função       | Descrição                     | Exemplo              |
|--------------|-------------------------------|----------------------|
| `ceil(x)`    | Arredonda para cima           | `math.ceil(4.2) → 5`|
| `floor(x)`   | Arredonda para baixo          | `math.floor(4.8) → 4`|
| `sqrt(x)`    | Raiz quadrada                 | `math.sqrt(9) → 3`  |
| `pow(x, y)`  | Potência (x elevado a y)      | `math.pow(2, 3) → 8`|
| `log(x)`     | Logaritmo natural (base e)    | `math.log(10)`      |
| `sin(x)`     | Seno (radianos)               | `math.sin(math.pi/2)`|
| `cos(x)`     | Cosseno (radianos)            | `math.cos(0) → 1`   |
| `pi`         | Constante π (3.1415...)       | `math.pi`           |
| `e`          | Constante de Euler (2.718...)| `math.e`            |

## 💻 Exemplos

### Exemplo 1: Trabalhando com date, time e datetime

```python
from datetime import date, time, datetime

data = date(2025, 12, 25)
hora = time(14, 30, 0)
dt = datetime(2025, 12, 25, 14, 30, 0)

print("Data:", data)
print("Hora:", hora)
print("Data e Hora:", dt)
print("Data de hoje:", date.today())
print("Agora:", datetime.now())
```

### Exemplo 2: Formatando datas com strftime

```python
from datetime import datetime

agora = datetime.now()
print("Formato padrão:", agora.strftime("%d/%m/%Y %H:%M:%S"))
print("Dia da semana:", agora.strftime("%A"))
print("Mês abreviado:", agora.strftime("%b"))
print("Ano com dois dígitos:", agora.strftime("%y"))
print("Formato BR completo:", agora.strftime("%d de %B de %Y"))
```

### Exemplo 3: Convertendo string para datetime com strptime

```python
from datetime import datetime

data_str = "15/03/2025 10:30:00"
data_obj = datetime.strptime(data_str, "%d/%m/%Y %H:%M:%S")
print("Objeto datetime:", data_obj)
print("Ano:", data_obj.year)
print("Mês:", data_obj.month)
print("Dia:", data_obj.day)
```

### Exemplo 4: Operações com timedelta

```python
from datetime import datetime, timedelta

hoje = datetime.now()
daqui_7_dias = hoje + timedelta(days=7)
ha_30_dias = hoje - timedelta(days=30)
daqui_2_horas = hoje + timedelta(hours=2)

print("Hoje:", hoje.strftime("%d/%m/%Y %H:%M"))
print("Daqui 7 dias:", daqui_7_dias.strftime("%d/%m/%Y %H:%M"))
print("Há 30 dias:", ha_30_dias.strftime("%d/%m/%Y %H:%M"))
print("Daqui 2 horas:", daqui_2_horas.strftime("%d/%m/%Y %H:%M"))
```

### Exemplo 5: Diferença entre duas datas

```python
from datetime import date

data1 = date(2025, 1, 1)
data2 = date(2025, 12, 25)
diferenca = data2 - data1

print(f"Diferença: {diferenca.days} dias")
print(f"Diferença em segundos: {diferenca.total_seconds():.0f} s")
```

### Exemplo 6: math.ceil, math.floor e math.sqrt

```python
import math

numero = 9.7

print(f"Número original: {numero}")
print(f"Arredondar para cima (ceil): {math.ceil(numero)}")
print(f"Arredondar para baixo (floor): {math.floor(numero)}")
print(f"Raiz quadrada de {numero}: {math.sqrt(numero):.2f}")
print(f"Raiz quadrada de 144: {math.sqrt(144)}")
```

### Exemplo 7: math.pow, pi, e e log

```python
import math

print(f"2 elevado a 10: {math.pow(2, 10)}")
print(f"Valor de pi: {math.pi}")
print(f"Valor de e: {math.e}")
print(f"Logaritmo natural de 100: {math.log(100):.2f}")
print(f"Logaritmo base 10 de 1000: {math.log(1000, 10)}")
```

### Exemplo 8: math.sin e math.cos

```python
import math

angulo_graus = 45
angulo_rad = math.radians(angulo_graus)

print(f"Seno de {angulo_graus}°: {math.sin(angulo_rad):.4f}")
print(f"Cosseno de {angulo_graus}°: {math.cos(angulo_rad):.4f}")
print(f"Seno de pi/2: {math.sin(math.pi/2)}")
print(f"Cosseno de pi: {math.cos(math.pi)}")
```

## ✅ Exercícios Resolvidos

### Exercício 1: Dias até o próximo aniversário

```python
from datetime import date

aniversario = date(2026, 3, 15)
hoje = date.today()
dias_restantes = (aniversario - hoje).days

if dias_restantes > 0:
    print(f"Faltam {dias_restantes} dias para seu aniversário!")
elif dias_restantes == 0:
    print("Feliz aniversário! 🎉")
else:
    print(f"Seu aniversário já passou há {abs(dias_restantes)} dias.")
```

### Exercício 2: Calculadora de hipotenusa

```python
import math

cateto_a = 6
cateto_b = 8
hipotenusa = math.sqrt(math.pow(cateto_a, 2) + math.pow(cateto_b, 2))

print(f"Catetos: {cateto_a} e {cateto_b}")
print(f"Hipotenusa: {hipotenusa:.2f}")
```

### Exercício 3: Arredondar valor para cima

```python
import math

preco_unitario = 7.49
quantidade = 12
total = preco_unitario * quantidade
total_arredondado = math.ceil(total * 100) / 100

print(f"Total sem arredondar: R$ {total:.2f}")
print(f"Total arredondado para cima: R$ {total_arredondado:.2f}")
```

## 📌 Resumo

- O módulo `datetime` fornece `date`, `time`, `datetime` e `timedelta` para manipular datas e horas
- `strftime` converte objeto datetime em string formatada; `strptime` faz o inverso
- `timedelta` permite somar/subtrair dias, horas, minutos etc.
- O módulo `math` oferece funções como `ceil`, `floor`, `sqrt`, `pow`, `log`, `sin`, `cos`
- `math.pi` e `math.e` são constantes importantes
- Use `math.radians()` para converter graus em radianos antes de usar `sin`/`cos`

## 💡 Curiosidades

- O módulo `datetime` não considera fusos horários por padrão. Para isso, use `pytz` ou `zoneinfo` (Python 3.9+)
- No módulo `math`, as funções trigonométricas trabalham em radianos, não em graus
- `math.ceil(4.0)` retorna 4 (não arredonda para cima se já for inteiro)
- `math.pow(2, 3)` retorna um `float` (8.0), enquanto `2 ** 3` retorna um `int` (8)

## ✅ Boas Práticas

- Sempre importe apenas as classes/funções que você vai usar (`from datetime import date`)
- Ao formatar datas, prefira os códigos de formatação padrão (%d, %m, %Y) em vez de concatenar manualmente
- Use `timedelta` para cálculos de tempo em vez de tentar somar manualmente dias ou horas
- Considere anos bissextos ao calcular diferenças longas entre datas
- Para operações matemáticas mais avançadas, explore o módulo `numpy` ou `statistics`

## ⚠️ Erros Comuns

| Erro | Causa | Solução |
|------|-------|---------|
| `ValueError: unconverted data remains` | String com caracteres extras ao converter com strptime | A string deve corresponder exatamente ao formato fornecido |
| `TypeError: can't subtract offset-naive and offset-aware datetimes` | Comparar datetime com e sem fuso horário | Use `datetime.now(timezone.utc)` ou normalize ambos para naive |
| `ValueError: month must be in 1..12` | Mês fora do intervalo 1-12 | Verifique sempre os valores passados ao construtor |
| `math.sqrt(-1)` causa `ValueError` | Raiz quadrada de número negativo | Use `cmath.sqrt()` para números complexos |
| `math.sin(45)` retorna valor inesperado | O argumento deve estar em radianos, não graus | Converta com `math.radians(45)` antes |
