## 📝 Introducao

**Classificacao** e o tipo mais comum de aprendizado supervisionado, onde o objetivo e prever uma **categoria** (classe) em vez de um valor numerico. Exemplos: spam ou nao spam, doenca ou saudavel, tipo de flor, etc.

### Algoritmos abordados:

- **Decision Tree**: estrutura de arvore com decisoes binarias, facil de interpretar
- **Logistic Regression**: apesar do nome, e para classificacao — estima probabilidades
- **K-Means**: algoritmo nao supervisionado que agrupa dados em K clusters
- **K-Nearest Neighbors (KNN)**: classifica pelo voto majoritario dos K vizinhos mais proximos
- **Grid Search**: busca exaustiva pelos melhores hiperparametros
- **Confusion Matrix**: matriz que mostra acertos e erros por classe

## 📚 Explicacao Detalhada

### Decision Tree

Algoritmo que divide os dados recursivamente baseado em regras (ex: "se idade < 30, va para a esquerda"). Cada no representa uma decisao, cada folha uma predicao.

```python
from sklearn.tree import DecisionTreeClassifier
modelo = DecisionTreeClassifier()
modelo.fit(X_train, y_train)
```

### Confusion Matrix

Matriz que compara predicoes com valores reais:

| | Predito Positivo | Predito Negativo |
|---|---|---|
| **Real Positivo** | TP (True Positive) | FN (False Negative) |
| **Real Negativo** | FP (False Positive) | TN (True Negative) |

- **Acuracia**: (TP + TN) / Total
- **Precision**: TP / (TP + FP)
- **Recall**: TP / (TP + FN)
- **F1-Score**: media harmonica de Precision e Recall

### Logistic Regression

Usa a funcao sigmoide para estimar probabilidades entre 0 e 1. Ideal para classificacao binaria.

### K-Means

Agrupa dados em K clusters minimizando a distancia dentro de cada cluster.

### KNN

Para classificar um novo ponto, olha os K vizinhos mais proximos e atribui a classe majoritaria.

### Grid Search

Testa combinacoes de hiperparametros e escolhe a melhor usando validacao cruzada.

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'max_depth': [3, 5, 10], 'min_samples_split': [2, 5, 10]}
grid = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5)
grid.fit(X_train, y_train)
```

## 💻 Exemplos

````python
# Exemplo 1: Decision Tree basica
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification

np.random.seed(42)
X, y = make_classification(n_samples=200, n_features=2, n_informative=2,
                           n_redundant=0, n_clusters_per_class=1, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

modelo = DecisionTreeClassifier(max_depth=3, random_state=42)
modelo.fit(X_train, y_train)

plt.figure(figsize=(12, 5))
plot_tree(modelo, filled=True, feature_names=['Feature 1', 'Feature 2'],
          class_names=['Classe 0', 'Classe 1'], rounded=True, fontsize=10)
plt.title('Arvore de Decisao (max_depth=3)')
plt.show()

print(f"Acuracia: {modelo.score(X_test, y_test):.4f}")
````

````python
# Exemplo 2: Confusion Matrix
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay

y_pred = modelo.predict(X_test)
cm = confusion_matrix(y_test, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Classe 0', 'Classe 1'])
disp.plot(cmap='Blues')
plt.title('Matriz de Confusao')
plt.show()

print("Relatorio de Classificacao:\n")
print(classification_report(y_test, y_pred, target_names=['Classe 0', 'Classe 1']))
````

````python
# Exemplo 3: Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

logreg = LogisticRegression(random_state=42)
logreg.fit(X_train_scaled, y_train)

y_pred_lr = logreg.predict(X_test_scaled)

print(f"Logistic Regression - Acuracia: {logreg.score(X_test_scaled, y_test):.4f}")
print(f"Coeficientes: {logreg.coef_[0]}")
print(f"Intercepto: {logreg.intercept_[0]:.4f}")

# Probabilidades
probabilidades = logreg.predict_proba(X_test_scaled[:3])
print(f"\nProbabilidades das primeiras 3 amostras:\n{probabilidades}")
````

````python
# Exemplo 4: K-Means (agrupamento nao supervisionado)
from sklearn.cluster import KMeans

np.random.seed(42)
# Gerar dados com 3 grupos naturais
X1 = np.random.normal(0, 1, (100, 2))
X2 = np.random.normal(5, 1, (100, 2))
X3 = np.random.normal(-5, 1, (100, 2))
X_cluster = np.vstack([X1, X2, X3])

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X_cluster)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.figure(figsize=(10, 6))
plt.scatter(X_cluster[:, 0], X_cluster[:, 1], c=labels, cmap='viridis', alpha=0.6, s=30)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='X',
            edgecolors='black', linewidth=2, label='Centroides')
plt.title('K-Means Clustering (K=3)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"Inercia: {kmeans.inertia_:.2f}")
````

````python
# Exemplo 5: Metodo do cotovelo (Elbow) para K-Means
inercias = []
K_range = range(1, 11)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_cluster)
    inercias.append(km.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K_range, inercias, 'bo-', linewidth=2)
plt.xlabel('Numero de Clusters (K)')
plt.ylabel('Inercia')
plt.title('Metodo do Cotovelo (Elbow Method)')
plt.axvline(x=3, color='red', linestyle='--', label=f'K=3 (cotovelo)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

````python
# Exemplo 6: K-Nearest Neighbors (KNN)
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)

y_pred_knn = knn.predict(X_test_scaled)

print(f"KNN (K=5) - Acuracia: {knn.score(X_test_scaled, y_test):.4f}")

# Testar diferentes valores de K
acus = []
for k in range(1, 21):
    knn_test = KNeighborsClassifier(n_neighbors=k)
    knn_test.fit(X_train_scaled, y_train)
    acus.append(knn_test.score(X_test_scaled, y_test))

plt.figure(figsize=(8, 5))
plt.plot(range(1, 21), acus, 'go-', linewidth=2)
plt.xlabel('Valor de K')
plt.ylabel('Acuracia')
plt.title('Efeito do K no KNN')
plt.grid(True, alpha=0.3)
plt.show()
````

````python
# Exemplo 7: Grid Search para otimizar hiperparametros
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid = GridSearchCV(DecisionTreeClassifier(random_state=42),
                    param_grid, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)

print(f"Melhores parametros: {grid.best_params_}")
print(f"Melhor acuracia (CV): {grid.best_score_:.4f}")
print(f"Acuracia no teste: {grid.score(X_test, y_test):.4f}")

# Comparar com modelo default
modelo_default = DecisionTreeClassifier(random_state=42)
modelo_default.fit(X_train, y_train)
print(f"\nModelo default - Acuracia: {modelo_default.score(X_test, y_test):.4f}")
````

````python
# Exemplo 8: Comparacao de algoritmos de classificacao
from sklearn.ensemble import RandomForestClassifier

modelos = {
    'Decision Tree': DecisionTreeClassifier(max_depth=5, random_state=42),
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=200),
    'KNN (K=5)': KNeighborsClassifier(n_neighbors=5),
    'Random Forest': RandomForestClassifier(n_estimators=50, random_state=42)
}

resultados = {}
for nome, modelo in modelos.items():
    modelo.fit(X_train_scaled, y_train)
    score = modelo.score(X_test_scaled, y_test)
    resultados[nome] = score

plt.figure(figsize=(10, 6))
barras = plt.bar(resultados.keys(), resultados.values(),
                color=['#3498db', '#e74c3c', '#2ecc71', '#9b59b6'])
plt.ylabel('Acuracia')
plt.title('Comparacao de Algoritmos de Classificacao')
plt.ylim(0, 1)

for barra, (nome, score) in zip(barras, resultados.items()):
    plt.text(barra.get_x() + barra.get_width()/2, barra.get_height() + 0.01,
             f'{score:.3f}', ha='center', fontweight='bold')

plt.show()

for nome, score in resultados.items():
    print(f"{nome:<25s}: {score:.4f}")
````

````python
# Exercicio 1: Classificacao de flores Iris com Decision Tree
from sklearn.datasets import load_iris

iris = load_iris()
X_iris, y_iris = iris.data, iris.target
X_tr, X_te, y_tr, y_te = train_test_split(X_iris, y_iris, test_size=0.3, random_state=42)

dt = DecisionTreeClassifier(max_depth=3, random_state=42)
dt.fit(X_tr, y_tr)

print(f"Acuracia: {dt.score(X_te, y_te):.4f}")
print(f"\nRelatorio:\n{classification_report(y_te, dt.predict(X_te), target_names=iris.target_names)}")
````

````python
# Exercicio 2: Encontrar K ideal para K-Means com elbow method
np.random.seed(42)
X_synth = np.vstack([
    np.random.normal([0, 0], 1, (100, 2)),
    np.random.normal([5, 5], 1, (100, 2)),
    np.random.normal([10, 0], 1, (100, 2)),
    np.random.normal([5, -5], 1, (100, 2))
])

inercias = []
for k in range(1, 11):
    km = KMeans(n_clusters=k, n_init=10, random_state=42)
    km.fit(X_synth)
    inercias.append(km.inertia_)

deltas = np.diff(inercias)
deltas2 = np.diff(deltas)
k_otimo = np.argmax(deltas2) + 2

print(f"K otimo estimado pelo metodo do cotovelo: {k_otimo}")
print(f"Inercias: {[f'{i:.0f}' for i in inercias]}")
````

## 📌 Resumo

- **Decision Tree**: arvore de decisoes binarias, interpretavel, propensa a overfitting
- **Confusion Matrix**: TP, TN, FP, FN — base para metricas de classificacao
- **Acuracia**: (TP+TN)/Total — pode enganar em classes desbalanceadas
- **Precision**: TP/(TP+FP) — quantos positivos preditos sao realmente positivos
- **Recall**: TP/(TP+FN) — quantos positivos reais foram encontrados
- **Logistic Regression**: funcao sigmoide para probabilidades, classificacao binaria
- **K-Means**: agrupamento nao supervisionado, escolher K com metodo do cotovelo
- **KNN**: classifica pelo voto dos K vizinhos mais proximos
- **Grid Search**: busca exaustiva de hiperparametros com `GridSearchCV`

## 💡 Curiosidades

- Decision Trees sao a base de algoritmos poderosos como **Random Forest** e **XGBoost**, que estao entre os melhores para dados tabulares.
- K-Means foi proposto por **Hugo Steinhaus** em 1956 e popularizado por **James MacQueen** em 1967.
- KNN e um dos algoritmos mais simples de ML: nao ha "treinamento" — apenas armazena os dados e compara distancias.
- Regressao Logistica foi desenvolvida pelo estatistico **David Cox** em 1958.
- Grid Search pode ser computacionalmente caro — uma alternativa e o **RandomizedSearchCV** que testa combinacoes aleatorias.

## ✅ Boas Praticas

- Escale os dados para Logistic Regression e KNN (sensiveis a escala)
- Use `random_state` para reprodutibilidade
- Sempre compare com um baseline (ex: classificador que chuta a classe majoritaria)
- Para classes desbalanceadas, use F1-score em vez de acuracia
- Limite a profundidade da Decision Tree (`max_depth`) para evitar overfitting
- Visualize a arvore (`plot_tree`) para entender as decisoes do modelo

## ⚠️ Erros Comuns

| Erro | Exemplo | Correcao |
|------|---------|----------|
| Usar acuracia em classes desbalanceadas | 95% de acuracia mas nunca detecta a classe rara | Usar F1-score ou balanced accuracy |
| Nao escalar para KNN/LogReg | Features em escalas diferentes distorcem distancia | Usar StandardScaler |
| Overfitting na Decision Tree | `max_depth=None` pode criar arvore gigante | Limitar profundidade e minimo de amostras |
| K-Means com K errado | Forcar 2 clusters onde ha 4 grupos naturais | Usar metodo do cotovelo ou silhouette score |
| KNN com K=1 | Muito sensivel a ruido | Usar K >= 3 e impar para evitar empates || Grid Search sem CV | Overfitting nos hiperparametros | Usar `cv >= 5` em GridSearchCV |
