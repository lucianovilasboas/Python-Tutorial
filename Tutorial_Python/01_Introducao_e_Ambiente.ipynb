## 📝 Introdução

Python é uma linguagem de programação criada por **Guido van Rossum** em 1991. Ela foi projetada com foco em **legibilidade** e **simplicidade**, permitindo que programadores escrevam código de forma clara e intuitiva.

Python é usada em diversas áreas:

- **Desenvolvimento Web** (Django, Flask)
- **Ciência de Dados** (Pandas, NumPy)
- **Inteligência Artificial e Machine Learning**
- **Automação de tarefas**
- **Desenvolvimento de jogos** (Pygame)
- **Aplicações desktop**

> **Curiosidade**: Python foi nomeada em homenagem ao grupo de comédia britânico *Monty Python's Flying Circus*, não à cobra píton!

## 📚 Explicação Detalhada

### 1. O que é Python?

Python é uma linguagem:
- **Interpretada**: o código é executado linha por linha
- **De alto nível**: próxima da linguagem humana
- **Tipada dinamicamente**: você não precisa declarar o tipo das variáveis
- **Multiparadigma**: suporta programação procedural, orientada a objetos e funcional

### 5. Executando Python

Existem duas formas principais de executar código Python:

**Terminal interativo:**
- Digite `python` no terminal
- O interpretador aguarda comandos
- Ideal para testar pequenos trechos

**Script (.py):**
- Crie um arquivo com extensão `.py`
- Execute com `python nome_do_arquivo.py`
- Ideal para programas completos

## 🏆 Desafios

2. **Desafio da Arte ASCII:** Use print() para criar um desenho usando caracteres:

```
    /\___/\

```

3. **Desafio da Calculadora de Viagem:** Calcule quantos litros de combustível serão necessários para uma viagem de 350 km, sabendo que o carro faz 12 km/l. Exiba o resultado formatado.

````python
# 🏆 Desafio 1: Cartão de Visita
print("=========================")
print("|   NOME: Maria Souza   |")
print("|   IDADE: 20 anos      |")
print("|   CIDADE: São Paulo   |")
print("=========================")
````

````python
# 🏆 Desafio 2: Arte ASCII
print("    /\\___/\\")
print("   (  o o  )")
print("   (  =^=  )")
print("    (---)")
````

````python
# 🏆 Desafio 3: Calculadora de Viagem
distancia = 350
consumo = 12
litros = distancia / consumo
print(f"Para uma viagem de {distancia} km, serão necessários {litros:.2f} litros de combustível.")
````

## ⚠️ Erros Comuns

| Erro | Exemplo | Como Evitar |
|------|---------|-------------|
| Esquecer aspas | `print(Ola)` | Sempre use `'` ou `"` ao redor de texto |
| Esquecer parênteses | `print "Olá"` | Use `print("...")` com parênteses |
| Misturar tipos de aspas | `print('Olá")` | Use o mesmo tipo dos dois lados |
| Erro de digitação | `prnit("Olá")` | Verifique a ortografia do código |
| Arquivo não encontrado | Executar script sem estar no diretório correto | Use `cd` para navegar até a pasta do arquivo |
| PATH não configurado | `python` não é reconhecido | Reinstale marcando "Add Python to PATH" |
